시계열 자료 중에는 시차가 증가해도 자기 상관계수가 비교적 오랫동안 0이 아닌 값으로 남아 있는 것들이 많다. 

#### 이러한 시계열에 MA 모형을 사용하면 차수가 너무 커진다. 이를 극복하기 위한 모형 중의 하나가 Autoregressive (AR) 모형이다. 

AR 모형은 백색 잡음의 현재값과 자기 자신의 과거값의 선형 가중합으로 이루어진 정상 확률 모형이다.

$$Y_t = -\phi_1 Y_{t-1} -\phi_2 Y_{t-2} -\cdots -\phi_p Y_{t-p} + \epsilon_t$$

자기 자신의 과거 값을 사용하기 때문에 Autoregressive 라는 이름이 붙여졌다. AR 모형은 임의의 계수 값을 가질 수 있는 MA 모형과 달리 정상 과정이 되기 위해서는 계수의 값에 제한 조건이 붙는다. 

구체적인 제한 조건은 차수에 따라 달라진다. AR 모형은 얼핏 보기에 일반 선형 확률 과정(general linear process)의 수식을 따르지 않는 것처럼 보이지만 일반 선형 확률 과정의 형태로 변환이 가능하다.

## AR(1) 모형

1차 AR 모형 즉, AR(1) 모형은 백색 잡음의 현재값과 1스텝 과거의 자기 자신의 값만의 가중합으로 이루어진 모형이다. 수식은 다음과 같다.

$$Y_t = -\phi Y_{t-1} + \epsilon_t$$

이 수식에서 계수  𝜙 는 다음 조건을 만족해야 한다.

$$-1 < \phi < 1$$

AR(1) 모형은 다음과 같이 일반 선형 확률 과정의 형태로 바꿀 수 있다.

\begin{aligned}
Y_t 
&= -\phi Y_{t-1} + \epsilon_t \\
&= -\phi \left( -\phi Y_{t-2} + \epsilon_{t-1} \right) + \epsilon_t \\
&= \phi^2 Y_{t-2} -\phi \epsilon_{t-1} + \epsilon_t \\
&= \phi^2  \left( -\phi Y_{t-3} + \epsilon_{t-2} \right) -\phi \epsilon_{t-1} + \epsilon_t \\
&= -\phi^3 Y_{t-3} + \phi^2 \epsilon_{t-2} -\phi \epsilon_{t-1} + \epsilon_t \\
&\vdots& \\
&= \epsilon_t -\phi \epsilon_{t-1} +\phi^2 \epsilon_{t-2} -\phi^3 \epsilon_{t-3} + \cdots  \\
\end{aligned}

AR(1) 모형의 기대값은 MA모형과 마찬가지로 0이다.

$$\text{E}[Y_t] = 0$$

이는 다음과 같이 증명할 수 있다.

\begin{aligned}
\text{E}[Y_t] = \mu 
&= \text{E} \left[ -\phi  Y_{t-1} - \epsilon_{t-1} \right] \\
&= -\phi\text{E} [ Y_{t-1} ] -\text{E} [ \epsilon_{t-1} ] \\
&= -\phi \mu -   0 \\ 
(1+\phi)\mu &= 0 \\
\mu &= 0 \; \text{ if } \phi \neq -1 \\
\end{aligned}

AR(1) 모형의 분산은 다음과 같다.

$$\gamma_0 = \dfrac{\sigma_\epsilon^2}{1-\phi^2}$$

이는 다음과 같이 증명할 수 있다.

우선  𝜖𝑡 는 과거의 값  𝑌𝑡−𝑙  (for  𝑙>1 )에 영향을 주지 않기 때문에 독립이다. 따라서

$$\text{E}[\epsilon_tY_{t-l}] = \text{E}[\epsilon_t]\text{E}[Y_{t-l}] = 0$$

이다.

\begin{aligned}
\text{Var}[Y_t] = \gamma_0 = \text{E}[Y_t^2] 
&= \text{E} \left[ (-\phi  Y_{t-1} - \epsilon_{t})^2 \right] \\
&= \text{E} \left[ \phi^2  Y_{t-1}^2 + 2\phi Y_{t-1} \epsilon_{t-1} + \epsilon_{t}^2 \right] \\
&= \phi^2\text{E} [ Y_{t-1}^2 ] + 2 \phi \text{E} [ Y_{t-1} \epsilon_{t} ] + \text{E} [ \epsilon_{t}^2 ] \\
&= \phi^2 \gamma_0 + 0 + \sigma_\epsilon^2
\end{aligned}

AR(1) 모형의 자기공분산은 다음과 같다.

$$\gamma_l = (-\phi)^l \dfrac{\sigma_\epsilon^2}{1-\phi^2}$$

자기공분산으로부터 자기상관계수는 다음과 같다.

$$\rho_l = \dfrac{\gamma_l}{\gamma_0} = (-\phi)^l$$

1과 -1사이의 여러가지  𝜙 에 대해 AR(1) 모형의 자기상관계수 함수 즉, ACF(이론적 자기상관계수)를 그리면 다음과 같다.

+ 𝜙<0  이면 ACF는 지수함수적으로 감소한다.
+ 𝜙>0  이면 ACF는 부호를 바꿔가면서 (진동하면서) 지수함수적으로 감소한다.

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
lag = np.arange(12)

plt.subplot(221)
acf = 0.9 ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-0.1, 1.1)
plt.xlabel("lag")
plt.ylabel("자기상관계수")
plt.title("$\phi=0.9$")

plt.subplot(222)
acf = 0.4 ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-0.1, 1.1)
plt.xlabel("lag")
plt.ylabel("자기상관계수")
plt.title("$\phi=0.4$")

plt.subplot(223)
acf = (-0.8) ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-1.1, 1.1)
plt.xlabel("lag")
plt.ylabel("자기상관계수")
plt.title("$\phi=-0.8$")

plt.subplot(224)
acf = (-0.5) ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-1.1, 1.1)
plt.xlabel("lag")
plt.ylabel("자기상관계수")
plt.title("$\phi=-0.5$")

plt.suptitle("AR(1) 모형의 자기상관계수 함수 사례", y=1.1)
plt.tight_layout()
plt.show()
```

앞의 그림에서 볼 수 있듯이  𝜙 의 절대값이 1에 가까울 수록 자기상관계수의 감소 속도가 작아진다. 

#### 즉, 시차가 증가해도 상관계수가 크게 작아지지 않는다. 

그럼 만약  𝜙 의 절대값이 1이거나 1보다 크다면 어떻게 될까?

#### 𝜙 값의 크기가 1 이상이 되면 분산이 계속 증가하는 비정상 과정이 된다.  𝜙=−1 인 경우를 보자.  (분산이 시간에 따라 변하기 때문)

\begin{aligned}
Y_t &= Y_{t-1} + \epsilon_t \\
\text{Var}[Y_t] &= \text{Var}[Y_{t-1} + \epsilon_t] \\
 &= \text{Var}[Y_{t-1}] + \text{Var}[\epsilon_t]  \;\; (\text{independence})\\
\end{aligned}

$$\text{Var}[Y_t] > \text{Var}[Y_{t-1}]$$

AR(1) 모형의 경우  |𝜙|<1  조건은 정상상태 조건(stationarity condition)이 된다.

### AR(1) 모형의 시뮬레이션

```python
import statsmodels.api as sm

np.random.seed(0)
p1 = sm.tsa.ArmaProcess([1, -0.9], [1])
y1 = p1.generate_sample(100, burnin=100)
plt.plot(y1, 'o-')
plt.title("1차 AR 모형의 시뮬레이션")
plt.show()
```

이 시계열 데이터의  𝑙 -시차 상관계수, 즉  𝑌𝑡 와  𝑌𝑡−𝑙 의 상관계수를 조사하기 위해 스캐터플롯을 그려보았다. 

#### 지속적으로 강한 상관관계를 보인다.

```python
import scipy as sp

plt.figure(figsize=(10, 4))

plt.subplot(131)
r, p = sp.stats.pearsonr(y1[1:], y1[:-1])
sns.scatterplot(y1[1:], y1[:-1])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-1}$")
plt.title("시차가 1인 경우의 상관계수\nr={0:.3f}, p-value={1:.3f}".format(r, p))

plt.subplot(132)
r, p = sp.stats.pearsonr(y1[2:], y1[:-2])
sns.scatterplot(y1[2:], y1[:-2])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-2}$")
plt.title("시차가 2인 경우의 상관계수\nr={0:.3f}, p-value={1:.3f}".format(r, p))

plt.subplot(133)
r, p = sp.stats.pearsonr(y1[3:], y1[:-3])
sns.scatterplot(y1[3:], y1[:-3])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-3}$")
plt.title("시차가 3인 경우의 상관계수\nr={0:.3f}, p-value={1:.3f}".format(r, p))

plt.tight_layout()
plt.show()
```

이론적 자기상관계수 함수와 표본 자기상관계수 함수를 그리면 다음과 같다.

```python
plt.subplot(211)
plt.stem(p1.acf(11), use_line_collection=True)  # 이론적 자기상관계수 구하기
plt.xlim(-1, 11)
plt.ylim(-0.4, 1.1)
plt.title("1차 AR 모형의 이론적 자기상관계수 함수")

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(y1, lags=10, ax=ax)
plt.xlim(-1, 11)
plt.ylim(-0.4, 1.1)
plt.title("1차 AR 모형의 표본 자기상관계수 함수")

plt.tight_layout()
plt.show()
```

### AR(2) 모형의 시뮬레이션

```python
np.random.seed(0)
p2 = sm.tsa.ArmaProcess([1, -1, 0.6], [1])
y2 = p2.generate_sample(300)
plt.plot(y2, 'o-')
plt.title("2차 AR 모형의 시뮬레이션")
plt.show()
```

상관관계를 위한 스캐터플롯, 이론적 자기상관계수 함수와 표본 자기상관계수 함수를 그리면 다음과 같다.

```python
plt.figure(figsize=(10, 4))

plt.subplot(131)
r, p = sp.stats.pearsonr(y2[1:], y2[:-1])
sns.scatterplot(y2[1:], y2[:-1])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-1}$")
plt.title("시차가 1인 경우의 상관계수\n(r={0:.3f}, p-value={1:.3f})".format(r, p))

plt.subplot(132)
r, p = sp.stats.pearsonr(y2[2:], y2[:-2])
sns.scatterplot(y2[2:], y2[:-2])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-2}$")
plt.title("시차가 2인 경우의 상관계수\n(r={0:.3f}, p-value={1:.3f})".format(r, p))

plt.subplot(133)
r, p = sp.stats.pearsonr(y2[3:], y2[:-3])
sns.scatterplot(y2[3:], y2[:-3])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-3}$")
plt.title("시차가 3인 경우의 상관계수\n(r={0:.3f}, p-value={1:.3f})".format(r, p))

plt.tight_layout()
plt.show()
```

자기상관계수는 시차 1일 경우 0.634이다가 시차 2인 경우에는 거의 제로로 나오고 다시 시차가 3인 경우 상관계수가 0.347이 나온다. 진동하는 것처럼 보인다.

```python
plt.subplot(211)
plt.stem(p2.acf(11), use_line_collection=True)
plt.xlim(-1, 11)
plt.ylim(-1.1, 1.1)
plt.title("2차 AR 모형의 이론적 자기상관계수 함수")

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(y2, lags=10, ax=ax)
plt.xlim(-1, 11)
plt.ylim(-1.1, 1.1)
plt.title("2차 AR 모형의 표본 자기상관계수 함수")

plt.tight_layout()
plt.show()
```
