#### ÏãúÍ≥ÑÏó¥  ùëåùë° ÏùÑ Ï∞®Î∂Ñ(difference)Ìïú Í≤∞Í≥ºÎ°ú ÎßåÎì§Ïñ¥ÏßÑ ÏãúÍ≥ÑÏó¥  ‚àáùëåùë°=ùëåùë°‚àíùëåùë°‚àí1 Ïù¥ ARMA Î™®ÌòïÏùÑ Îî∞Î•¥Î©¥ ÏõêÎûòÏùò ÏãúÍ≥ÑÏó¥  ùëåùë° Î•º ARIMA(Autoregressive Integrated Moving Average) Î™®ÌòïÏù¥ÎùºÍ≥† ÌïúÎã§.

ÎßåÏïΩ  ùëë Î≤à Ï∞®Î∂ÑÌïú ÌõÑÏóêÏïº ÏãúÍ≥ÑÏó¥  ‚àáùëëùëåùë° Í∞Ä ARMA(p,q) Î™®ÌòïÏùÑ Îî∞Î•∏Îã§Î©¥ Ï†ÅÎ∂Ñ Ï∞®Ïàò(order of integration)Í∞Ä  ùëë Ïù∏ ARIMA Î™®ÌòïÏúºÎ°ú ARIMA(p, d, q)Î°ú ÌëúÍ∏∞ÌïúÎã§.  ùëû=0 Ïù∏ Í≤ΩÏö∞ÏóêÎäî ARI(p,d),  ùëù=0 Ïù∏ Í≤ΩÏö∞ÏóêÎäî IMA(d,q)Î°ú ÌëúÍ∏∞ÌïúÎã§.



## Îã®ÏúÑÍ∑º ÌäπÏÑ±

#### ARIMA(p, 1, q) Î™®ÌòïÏùÄ ÌäπÏÑ± Î∞©Ï†ïÏãù(characteristic equation)Ïù¥  ùë•=1 Ïù¥ÎùºÎäî Îã®ÏúÑÍ∑º(unit root)Î•º Í∞ÄÏßÑÎã§. 

Ïù¥ ÌäπÏÑ±ÏúºÎ°ú Ïù∏Ìï¥ ÌôïÎ•† Í≥ºÏ†ïÏù¥ ARIMA(p, 1, q)Ïóê Ìï¥ÎãπÌïòÎäîÏßÄ ÌôïÏù∏ÌïòÎäî Í≤ÄÏ†ï Î∞©Î≤ïÏùÑ ÌÜµÌãÄÏñ¥ Îã®ÏúÑÍ∑º Í≤ÄÏ†ï(unit root test)Ïù¥ÎùºÍ≥†ÎèÑ ÌïúÎã§.

Îã®ÏúÑÍ∑º ÌäπÏÑ±ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÌôïÏù∏ÌïúÎã§. ARIMA(p,1,q) Î™®ÌòïÏùÑ Ï∞®Î∂ÑÌïú Í∞íÏùÑ  ùëäùë°=ùëåùë°‚àíùëåùë°‚àí1 ÎùºÍ≥† ÌïòÏûê.  ùëäùë° Îäî ARMA(p,q) Î™®ÌòïÏù¥ÎØÄÎ°ú Îã§ÏùåÍ≥º Í∞ôÏù¥ ÌëúÌòÑÌï† Ïàò ÏûàÎã§.

$$W_t + \phi_1 W_{t-1} + \phi_2 W_{t-2} + \cdots + \phi_p W_{t-p} = \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q}$$

Ïù¥Î•º Îã§Ïãú Ïõê ÏãúÍ≥ÑÏó¥  ùëåùë° Î°ú ÌëúÌòÑÌïòÎ©¥,

$$Y_t-Y_{t-1} + \phi_1 (Y_{t-1}-Y_{t-2}) + \phi_2 (Y_{t-2}-Y_{t-3}) + \cdots + \phi_p (Y_{t-p}-Y_{t-p-1}) = \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q}$$

$$Y_t + (-1+\phi_1)Y_{t-1}+(\phi_2-\phi_1)Y_{t-2}+\cdots+(\phi_p-\phi_{p-1})Y_{t-p}-\phi_pY_{t-p-1} = \epsilon_t+\theta_1\epsilon_{t-1}+\theta_2 \epsilon_{t-2}+\cdots+\theta_q \epsilon_{t-q}$$

ÏúÑ Î™®ÌòïÏóê ÎåÄÌïú ÌäπÏÑ± Î∞©Ï†ïÏãùÏùÄ

$$1+(-1+\phi_1)x+(\phi_2-\phi_1)x^2+\cdots+(\phi_p-\phi_{p-1})x^p-\phi_px^{p+1} = 0$$

Ïù¥Î•º Ïù∏ÏàòÎ∂ÑÌï¥ÌïòÎ©¥

$$(1-x)(1+\phi_1x+\phi_2x^2+\cdots+\phi_px^p) = 0$$

Ïù¥Îã§. Ï¶â ÌäπÏÑ± Î∞©Ï†ïÏãùÏù¥ Îã®ÏúÑÍ∑º( ùë•=1 )ÏùÑ Í∞ÄÏßÑÎã§.

Ïù¥Î≤àÏóêÎäî Ï∞®Î∂ÑÌïú ARMA(p,q)Î™®ÌòïÏù¥ ÏïÑÎãàÎùº Ï∞®Î∂ÑÌïòÏßÄ ÏïäÏùÄ ÌòïÌÉúÏùò ARMA(p,q)Î™®Ìòï

$$Y_t = \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} + \epsilon_t - \theta_1 \epsilon_{t-1}  - \theta_2 \epsilon_{t-2} - \cdots  - \theta_q \epsilon_{t-q}$$

ÏùÑ ÏÇ¥Ìé¥Î≥¥Ïûê.

Ïù¥ Î™®ÌòïÏùò ÌäπÏÑ±Î∞©Ï†ïÏãùÏùÄ

$$\phi_p x^p + \cdots + \phi_1 x - 1 = 0$$

Ïù¥Í≥† Ïù¥ Î∞©Ï†ïÏãùÏùò Ìï¥Í∞Ä  ùë•=1 Ïù¥ÎØÄÎ°ú ÎåÄÏûÖÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Í¥ÄÍ≥ÑÎ•º ÏñªÏùÑ Ïàò ÏûàÎã§.

$$\sum_{i=1}^p \phi_i = 1$$

## IMA(1,1) Î™®Ìòï

ARIMA Î™®ÌòïÏùò Í∞ÄÏû• Îã®ÏàúÌïú ÌòïÌÉúÏù∏ IMA(1,1)ÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.  Ï¶â, p=0Ïù¥Í≥† 1Î≤à Ï∞®Î∂ÑÌïú Î™®Ìòï

$$Y_t = Y_{t-1} + \epsilon_t - \theta \epsilon_{t-1}$$

Ïù¥ ÏãùÏùÑ ÏùºÎ∞ò ÏÑ†Ìòï ÌôïÎ•† Í≥ºÏ†ïÏúºÎ°ú ÌëúÌòÑÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$Y_t = \epsilon_t+(1-\theta)\epsilon_{t-1}+(1-\theta)\epsilon_{t-2}+(1-\theta)\epsilon_{t-3}+\cdots$$

Ï¶â Í≥ºÍ±∞Ïùò Î∞±ÏÉâ Ïû°ÏùåÏùÑ ÎàÑÏ†Å(cumulation)Ìïú Í≤ÉÏù¥ÎùºÍ≥† Î≥º Ïàò ÏûàÎã§.

IMA(1,1) Î™®ÌòïÏùò ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÎäî ÏãúÏ∞®(lag)Ïóê ÏÉÅÍ¥ÄÏóÜÏù¥ Í±∞Ïùò  1 Ïù¥Îã§. 

#### Ï¶â, Ï†ïÏÉÅÍ≥ºÏ†ïÏ≤òÎüº ÏãúÏ∞®Í∞Ä Ï¶ùÍ∞ÄÌï¥ÎèÑ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä Í∞êÏÜåÌïòÏßÄ ÏïäÎäîÎã§.

$$\rho_l = \text{corr}[Y_t,Y_{t-l}] \approx 1$$

Ïù¥Î†áÍ≤å ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä Îπ†Î•¥Í≤å Í∞êÏÜåÌïòÏßÄ ÏïäÎäî Í≤ÉÏù¥ ARIMAÏôÄ Í∞ôÏùÄ Ï†ÅÎ∂Ñ Í≥ºÏ†ï(integrated process)Ïùò ÌäπÏßïÏù¥Îã§.

## IMA(2,1) Î™®Ìòï

Îã§ÏùåÍ≥º Ï†ÅÎ∂Ñ Ï∞®ÏàòÍ∞Ä 2Ïù∏ IMA(2,1)ÏùÑ ÏÇ¥Ìé¥Î≥¥Ïûê

$$\nabla^2 Y_t = \epsilon_t - \theta \epsilon_{t-1}$$

Ïù¥ Î™®ÌòïÏùÑ ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïòÏó¨ ACFÎ•º ÏÇ¥Ìé¥Î≥¥Î©¥ ÏïÑÎûòÏôÄ Í∞ôÎã§. statsmodelsÎäî ARIMA Î™®ÌòïÏùÑ ÏúÑÌïú Î≥ÑÎèÑÏùò ÌÅ¥ÎûòÏä§Í∞Ä ÏóÜÍ∏∞ ÎïåÎ¨∏Ïóê ArmaProcessÎ•º ÏÇ¨Ïö©Ìïú ÌõÑ ÎàÑÏ†ÅÌï©ÏùÑ Íµ¨ÌïòÎäî Î∞©ÏãùÏúºÎ°ú ÏãúÎÆ¨Î†àÏù¥ÏÖò ÌïúÎã§.

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)

import statsmodels.api as sm
```

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1], [1, 0.6])
y2 = p.generate_sample(100).cumsum().cumsum()  # ARIMA d=2
y1 = np.diff(y2) # ÎØ∏Î∂Ñ
y0 = np.diff(y1)

plt.subplot(311)
plt.title("ARIMA Î™®ÌòïÏùò ÏãúÎÆ¨Î†àÏù¥ÏÖò")
plt.plot(y2, "o-")

plt.subplot(312)
plt.plot(y1, "o-")

plt.subplot(313)
plt.plot(y0, "o-")

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.show()
```

ÏúÑÏùò Í∞Å Î™®ÌòïÏóê ÎåÄÌïú ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò Íµ¨ÌïòÍ∏∞

```python
ax1 = plt.subplot(311)
sm.graphics.tsa.plot_acf(y2, ax=ax1)
ax2 = plt.subplot(312)
sm.graphics.tsa.plot_acf(y1, ax=ax2)
ax3 = plt.subplot(313)
sm.graphics.tsa.plot_acf(y0, ax=ax3)

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.show()
```

ARIMA Î™®ÌòïÏùò Í≤ΩÏö∞, ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä ÏÑúÏÑúÌûà Í∞êÏÜåÌïòÍ≥† ÏûàÏúºÎÇò, ÎßàÏßÄÎßâ Ï∞®Î∂ÑÏùÑ ÎëêÎ≤àÌïú Î™®ÌòïÏùò Í≤ΩÏö∞Îäî ÏßÑÎèôÌïòÎäî ÌòïÌÉúÎ•º Î≥¥Ïù¥Í≥† ÏûàÎã§.

## Í≥ºÏ∞®Î∂Ñ

ÏãúÍ≥ÑÏó¥ ÏûêÎ£åÎ•º ÏõêÎûòÏùò Ï†ÅÎ∂Ñ Ï∞®Ïàò Ïù¥ÏÉÅÏúºÎ°ú Ï∞®Î∂ÑÌïòÎäî Í≤ÉÏùÑ Í≥ºÏ∞®Î∂Ñ(over-differencing)Ïù¥ÎùºÍ≥† ÌïúÎã§.

ÏòàÎ•º Îì§Ïñ¥ ARIMA(0,d,0) Î™®ÌòïÏùò ÏãúÍ≥ÑÏó¥ÏùÑ Ï†ÅÎ∂Ñ Ï∞®ÏàòÏóê ÎßûÍ≤å Ï∞®Î∂ÑÌïòÎ©¥ ÏµúÏ¢ÖÏ†ÅÏúºÎ°ú Î∞±ÏÉâÏû°ÏùåÏù¥ ÎêúÎã§. ÎßåÏïΩ Ïó¨Í∏∞ÏóêÏÑú Ï∂îÍ∞ÄÏ†ÅÏúºÎ°ú Ï∞®Î∂ÑÏùÑ ÌïòÎ©¥ Î∞±ÏÉâ Ïû°ÏùåÏóê ÎåÄÌïú MA(1) Î™®ÌòïÏù¥ ÎêòÏñ¥ Î≤ÑÎ¶∞Îã§. Ìïú Î≤à Îçî Ï∞®Î∂ÑÌïúÎã§Î©¥ Ïù¥Î≤àÏóêÎäî MA(2) Î™®ÌòïÏù¥ ÎêúÎã§.

$$\nabla^d Y_t = \epsilon_t$$

$$\nabla^{d+1} Y_t = \epsilon_t - \epsilon_{t-1}$$

$$\nabla^{d+2} Y_t = \epsilon_t - 2\epsilon_{t-1} + \epsilon_{t-2}$$

Ï¶â Î™®Ìòï Ï∞®ÏàòÍ∞Ä Î∂àÌïÑÏöîÌïòÍ≤å Ï¶ùÍ∞ÄÌïòÍ≤å ÎêòÎØÄÎ°ú ACFÎ•º Î≥¥Î©¥ ÌÅ∞ ÏãúÏ∞®Ïóê ÎåÄÌïú ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞íÏù¥ ÎÇ®ÏïÑÏûàÎäî Í≤ÉÏùÑ Î≥º Ïàò ÏûàÎã§.

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1], [1, 0.6])
y2 = p.generate_sample(100, burnin=100)  # Ï†ÅÎ∂Ñ ÏïàÌïú ÏÉÅÌÉú
y1 = np.diff(y2)
y0 = np.diff(y1)

ax1 = plt.subplot(311)
sm.tsa.graphics.plot_acf(y2, lags=10, ax=ax1)
ax2 = plt.subplot(312)
sm.tsa.graphics.plot_acf(y1, lags=10, ax=ax2)
ax3 = plt.subplot(313)
sm.tsa.graphics.plot_acf(y0, lags=10, ax=ax3)

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.show()
```

## Dickey-Fuller Îã®ÏúÑÍ∑º Í≤ÄÏ†ï

DF Í≤ÄÏ†ï(Dickey-Fuller Îã®ÏúÑÍ∑º Í≤ÄÏ†ï)Ïùò ÌöåÍ∑ÄÎ∂ÑÏÑùÏùò Í≤∞Í≥ºÎ°ú ÎÇòÌÉÄÎÇòÎäî Í∞ÄÏ§ëÏπò Í≥ÑÏàòÎ•º Í≤ÄÏ†ïÌÜµÍ≥ÑÎüâÏúºÎ°ú ÏÇ¨Ïö©ÌïúÎã§.

ÏãúÍ≥ÑÏó¥  ùëåùë° Ïù¥ AR(p)Î™®ÌòïÏù¥Î©¥

$$Y_t + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} = \epsilon_t$$

Ïù¥Í≥† Ïù¥ ÏãùÏùÑ Ï†ïÎ¶¨ÌïòÎ©¥

\begin{eqnarray}
Y_t 
&=& \sum_{i=1}^p \phi_i Y_{t-1} - \sum_{i=2}^p\phi_2 (Y_{t-1} - Y_{t-2}) - \cdots - \phi_p (Y_{t-p+1} - Y_{t-p}) + \epsilon_t \\
&=& a_0 Y_{t-1} + a_1 \nabla Y_{t-1} + \cdots + a_p \nabla Y_{t-p} + \epsilon_t \\
\end{eqnarray}

Ïù¥Îã§. ÏñëÎ≥ÄÏóêÏÑú  ùëåùë°‚àí1 ÏùÑ ÎπºÎ©¥

$$\nabla Y_t = (a_0 - 1) Y_{t-1} + a_1 \nabla Y_{t-1} + a_2 \nabla Y_{t-2} + \cdots + a_p \nabla Y_{t-p} + \epsilon_t$$

ÏùÑ ÏñªÏùÑ Ïàò ÏûàÎã§.

Îî∞ÎùºÏÑú  ‚àáùëåùë° Î•º Ï¢ÖÏÜçÎ≥ÄÏàòÎ°ú, Ïõê ÏãúÍ≥ÑÏó¥  ùëåùë° ÏôÄ Ï∞®Î∂ÑÍ∞íÏùò ÏßÄÏó∞ Í∞íÎì§  ‚àáùëåùë°‚àí1,‚àáùëåùë°‚àí2,‚ãØ,‚àáùëåùë°‚àíùëô ÏùÑ ÎèÖÎ¶ΩÎ≥ÄÏàòÎ°ú ÌöåÍ∑ÄÎ∂ÑÏÑùÏùÑ ÌïòÎ©¥  ùëé0,‚Ä¶,ùëéùëù Î•º Íµ¨Ìï† Ïàò ÏûàÎã§.

Ï†ÅÎ∂ÑÏ∞®ÏàòÍ∞Ä 1Ïù¥ÏÉÅÏù¥Î©¥

$$a_0 - 1 = \sum_{i=1}^p \phi_i - 1 = 0$$

Ïù¥ÎØÄÎ°ú Ïù¥ Í∞íÏù¥ Í≤ÄÏ†ï ÌÜµÍ≥ÑÎüâÏù¥ ÎêúÎã§. 

Ïó¨Í∏∞ÏóêÏÑúÎäî AR(p)Î™®ÌòïÏùÑ ÏÇ¨Ïö©ÌïòÏòÄÏßÄÎßå ARMA(p,q)Î™®ÌòïÏóêÏÑúÎèÑ Î∞±ÏÉâ Ïû°ÏùåÏùÑ Î≥ÄÌòïÌïòÎ©¥ Í∞ôÏùÄ Í≤∞Í≥ºÎ•º ÏñªÏùÑ Ïàò ÏûàÎã§.

DF Í≤ÄÏ†ïÏùò Í≤ÄÏ†ï ÌÜµÍ≥ÑÎüâ  ùëé Îäî ÏùºÎ∞òÏ†ÅÏù∏ Í≤ÄÏ†ïÍ≥º Îã¨Î¶¨ Ï†ïÍ∑úÎ∂ÑÌè¨ Îì±Ïùò Í∞ÑÎã®Ìïú ÏàòÏãùÏúºÎ°ú ÎÇòÌÉÄÎÇòÏßÄ ÏïäÍ≥† ÎØ∏Î¶¨ Í≥ÑÏÇ∞Îêú ÌëúÎ•º ÏÇ¨Ïö©ÌïòÍ±∞ÎÇò Í∑ºÏÇ¨ÏãùÏùÑ ÏÇ¨Ïö©ÌïúÎã§.

## ADF Í≤ÄÏ†ï

Augmented Dickey-Fuller (ADF) Í≤ÄÏ†ïÏùÄ 1Ï∞® ÎàÑÏ†ÅÏóê ÏùòÌïú ÌôïÎ•†Ï†Å Ï∂îÏÑ∏Îøê ÏïÑÎãàÎùº 2Ï∞®Ìï≠ÏúºÎ°ú ÎÇòÌÉÄÎÇòÎäî Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏Î•º Ìè¨Ìï®ÌïòÎäî ÏãúÍ≥ÑÏó¥Ïóê ÎåÄÌï¥ÏÑú Îã®ÏúÑ Í∑º Í≤ÄÏ†ïÏùÑ Ìï† Ïàò ÏûàÎèÑÎ°ù DF Í≤ÄÏ†ïÏùÑ ÏùºÎ∞òÌôî Ìïú Í≤ÉÏù¥Îã§.

ADF Í≤ÄÏ†ïÎèÑ DF Í≤ÄÏ†ïÍ≥º ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú ÌöåÍ∑Ä Î∂ÑÏÑù Í≤∞Í≥ºÎ°ú ÎÇòÏò® Í≥ÑÏàòÎ•º Í≤ÄÏ†ï ÌÜµÍ≥ÑÎüâÏúºÎ°ú ÏÇ¨Ïö©ÌïòÎ©∞ ÌöåÍ∑Ä Î∞©Ï†ïÏãùÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$\nabla Y_t = \alpha + \beta t +  (a_0 - 1) Y_{t-1} + a_1 \nabla Y_{t-1} + a_2 \nabla Y_{t-2} + \cdots + a_p \nabla Y_{t-p} + \epsilon_t$$

ADF Í≤ÄÏ†ïÏùÄ DF Í≤ÄÏ†ïÏùÑ Ìè¨Ìï®ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê ÎåÄÎ∂ÄÎ∂ÑÏùò ÌÜµÍ≥Ñ Ìå®ÌÇ§ÏßÄÎäî ADFÎßåÏùÑ ÏßÄÏõêÌïúÎã§.

### statsmodelsÏùÑ ÏÇ¨Ïö©Ìïú ADF Í≤ÄÏ†ï

statsmodels Ìå®ÌÇ§ÏßÄÎäî statsmodels.tsa.adfuller ÎùºÎäî ADF Í≤ÄÏ†ï Ìï®ÏàòÎ•º Ï†úÍ≥µÌïúÎã§.

ÏûÖÎ†• Ïù∏ÏàòÏôÄ Î∞òÌôòÍ∞íÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

+ ÏûÖÎ†• Ïù∏Ïàò

    + x : ÏãúÍ≥ÑÏó¥ ÏûêÎ£å
    + maxlag : ADF Í≤ÄÏ†ïÏóê ÏÇ¨Ïö©Ìï† ÏãúÏ∞®Ïùò Ïàò. ÎîîÌè¥Ìä∏ 12*(nobs/100)^{1/4}
    + regression : Í≤ÄÏ†ï ÏÇ¨Ïö©Ìï† Î™®Ìòï.
        + 'nc': Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏ ÏóÜÏùå. DF Í≤ÄÏ†ï
        + 'c': ÏÉÅÏàòÌï≠Îßå ÏÇ¨Ïö©
        + 'ct': ÏÉÅÏàòÌï≠Í≥º 1Ï∞® Ï∂îÏÑ∏ ÏÇ¨Ïö©
        + 'ctt': ÏÉÅÏàòÌï≠Í≥º 1Ï∞® Ï∂îÏÑ∏, 2Ï∞® Ï∂îÏÑ∏ ÏÇ¨Ïö©
    + autolag : Í≤ÄÏ†ï AR Î™®ÌòïÏùò Ï∞®ÏàòÎ•º ÏûêÎèô Í≤∞Ï†ïÌïòÎäî ÏïåÍ≥†Î¶¨Ï¶ò {‚ÄòAIC‚Äô, ‚ÄòBIC‚Äô, ‚Äòt-stat‚Äô, None}


+ Î∞òÌôòÍ∞í
    + adf : Í≤ÄÏ†ï ÌÜµÍ≥ÑÎüâ
    + pvalue : MacKinnon(1994) Î∞©ÏãùÏùÑ ÏÇ¨Ïö©Ìïú p-value Ï∂îÏ†ïÏπò
    + usedlag : ÏÇ¨Ïö©Îêú ÏãúÏ∞®Ïùò Ïàò
    + nobs : Î∂ÑÏÑùÏóê ÏûêÎ£åÏùò Ïàò
    + critical values : 1%, 5%, 10% ÏàòÏ§ÄÏóê Ìï¥ÎãπÎêòÎäî Í≤ÄÏ†ï ÌÜµÍ≥ÑÎüâÏùò Í∞í

Ïã§Ï†úÎ°ú ÏãúÎÆ¨Î†àÏù¥ÏÖòÏúºÎ°ú ÏÉùÏÑ±Îêú ÏãúÍ≥ÑÏó¥ ÏûêÎ°úÏóê ÎåÄÌï¥ ADF Í≤ÄÏ†ïÏùÑ Ïã§ÏãúÌï¥ Î≥∏Îã§.

```python
p = sm.tsa.ArmaProcess([1], [-1, 0.6])
y2 = p.generate_sample(100, burnin=100).cumsum().cumsum()  # ARIMA d=2
y1 = np.diff(y2)
y0 = np.diff(y1)
```

ÏúÑ ÏΩîÎìúÏóêÏÑú y2Îäî Ï†ÅÎ∂ÑÏ∞®ÏàòÍ∞Ä 2, y1Îäî Ï†ÅÎ∂ÑÏ∞®ÏàòÍ∞Ä 1, y0Îäî Ï†ÅÎ∂ÑÏ∞®ÏàòÍ∞Ä 0Ïù¥ÎØÄÎ°ú 

y2, y1Îäî  ùëé0=0 ÏúºÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§Ïù¥ Ï±ÑÌÉùÎêòÏñ¥Ïïº ÌïòÍ≥† y0Îäî  ùëé0‚â†0 ÏúºÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§Ïù¥ Í∏∞Í∞ÅÎêòÏñ¥Ïïº ÌïúÎã§. Îã§Ïùå Í≤ÄÏ†ïÏóêÏÑú Ïù¥ Í≤∞Í≥ºÎ•º ÌôïÏù∏Ìï† Ïàò ÏûàÎã§.

```python
sm.tsa.adfuller(y2)
```

```python
sm.tsa.adfuller(y1)
```

```python
sm.tsa.adfuller(y0)
```
