다음과 같은 ARMA(p,q) 모형과 표본 시계열 자료가 있는 경우 
#### 최대 가능도 추정법으로 모수  𝜙𝑖 ,  𝜃𝑖 를 구할 수 있다.

$$Y_t + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} 
= \epsilon_t + \theta_1 \epsilon_{t-1}  + \theta_2 \epsilon_{t-2} - \cdots  + \theta_q \epsilon_{t-q}$$

MLE를 사용하기 위한 가능도 함수는 다음과 같다.

\begin{eqnarray}
\mathcal{L}(\theta; \{y_j\}) 
&=& \mathcal{L}(\phi_1, \cdots, \phi_p, \theta_1, \cdots, \theta_q; y_1, y_2, \cdots, y_N) \\
&=& f_{Y_1,\cdots , Y_N}(y_1, y_2, \cdots, y_N;\phi_1, \cdots, \phi_p, \theta_1, \cdots, \theta_q) \\
\end{eqnarray}

## ARMA(p,q)의 가능도 함수

확률변수와 달리 확률과정 모형에서는  𝜖  값을 알아야 하기 때문에 다음과 같은 반복적 방법을 사용한다.

#### (1) 𝜖  값의 초기값 가정과 선형회귀방법을 통한 모수 추정

우선 임의의  𝜖  값을 초기값으로 가정한다.

𝑌𝑡 는  𝑌𝑡−1,⋯,𝑌𝑡−𝑝,𝜖𝑡−1,⋯,𝜖𝑡−𝑞  에 의존하는 정규 분포 확률변수가 된다.

\begin{eqnarray}
Y_{t} 
&\sim& 
-\phi_1 Y_{t-1} -\phi_2 Y_{t-2} -\cdots -\phi_p Y_{t-p} 
+\theta_1 \epsilon_{t-1} +\theta_2 \epsilon_{t-2} +\cdots +\theta_q \epsilon_{t-q}
\end{eqnarray}

선형회귀방법으로 모수  𝜙1,⋯,𝜙𝑝,𝜃1,⋯,𝜃𝑞 를 추정한다.

#### (2) 다시 𝜖  값을 계산하고 회귀분석을 실시하여 모수 추정 


𝜖  값을 계산한다. 우선 다음과 같은 초기값 가정을 하자.

$$\epsilon_p = \epsilon_{p-1} = \cdots = \epsilon_{p-q} = 0$$

이  𝜖𝑝+1,⋯,𝜖𝑁 은 다음 공식에서 구할 수 있다.

$$\epsilon_{p+1}=Y_{p+1}+\phi_1Y_{p}+\phi_2 Y_{p-1} + \cdots + \phi_p Y_{1}$$

$$\epsilon_{p+2}=Y_{p+2}+\phi_1Y_{p+1}+\phi_2Y_{p}+\cdots+\phi_pY_{2}-\theta_1\epsilon_{p+1}$$

$$\vdots$$

이  𝜖  값을 이용하여 다시 (1)의 회귀분석을 실시하여 모수  𝜙1,⋯,𝜙𝑝,𝜃1,⋯,𝜃𝑞 를 추정한다.

## statsmodels 패키지를 사용한 ARMA 모수 추정

statsmodels 패키지는 ARMA 모수 추정을 위한 ARMA라는 클래스와 이 클래스의 fit 메서드를 제공한다. ARMA 클래스의 인수로 시계열 표본과 차수를 넣어 인스턴스를 생성한 뒤 fit 메서드를 호출하면 추정 결과값을 가진 ARMAResults 클래스 자료를 반환한다. 

#### 다만 주의할 점은 ARMAResults에서는 AR계수  𝜙 에 대해 ArmaProcess 클래스에서 정한 부호와 반대의 부호를 출력한다.

ARMA 클래스와 몇가지 이론적 모형에서 시뮬레이션하여 나온 시계열 자료를 사용하여 모수를 측정하여 보자.

## MA(1) 모수 추정의 예

$$Y_t = \epsilon_t + 0.9 \epsilon_{t-1}$$

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)

import statsmodels.api as sm
```

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1, -0.9], [1]) # MA 모형 생성
y = p.generate_sample(1000)  # 샘플 생성
m = sm.tsa.ARMA(y, (1, 0))   # 추정값 생성을 위해 시계열 표본과 차수를 넣은 후 인스턴스 생성
r = m.fit()  # 추정결과값
print(r.summary())
```

ar.L1.y 값이 0.9144로 모형의 0.9와 거의 일치한 값이 추정되고 있다.

## AR(2) 모수 추정의 예

$$Y_t = 1.5 Y_{t-1} - 0.75 Y_{t-2} + \epsilon_t$$

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1, -1.5, 0.75], [1])
y = p.generate_sample(1000)
m = sm.tsa.ARMA(y, (2, 0))
r = m.fit()
print(r.summary())
```

ar.L1.y 1.49, ar.L2.y -0.74로 원래 모형의 모수값과 거의 일치한다. 상수항은 14%로 귀무가설 채택으로 0라고 할 수 있다.

## ARMA(1,1) 모수 추정의 예

$$Y_t = 0.6 Y_{t-1} + \epsilon_t + 0.3 \epsilon_{t-1}$$

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1, -0.6], [1, 0.3])
y = p.generate_sample(1000)
m = sm.tsa.ARMA(y, (1, 1))
r = m.fit()
print(r.summary())
```

ar.L1.y 0.62, ma.L1.y 0.26으로 원래 모형의 모수 0.6, 0.3과 유사하게 추정되고 있다.

## 실제 시계열 자료에 대한 ARMA 모형 계수 추정

이번에는 실제 시계열 자료에 대해 ARMA 모형 계수를 추정해 보자.

### 사례1. 황체형성 호르몬 모형 계수 추정

황체형성 호르몬의 모형 차수는 AR(1)로 추정했었다. 이 때 계수는 다음처럼 구한다.

```python
data = sm.datasets.get_rdataset("lh")
df = data.data
df.plot(x="time", y="value", style=["o-"])
plt.show()
```

이 시계열을 AR(1) 모형으로 보고 모수를 추정한 결과는 다음과 같다.

```python
m = sm.tsa.ARMA(df.value, (1, 0))
r = m.fit()
print(r.summary())
```

이 모수를 사용하여 다시 시뮬레이션한 결과는 다음과 같다. 굵은 실선이 원래 자료이다.

```python
plt.figure(figsize=(10, 10))
#plt.plot(r.fittedvalues, label="추정한 데이터")
np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], [1])
plt.subplot(6, 1, 1)
plt.plot(df.value, lw=5)  # 원래 데이터

for i in range(5):
    plt.subplot(6, 1, i+2)
    y = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0] # 각 샘플 값에 편차를 곱한 후 상수항 더함
    plt.plot(y)
    
plt.tight_layout()
plt.show()
```

MA(1) 모형으로 추정한 결과는 다음과 같다. AR(1) 모형보다 성능이 좋지 않음을 알 수 있다.  (Loglikelihood : 로그 가능도, AIC, BIC 비교)

```python
m = sm.tsa.ARMA(df.value, (0, 1))
r = m.fit()
print(r.summary())
```

ARMA(1,1)모형으로 추정하면 가능도는 높지만 AIC, BIC가 나쁘다.

```python
m = sm.tsa.ARMA(df.value, (1, 1))
r = m.fit()
print(r.summary())
```

### 사례2. 호흡기질환 사망자수

```python
data = sm.datasets.get_rdataset("deaths", "MASS")
df = data.data
df.tail()
```

```python
df.plot(x="time", y="value")
plt.show()
```

```python
# 추세 제거를 위해 로그변환
df["logvalue"] = np.log(df.value)
df.plot(x="time", y="logvalue")
plt.show()
```

이 시계열을 AR(2) 모형으로 보고 모수를 추정한 결과는 다음과 같다.

```python
m = sm.tsa.ARMA(df.logvalue, (2, 0))
r = m.fit()
print(r.summary())
```

이 모수를 사용하여 다시 시뮬레이션한 결과는 다음과 같다. 굵은 실선이 원래 자료이다.

```python
plt.figure(figsize=(10, 10))
np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], [1])
plt.subplot(6, 1, 1)
plt.plot(df.logvalue, lw=5)

for i in range(5):
    plt.subplot(6, 1, i+2)
    y = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0]
    plt.plot(y)
    
plt.tight_layout()
plt.show()
```

ARMA(1,0), ARMA(1,1), ARMA(2,0), ARMA(2,1), ARMA(1,2), ARMA(2,2) 모형을 사용한 결과는 다음과 같다.

```python
from itertools import product

result = []
for p, q in product(range(3), range(3)):
    if (p == 0): 
        continue 
    m = sm.tsa.ARMA(df.logvalue, (p, q))
    try:
        r = m.fit()
        result.append({"p": p, "q": q, "LLF": r.llf, "AIC": r.aic, "BIC": r.bic})
    except:
        pass
```

```python
import pandas as pd

pd.DataFrame(result)
```

위의 결과를 통해 보면, log likelihood 기준으로 볼때 ARMA(2, 1) 모형이 가장 성능이 좋다고 할 수 있다. 따라서 다시 이 모형을 기준으로 모수를 추정하면 다음과 같다.

```python
m = sm.tsa.ARMA(df.logvalue.astype(float), (2, 1))
r = m.fit()
print(r.summary())
```

위의 모수추정 결과로 다시 시뮬레이션하면 다음과 같다.

```python
plt.figure(figsize=(10, 10))

np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], [1, r.maparams])
plt.subplot(6, 1, 1)
plt.plot(df.logvalue, lw=5)
for i in range(5):
    plt.subplot(6, 1, i+2)
    y = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0]
    plt.plot(y)
    
plt.tight_layout()
plt.show()
```

### 사례3. 호흡기 질환 사망자 수의 결정론적 Seasonality 모형

```python
def yearfraction2datetime(yearfraction, startyear=0):
    import datetime
    import dateutil
    year = int(yearfraction) + startyear
    month = int(round(12 * (yearfraction - year)))
    delta = dateutil.relativedelta.relativedelta(months=month)
    date = datetime.datetime(year, 1, 1) + delta
    return date

df["datetime"] = df.time.map(yearfraction2datetime)
df["month"] = df.datetime.dt.month

df.tail()
```

선형회귀분석을 통해 계절성을 추정한다.

```python
result = sm.OLS.from_formula('logvalue ~ C(month) + time + 0', data=df).fit()
y_seasonal = result.fittedvalues # 계절성 추정한 값
y_nonseasonal = df.logvalue - y_seasonal # 계절성 제거
```

```python
from itertools import product
```

```python
result = []
for p, q in product(range(3), range(3)):
    if (p==0):
        continue
    m = sm.tsa.ARMA(y_nonseasonal, (p, q))
    try:
        r = m.fit()
        result.append({"p": p, "q": q, "LLF": r.llf, "AIC": r.aic, "BIC": r.bic})
    except:
        pass
    
pd.DataFrame(result)
```

#### 계절성 추세를 제거한 후에 다시 모수 추정 결과 ARMA(2, 2)모형이 가장 성능이 좋다. ARMA(2, 2) 모형으로 시뮬레이션하면 다음과 같다.

```python
m = sm.tsa.ARMA(y_nonseasonal, (2, 2))
r = m.fit()
```

```python
plt.figure(figsize=(10, 10))
np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], np.r_[1, r.maparams],)
plt.subplot(6, 1, 1)
plt.plot(df.logvalue, lw=5)
for i in range(5):
    plt.subplot(6, 1, i + 2)
    y2 = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0] 
    y = y_seasonal + y2
    plt.plot(y)
plt.tight_layout()
plt.show()
```
