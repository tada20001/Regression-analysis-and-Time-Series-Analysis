## Î†àÎ≤ÑÎ¶¨ÏßÄ

###### Í∞úÎ≥ÑÏ†ÅÏù∏ Îç∞Ïù¥ÌÑ∞ ÌëúÎ≥∏ ÌïòÎÇòÌïòÎÇòÍ∞Ä ÌöåÍ∑ÄÎ∂ÑÏÑù Í≤∞Í≥ºÏóê ÎØ∏ÏπòÎäî ÏòÅÌñ•Î†•ÏùÄ Î†àÎ≤ÑÎ¶¨ÏßÄ Î∂ÑÏÑùÏù¥ÎÇò ÏïÑÏõÉÎùºÏù¥Ïñ¥ Î∂ÑÏÑùÏùÑ ÌÜµÌï¥ Ïïå Ïàò ÏûàÎã§.

##### Î†àÎ≤ÑÎ¶¨ÏßÄ(leverage)Îäî Ïã§Ï†ú Ï¢ÖÏÜçÎ≥ÄÏàòÍ∞í  ùë¶ Í∞Ä ÏòàÏ∏°Ïπò(predicted target)  ùë¶ÃÇ  Ïóê ÎØ∏ÏπòÎäî ÏòÅÌñ•ÏùÑ ÎÇòÌÉÄÎÇ∏ Í∞íÏù¥Îã§. self-influence, self-sensitivity ÎùºÍ≥†ÎèÑ ÌïúÎã§.

Í∞ÄÏ§ëÏπò Î≤°ÌÑ∞Ïùò Í≤∞Í≥ºÍ∞íÏùÑ ÏòàÏ∏°ÏãùÏóê ÎåÄÏûÖÌïòÏó¨  ùë¶  ÏôÄ  ùë¶ÃÇ  Ïùò Í¥ÄÍ≥ÑÎ•º Îã§ÏùåÍ≥º Í∞ôÎã§Îäî Í≤ÉÏùÑ Î≥¥ÏòÄÏóàÎã§.

$$\hat{y} =  Hy$$

Ïù¥ ÌñâÎ†¨  ùêª Î•º ÏòÅÌñ•ÎèÑ ÌñâÎ†¨(influence matrix) ÎòêÎäî hat ÌñâÎ†¨(hat matrix)Ïù¥ÎùºÍ≥† ÌïúÎã§Í≥† ÌïòÏòÄÎã§.

ÌñâÎ†¨  ùêª Ïùò  ùëñ Î≤àÏß∏ Ìñâ,  ùëó Î≤àÏß∏ Ïó¥ ÏÑ±Î∂ÑÏùÑ  ‚Ñéùëñùëó ÎùºÍ≥† ÌïòÎ©¥ Ïã§Ï†ú Í≤∞Í≥ºÍ∞í  ùë¶ùëñ Í≥º ÏòàÏ∏°Í∞í  ùë¶ÃÇ ùëñ ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ Í¥ÄÍ≥ÑÎ•º Í∞ÄÏßÑÎã§.

$$\hat{y}_i = h_{i1} y_1 + h_{i2} y_2 + \cdots + h_{ii} y_i + \cdots + h_{iN} y_N$$

Î†àÎ≤ÑÎ¶¨ÏßÄÎäî ÏàòÌïôÏ†ÅÏúºÎ°ú ÏòÅÌñ•ÎèÑ ÌñâÎ†¨Ïùò ÎåÄÍ∞ÅÏÑ±Î∂Ñ  ‚Ñéùëñùëñ ÏúºÎ°ú Ï†ïÏùòÎêúÎã§. 

##### Ï¶â, Î†àÎ≤ÑÎ¶¨ÏßÄÎäî Ïã§Ï†úÏùò Í≤∞Í≥ºÍ∞í  ùë¶ùëñ Ïù¥ ÏòàÏ∏°Í∞í  ùë¶ÃÇ ùëñ Ïóê ÎØ∏ÏπòÎäî ÏòÅÌñ•, Ï¶â ÏòàÏ∏°Ï†êÏùÑ ÏûêÍ∏∞ ÏûêÏã†Ïùò ÏúÑÏπòÎ°ú ÎÅåÏñ¥ ÎãπÍ∏∞Îäî Ï†ïÎèÑÎ•º ÎÇòÌÉÄÎÇ∏ Í≤ÉÏù¥Îã§.

ÎßåÏïΩ  ‚Ñéùëñùëñ  Í∞íÏù¥ 1Ïù¥ ÎêòÍ≥† ÎÇòÎ®∏ÏßÄ ÏÑ±Î∂ÑÎì§Ïù¥ Î™®Îëê 0Ïù¥ Îê† ÏàòÎßå ÏûàÎã§Î©¥ Î™®Îì† ÌëúÎ≥∏ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ Ïã§Ï†ú Í≤∞Í≥ºÍ∞íÍ≥º ÏòàÏ∏°Í∞íÏù¥ ÏùºÏπòÌïòÍ≤å Îê† Í≤ÉÏù¥Îã§.

$$h_{ii} = 1, h_{ij} = 0 \; (\text{for } i \neq j) \;\; \rightarrow \;\; \hat{y}_i = y_i$$

ÌïòÏßÄÎßå Í≥ß Ïïå Ïàò ÏûàÎìØÏù¥ Ïù¥Îü¨Ìïú ÏùºÏùÄ Î∞úÏÉùÌïòÏßÄ ÏïäÎäîÎã§. 


Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞íÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ ÌäπÏÑ±ÏùÑ Í∞ÄÏßÑÎã§Îäî Í≤ÉÏùÑ ÏàòÌïôÏ†ÅÏúºÎ°ú Ï¶ùÎ™ÖÌï† Ïàò ÏûàÎã§.

1. 1Î≥¥Îã§ Í∞ôÍ±∞ÎÇò ÏûëÏùÄ ÏñëÏàò ÌòπÏùÄ 0Ïù¥Îã§.

$$0 \leq h_{ii} \leq 1$$

2. Î†àÎ≤ÑÎ¶¨ÏßÄÏùò Ìï©ÏùÄ Î™®ÌòïÏóê ÏÇ¨Ïö©Îêú Î™®ÏàòÏùò Í∞ØÏàò  ùêæ ÏôÄ Í∞ôÎã§. Î™®ÏàòÏóêÎäî ÏÉÅÏàòÌï≠ÎèÑ Ìè¨Ìï®ÎêòÎØÄÎ°ú ÏÉÅÏàòÌï≠Ïù¥ ÏûàÎäî 1Ï∞®Ïõê Î™®ÌòïÏóêÏÑúÎäî  ùêæ=2 Í∞Ä ÎêúÎã§.

$$\text{tr}(H) = \sum_i^N h_{ii} = K$$

Ïù¥ ÎëêÍ∞ÄÏßÄ ÏÑ±ÏßàÎ°úÎ∂ÄÌÑ∞ Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞íÏùÄ  ùëÅ Í∞úÏùò Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞íÏùÄ ÏñëÏàòÏù¥Í≥† Í∑∏ Ìï©Ïù¥  ùêæ Í∞Ä ÎêúÎã§Îäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§. 

Ï¶â  ùêæ ÎùºÍ≥† ÌïòÎäî Í∞íÏùÑ  ùëÅ Í∞úÏùò Î≥ÄÏàòÍ∞Ä ÎÇòÎàÑÏñ¥ Í∞ÄÏßÄÎäî Í≤ÉÍ≥º Í∞ôÎã§. 

##### ÌòÑÏã§Ï†ÅÏúºÎ°ú Îç∞Ïù¥ÌÑ∞Ïùò Í∞ØÏàò  ùëÅ Îäî Î™®ÏàòÏùò Í∞ØÏàò  ùêæ Î≥¥Îã§ Ìõ®Ïî¨ ÎßéÍ∏∞ ÎïåÎ¨∏Ïóê ÏúÑÏóêÏÑú ÎßêÌïú ÌäπÏÑ±ÏùÑ Ï†ÅÏö©ÌïòÎ©¥ Î™®Îì† Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞íÏù¥ ÎèôÏãúÏóê 1Ïù¥ ÎêòÎäî Í≤ÉÏùÄ Î∂àÍ∞ÄÎä•ÌïòÎã§.

ÏúÑ ÏãùÏùÑ Ïù¥Ïö©ÌïòÎ©¥ Î†àÎ≤ÑÎ¶¨ÏßÄÏùò ÌèâÍ∑†Í∞íÎèÑ Íµ¨Ìï† Ïàò ÏûàÎã§.

$$h_{ii} \approx \dfrac{K}{N}$$

##### Î≥¥ÌÜµ Ïù¥ ÌèâÍ∑†Í∞íÏùò 2 ~ 4Î∞∞Î≥¥Îã§ Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞íÏù¥ ÌÅ¨Î©¥ Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ¨Îã§Í≥† Ïù¥ÏïºÍ∏∞ÌïúÎã§.

## statsmodelsÎ•º Ïù¥Ïö©Ìïú Î†àÎ≤ÑÎ¶¨ÏßÄ Í≥ÑÏÇ∞

Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞íÏùÄ RegressionResults ÌÅ¥ÎûòÏä§Ïùò get_influence Î©îÏÑúÎìúÎ°ú Îã§ÏùåÍ≥º Í∞ôÏù¥ Íµ¨Ìï† Ïàò ÏûàÎã§. Ïö∞ÏÑ† Îã§ÏùåÍ≥º Í∞ôÏùÄ Í∞ÄÏÉÅÏùò 1Ï∞® Îç∞Ïù¥ÌÑ∞Î°ú ÌöåÍ∑ÄÎ∂ÑÏÑù ÏòàÏ†úÎ•º ÌíÄÏñ¥Î≥¥Ïûê.

```python
import numpy as np
import statsmodels.api as sm
from sklearn.datasets import make_regression

X0, y, coef = make_regression(n_samples=100, n_features=1, noise=20, coef=True, random_state=1)

# Î†àÎ≤ÑÎ¶¨Í∞Ä ÎÜíÏùÄ Í∞ÄÏÉÅÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï∂îÍ∞Ä
data_100 = (4, 300)
data_101 = (3, 150)
X0 = np.vstack([X0, np.array([data_100[:1], data_101[:1]])])
X = sm.add_constant(X0)
y = np.hstack([y, [data_100[1], data_101[1]]])
```

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
plt.scatter(X0, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Í∞ÄÏÉÅÏùò ÌöåÍ∑ÄÎ∂ÑÏÑùÏö© Îç∞Ïù¥ÌÑ∞")
plt.show()
```

Ïù¥ Îç∞Ïù¥ÌÑ∞Î•º Ïù¥Ïö©ÌïòÏó¨ ÏÑ†ÌòïÌöåÍ∑ÄÎ•º Ìïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
import pandas as pd

model = sm.OLS(pd.DataFrame(y), pd.DataFrame(X))
result = model.fit()
print(result.summary())
```

ÏÑ†ÌòïÌöåÍ∑Ä Í≤∞Í≥ºÏóêÏÑú get_influence Î©îÏÑúÎìúÎ•º Ìò∏Ï∂úÌïòÎ©¥ ÏòÅÌñ•ÎèÑ Ï†ïÎ≥¥ Í∞ùÏ≤¥Î•º Íµ¨Ìï† Ïàò ÏûàÎã§. Ïù¥ Í∞ùÏ≤¥Îäî hat_matrix_diag ÏÜçÏÑ±ÏúºÎ°ú Î†àÎ≤ÑÎ¶¨ÏßÄ Î≤°ÌÑ∞Ïùò Í∞íÏùÑ Í∞ÄÏßÄÍ≥† ÏûàÎã§.

```python
influence = result.get_influence()
hat = influence.hat_matrix_diag  # Î†àÎ≤ÑÎ¶¨ÏßÄ Î≤°ÌÑ∞ Í∞í
```

```python
hat.shape
```

```python
plt.figure(figsize=(10, 2))
plt.stem(hat)
plt.axhline(0.02, c='g', ls="--")  #Î†àÎ≤ÑÎ¶¨ÏßÄ ÌèâÍ∑† ÌëúÏãú K / N = D+1/N = 2 / 102
plt.title("Í∞Å Îç∞Ïù¥ÌÑ∞Ïùò Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞í")
plt.show()
```

ÏàòÎèôÏúºÎ°ú Ï∂îÍ∞ÄÌïú ÎßàÏßÄÎßâ Îëê Í∞úÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï†úÏô∏ÌïòÎ©¥ ÎåÄÎ∂ÄÎ∂ÑÏùò Îç∞Ïù¥ÌÑ∞Îäî Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞íÏù¥ 0.02 Í∑ºÏ≤òÏùò ÎÇÆÏùÄ Í∞íÏùÑ Í∞ÄÏßÑÎã§.

$$\text{Î†àÎ≤ÑÎ¶¨ÏßÄ ÌèâÍ∑†} \approx \dfrac{2}{102} \approx 0.02$$

Î†àÎ≤ÑÎ¶¨ÏßÄÏùò Ìï©Ïù¥ 2ÏôÄ Í∞ôÏïÑÏßÑÎã§Îäî Í≤ÉÎèÑ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÌôïÏù∏Ìï† Ïàò ÏûàÎã§.

```python
hat.sum()
```

```python
hat[100], hat[101]  # ÎëòÎã§ ÌèâÍ∑† Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞íÎ≥¥Îã§ 4Î∞∞Ïù¥ÏÉÅ ÌÅ¨Í∏∞ ÎïåÎ¨∏Ïóê Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ¨Îã§Í≥† Î≥º Ïàò ÏûàÎã§.
```

ÏúÑ Í≤∞Í≥ºÎ•º Ïù¥Ïö©ÌïòÏó¨ Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞Îßå ÌëúÏãúÌïòÎ©¥ ÏïÑÎûòÏôÄ Í∞ôÎã§. Ïó¨Í∏∞ÏóêÏÑúÎäî Î†àÎ≤ÑÎ¶¨ÏßÄ Í∞íÏù¥ 0.05Î•º ÎÑòÎäî Îç∞Ïù¥ÌÑ∞Îßå ÌÅ∞ ÏõêÏúºÎ°ú ÌëúÏãúÌïòÏòÄÎã§.

```python
ax = plt.subplot()
plt.scatter(X0, y)
sm.graphics.abline_plot(model_results=result, ax=ax)

idx = hat > 0.05
plt.scatter(X0[idx], y[idx], s=300, c='r', alpha=0.3)
plt.title("ÌöåÍ∑ÄÎ∂ÑÏÑù Í≤∞Í≥ºÏôÄ Î†àÎ≤ÑÎ¶¨ÏßÄ Ìè¨Ïù∏Ìä∏")
plt.show()
```

ÏúÑ Í∑∏Î¶ºÏóêÏÑú Ïïå Ïàò ÏûàÎìØÏù¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Î¨¥Î¶¨ÏßÄÏñ¥ ÏûàÏßÄ ÏïäÍ≥† Îã®ÎèÖÏúºÎ°ú Ï°¥Ïû¨Ìï†ÏàòÎ°ù Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä Ïª§ÏßêÏùÑ Ïïå Ïàò ÏûàÎã§.

## Î†àÎ≤ÑÎ¶¨ÏßÄÏùò ÏòÅÌñ•

Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞Í∞Ä Î™®ÌòïÏóê Ï£ºÎäî ÏòÅÌñ•ÏùÑ Î≥¥Í∏∞ ÏúÑÌï¥ Ïù¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Ìè¨Ìï®Îêú Í≤ΩÏö∞Ïùò Î™®ÌòïÍ≥º Ìè¨Ìï®ÎêòÏßÄ ÏïäÏùÄ Í≤ΩÏö∞Ïùò Î™®ÌòïÏùÑ ÏïÑÎûòÏóê ÎπÑÍµêÌïòÏòÄÎã§. Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞Îäî Ìè¨Ìï®ÎêòÍ±∞ÎÇò Ìè¨Ìï®ÎêòÏßÄ ÏïäÎäîÍ∞ÄÏóê Îî∞Îùº Î™®ÌòïÏóê Ï£ºÎäî ÏòÅÌñ•Ïù¥ ÌÅ∞ Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§.

```python
model2 = sm.OLS(y[:-1], X[:-1])
result2 = model2.fit()

ax = plt.subplot()
plt.scatter(X0, y)
sm.graphics.abline_plot(model_results=result, c='r', linestyle="--", alpha=0.7, ax=ax)
sm.graphics.abline_plot(model_results=result2, c='g', alpha=0.7, ax=ax)

plt.plot(X0[-1], y[-1], marker="x", c="m", ms=20, mew=5)
plt.legend(["Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞Î•º Ìè¨Ìï®Ìïú Í≤ΩÏö∞", "Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞Î•º Ìè¨Ìï®ÌïòÏßÄ ÏïäÏùÄ Í≤ΩÏö∞"], loc="upper left")
plt.title("Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÎÜíÏùÄ Îç∞Ïù¥ÌÑ∞Í∞Ä ÌöåÍ∑ÄÎ∂ÑÏÑùÏóê ÎØ∏ÏπòÎäî ÏòÅÌñ•")
plt.show()
```

Î∞òÎåÄÎ°ú Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Îäî Ìè¨Ìï®ÎêòÍ±∞ÎÇò Ìè¨Ìï®ÎêòÏßÄ ÏïäÍ±∞ÎÇò Î™®ÌòïÏù¥ Î≥ÑÎ°ú Îã¨ÎùºÏßÄÏßÄ ÏïäÎäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§.

```python
model3 = sm.OLS(y[1:], X[1:])
result3 = model3.fit()

ax = plt.subplot()
plt.scatter(X0, y)
sm.graphics.abline_plot(model_results=result, c='r', linestyle="--", ax=ax)
sm.graphics.abline_plot(model_results=result3, c='g', ax=ax)
plt.plot(X0[0], y[0], marker="x", c="m", ms=20, mew=5)
plt.legend(["Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Î•º Ìè¨Ìï®Ìïú Í≤ΩÏö∞", "Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Î•º Ìè¨Ìï®ÌïòÏßÄ ÏïäÏùÄ Í≤ΩÏö∞"], loc="upper left")
plt.title("Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Í∞Ä ÌöåÍ∑ÄÎ∂ÑÏÑùÏóê ÎØ∏ÏπòÎäî ÏòÅÌñ•")
plt.show()
```

ÌòπÏùÄ Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ¨ÎçîÎùºÎèÑ Ïò§Ï∞®Í∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Îäî Ìè¨Ìï®ÎêòÍ±∞ÎÇò Ìè¨Ìï®ÎêòÏßÄ ÏïäÍ±∞ÎÇò Î™®ÌòïÏù¥ Î≥ÑÎ°ú Îã¨ÎùºÏßÄÏßÄ ÏïäÎäîÎã§.

```python
idx = np.array(list(range(100)) + [101])
model4 = sm.OLS(y[idx], X[idx, :])
result4 = model4.fit()

ax = plt.subplot()
plt.scatter(X0, y)
sm.graphics.abline_plot(model_results=result, c='r', linestyle="--", alpha=0.7, ax=ax)
sm.graphics.abline_plot(model_results=result4, c='g', alpha=0.7, ax=ax)

plt.plot(X0[-2], y[-2], marker='x', ms=20, mew=5, c='m')
plt.legend(["Ïò§Ï∞®Í∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Î•º Ìè¨Ìï®Ìïú Í≤ΩÏö∞", "Ïò§Ï∞®Í∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Î•º Ìè¨Ìï®ÌïòÏßÄ ÏïäÏùÄ Í≤ΩÏö∞"], loc="upper left")
plt.title("Î†àÎ≤ÑÎ¶¨ÏßÄÎäî ÎÜíÏßÄÎßå Ïò§Ï∞®Í∞Ä ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Í∞Ä ÌöåÍ∑ÄÎ∂ÑÏÑùÏóê ÎØ∏ÏπòÎäî ÏòÅÌñ•")
plt.show()
```

## ÏïÑÏõÉÎùºÏù¥Ïñ¥

##### Î™®ÌòïÏóêÏÑú ÏÑ§Î™ÖÌïòÍ≥† ÏûàÎäî Îç∞Ïù¥ÌÑ∞ÏôÄ ÎèôÎñ®Ïñ¥ÏßÑ Í∞íÏùÑ Í∞ÄÏßÄÎäî Îç∞Ïù¥ÌÑ∞, Ï¶â ÏûîÏ∞®Í∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞Î•º ÏïÑÏõÉÎùºÏù¥Ïñ¥(outlier)ÎùºÍ≥† ÌïúÎã§. 

Í∑∏Îü∞Îç∞ ÏûîÏ∞®Ïùò ÌÅ¨Í∏∞Îäî ÎèÖÎ¶Ω Î≥ÄÏàòÏùò ÏòÅÌñ•ÏùÑ Î∞õÏúºÎØÄÎ°ú ÏïÑÏõÉÎùºÏù¥Ïñ¥Î•º Ï∞æÏúºÎü¨Î©¥ Ïù¥ ÏòÅÌñ•ÏùÑ Ï†úÍ±∞Ìïú ÌëúÏ§ÄÌôîÎêú ÏûîÏ∞®Î•º Í≥ÑÏÇ∞Ìï¥Ïïº ÌïúÎã§.

### ÌëúÏ§ÄÌôî ÏûîÏ∞®
Í∞úÎ≥ÑÏ†ÅÏù∏ ÏûîÏ∞®Ïùò ÌëúÏ§ÄÌé∏Ï∞®Î•º Íµ¨ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

\begin{eqnarray}
\text{Cov}[e] 
&=& \text{E}[M \epsilon\epsilon^T M^T] \\
&=& M \text{E}[\epsilon\epsilon^T] M^T \\
&=& M \sigma^2 I M^T \\
&=& \sigma^2MM^T \\
&=& \sigma^2MM \\
&=& \sigma^2M \\
&=& \sigma^2(I - H) \\
\end{eqnarray}

ÏúÑ ÏãùÏóêÏÑúÎäî  ùëÄ Ïù¥ ÎåÄÏπ≠ ÌñâÎ†¨Ïù¥Í≥†  ùëÄ2=ùëÄ ÎùºÎäî ÏÇ¨Ïã§ÏùÑ Ïù¥Ïö©ÌïòÏóàÎã§.

ÎåÄÍ∞Å ÏÑ±Î∂ÑÎßå Î≥¥Î©¥ Í∞úÎ≥ÑÏ†ÅÏù∏ ÏûîÏ∞®Ïùò ÌëúÏ§ÄÌé∏Ï∞®Îäî Îã§ÏùåÍ≥º Í∞ôÎã§.

$$\text{Var}[e_i] = \sigma^2 (1 - h_{ii})$$

Ï¶â, Ïò§Ï∞®Ïùò ÌëúÏ§Ä Ìé∏Ï∞®Îäî Î™®Îì† ÌëúÎ≥∏Ïóê ÎåÄÌï¥ Í∞ôÏßÄÎßå Í∞úÎ≥ÑÏ†ÅÏù∏ ÏûîÏ∞®Ïùò ÌëúÏ§ÄÌé∏Ï∞®Îäî Î†àÎ≤ÑÎ¶¨ÏßÄÏóê Îî∞Îùº Îã¨ÎùºÏßÄÎäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§.

Ïò§Ï∞®Ïùò Î∂ÑÏÇ∞ÏùÄ ÏïåÏßÄ Î™ªÌïòÎØÄÎ°ú ÏûîÏ∞® Î∂ÑÏÇ∞ÏúºÎ°úÎ∂ÄÌÑ∞ Ï∂îÏ†ïÌïúÎã§.

$$\text{Var}[e_i] \approx s^2 (1 - h_{ii})$$

Ïù¥ ÏãùÏóêÏÑú  ùë† Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Íµ¨Ìïú Ïò§Ï∞®Ïùò ÌëúÏ§ÄÌé∏Ï∞® Ï∂îÏ†ïÍ∞íÏù¥Îã§.

$$s^2 = \dfrac{e^Te}{N-K} = \dfrac{RSS}{N-K}$$

##### ÏûîÏ∞®Î•º Î†àÎ≤ÑÎ¶¨ÏßÄÏôÄ ÏûîÏ∞®Ïùò ÌëúÏ§Ä Ìé∏Ï∞®Î°ú ÎÇòÎàÑÏñ¥ ÎèôÏùºÌïú ÌëúÏ§Ä Ìé∏Ï∞®Î•º Í∞ÄÏßÄÎèÑÎ°ù Ïä§ÏºÄÏùºÎßÅÌïú Í≤ÉÏùÑ ÌëúÏ§ÄÌôî ÏûîÏ∞®(standardized residual ÎòêÎäî normalized residual ÎòêÎäî studentized residual)ÎùºÍ≥† ÌïúÎã§.

$$r_i = \dfrac{e_i}{s\sqrt{1-h_{ii}}}$$

## statsmodelsÎ•º Ïù¥Ïö©Ìïú ÌëúÏ§ÄÌôî ÏûîÏ∞® Í≥ÑÏÇ∞

## Cook's Distance

ÌöåÍ∑Ä Î∂ÑÏÑùÏóêÎäî ÏûîÏ∞®Ïùò ÌÅ¨Í∏∞Í∞Ä ÌÅ∞ Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÏõÉÎùºÏù¥Ïñ¥Í∞Ä ÎêòÎäîÎç∞ Ïù¥ Ï§ëÏóêÏÑúÎèÑ Ï£ºÎ°ú Í¥ÄÏã¨ÏùÑ Í∞ÄÏßÄÎäî Í≤ÉÏùÄ Î†àÎ≤ÑÎ¶¨ÏßÄÏôÄ ÏûîÏ∞®Ïùò ÌÅ¨Í∏∞Í∞Ä Î™®Îëê ÌÅ∞ Îç∞Ïù¥ÌÑ∞Îì§Ïù¥Îã§. ÏûîÏ∞®ÏôÄ Î†àÎ≤ÑÎ¶¨ÏßÄÎ•º ÎèôÏãúÏóê Î≥¥Í∏∞ ÏúÑÌïú Í∏∞Ï§ÄÏúºÎ°úÎäî Cook's DistanceÍ∞Ä ÏûàÎã§. Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï†ïÏùòÎêòÎäî Í∞íÏúºÎ°ú Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä Ïª§ÏßÄÍ±∞ÎÇò ÏûîÏ∞®Ïùò ÌÅ¨Í∏∞Í∞Ä Ïª§ÏßÄÎ©¥ Cook's Distance Í∞íÏù¥ Ïª§ÏßÑÎã§.

$$D_i = \frac{r_i^2}{\text{RSS}}\left[\frac{h_{ii}}{(1-h_{ii})^2}\right]$$

Fox' Outlier Recommendation ÏùÄ Cook's DistanceÍ∞Ä Îã§ÏùåÍ≥º Í∞ôÏùÄ Í∏∞Ï§ÄÍ∞íÎ≥¥Îã§ ÌÅ¥ Îïå ÏïÑÏõÉÎùºÏù¥Ïñ¥Î°ú ÌåêÎã®ÌïòÏûêÎäî Í≤ÉÏù¥Îã§.

$$D_i > \dfrac{4}{N ‚àí K - 1}$$

Î™®Îì† Îç∞Ïù¥ÌÑ∞Ïùò Î†àÎ≤ÑÎ¶¨ÏßÄÏôÄ ÏûîÏ∞®Î•º ÎèôÏãúÏóê Î≥¥Î†§Î©¥ plot_leverage_resid2 Î™ÖÎ†πÏùÑ ÏÇ¨Ïö©ÌïúÎã§. Ïù¥ Î™ÖÎ†πÏùÄ xÏ∂ïÏúºÎ°ú ÌëúÏ§ÄÌôî ÏûîÏ∞®Ïùò Ï†úÍ≥±ÏùÑ ÌëúÏãúÌïòÍ≥† yÏ∂ïÏúºÎ°ú Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞íÏùÑ ÌëúÏãúÌïúÎã§. Îç∞Ïù¥ÌÑ∞ ÏïÑÏù¥ÎîîÍ∞Ä ÌëúÏãúÎêú Îç∞Ïù¥ÌÑ∞Îì§Ïù¥ Î†àÎ≤ÑÎ¶¨ÏßÄÍ∞Ä ÌÅ∞ ÏïÑÏõÉÎùºÏù¥Ïñ¥Ïù¥Îã§.

```python
sm.graphics.plot_leverage_resid2(result)
plt.show()
```

## Î≥¥Ïä§ÌÑ¥ ÏßëÍ∞í ÏòàÏ∏° Î¨∏Ï†ú

Î≥¥Ïä§ÌÑ¥ ÏßëÍ∞í Î¨∏Ï†úÏóê ÏïÑÏõÉÎùºÏù¥Ïñ¥Î•º Ï†ÅÏö©Ìï¥ Î≥¥Ïûê. MEDVÍ∞Ä 50Ïù∏ Îç∞Ïù¥ÌÑ∞Îäî ÏÉÅÏãùÏ†ÅÏúºÎ°ú ÏÉùÍ∞ÅÌï¥ÎèÑ Ïù¥ÏÉÅÌïú Îç∞Ïù¥ÌÑ∞Ïù¥ÎØÄÎ°ú ÏïÑÏõÉÎùºÏù¥Ïñ¥ÎùºÍ≥† ÌåêÎã®Ìï† Ïàò ÏûàÎã§. ÎÇòÎ®∏ÏßÄ Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú Ìè≠Ïä§ Ï∂îÏ≤úÍ≥µÏãùÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏïÑÏõÉÎùºÏù¥Ïñ¥Î•º Ï†úÏô∏Ìïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
from sklearn.datasets import load_boston

boston = load_boston()
dfX0 = pd.DataFrame(boston.data, columns=boston.feature_names)
dfX = sm.add_constant(dfX0)
dfy = pd.DataFrame(boston.target, columns=["MEDV"])

model_boston = sm.OLS(dfy, dfX)
result_boston = model_boston.fit()
pred = result_boston.predict(dfX)  # hat y
```

```python
influence_boston = result_boston.get_influence()
cooks_d2, pvals = influence_boston.cooks_distance
cooks_d2.shape
```

```python
K = influence_boston.k_vars
fox_cr = 4 / (len(dfy) - K - 1)
idx = np.where(cooks_d2 > fox_cr)[0]
idx
```

```python
fox_cr
```

```python
# MEDV = 50 Ï†úÍ±∞  : ÏúÑÏùò idxÏóê target Í∞íÏù¥ 50Ïù∏ idx Ï∂îÍ∞Ä
idx = np.hstack([idx, np.where(boston.target == 50)[0]])
idx
```

```python
%matplotlib inline
ax = plt.subplot()
plt.scatter(dfy, pred)
plt.scatter(dfy.MEDV[idx], pred[idx], s=300, c='r', alpha=0.3)
utils.annotate_axes(range(len(idx)), idx, list(zip(dfy.MEDV[idx], pred[idx])), [(-20, 15)] * len(idx), size="small", ax=ax)
plt.title("Î≥¥Ïä§ÌÑ¥ ÏßëÍ∞í Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïù¥ÏõÉÎùºÏù¥Ïñ¥ ÏÑ†ÌÉù", y=1.1)
plt.show()
```

Îã§ÏùåÏùÄ Ïù¥Î†áÍ≤å ÏïÑÏõÉÎùºÏù¥Ïñ¥Î•º Ï†úÏô∏Ìïú ÌõÑÏóê Îã§Ïãú ÌöåÍ∑ÄÎ∂ÑÏÑùÏùÑ Ìïú Í≤∞Í≥ºÏù¥Îã§.

```python
idx2 = list(set(range(len(dfX))).difference(idx)) # Ï†ÑÏ≤¥ idxÏóêÏÑú ÏïÑÏõÉÎùºÏù¥Ïñ¥ idx Ï†úÍ±∞
dfX = dfX.iloc[idx2, :].reset_index(drop=True)
dfy = dfy.iloc[idx2, :].reset_index(drop=True)
model_boston2 = sm.OLS(dfy, dfX)
result_boston2 = model_boston2.fit()
print(result_boston2.summary())
```

```python
# Ï°∞Í±¥ÏàòÍ∞Ä ÌÅ¨Í∏∞ ÎïåÎ¨∏Ïóê Ïä§ÏºÄÏùºÎßÅÏùÑ ÌïòÏó¨ Î™®Îç∏ÎßÅÌï®
feature_names = list(boston.feature_names)
feature_names.remove("CHAS")
feature_names = ["scale({})".format(name) for name in feature_names] + ["CHAS"]
df = pd.concat([dfX, dfy], axis=1)

model_boston3 = sm.OLS.from_formula("MEDV ~" + "+".join(feature_names), data=df)
result_boston3 = model_boston3.fit()
print(result_boston3.summary())
```
