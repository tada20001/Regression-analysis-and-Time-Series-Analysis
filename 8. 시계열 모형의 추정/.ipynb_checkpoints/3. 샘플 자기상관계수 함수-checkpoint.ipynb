## ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏùò Ï†ïÏùò

ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò(ACF: autocorrelation function)  ùúåùëò Îäî Ï†ïÏÉÅ ÌôïÎ•† Í≥ºÏ†ïÏùÑ ÌëúÌòÑÌï† Ïàò ÏûàÎäî Í∞ÄÏû• ÎåÄÌëúÏ†ÅÏù∏ ÌäπÏÑ±Ïù¥Îã§. Í∑∏Îü¨ÎÇò ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò  ùúå(ùëò) Ïùò Ï†ïÌôïÌïú Í∞íÏùÑ ÏïåÍ∏∞ ÏúÑÌï¥ÏÑúÎäî Ìï¥Îãπ Ï†ïÏÉÅ ÌôïÎ•† Í≥ºÏ†ïÏóê ÎåÄÌïú ÏàòÌïôÏ†ÅÏù∏ Î™®ÌòïÏùÑ ÏïåÍ≥† ÏûàÏñ¥Ïïº ÌïúÎã§. ÎßåÏïΩ ÏàòÌïôÏ†Å Î™®ÌòïÏù¥ ÏóÜÍ≥† ÏãúÍ≥ÑÏó¥ ÏûêÎ£å 

#### Ï¶â, ÌôïÎ•† Í≥ºÏ†ïÏùò ÏÉòÌîåÎßå Ï°¥Ïû¨ÌïòÎäî Í≤ΩÏö∞ÏóêÎäî Ïù¥Î°úÎ∂ÄÌÑ∞ ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò(Sample ACF)Î•º Íµ¨ÌïòÏó¨ Ïù¥Î°†Ï†ÅÏù∏ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏùò ÌòïÌÉúÎ•º Ïú†Ï∂îÌï¥Ïïº ÌïúÎã§.

ÏãúÍ≥ÑÏó¥ ÏûêÎ£å  {ùë¶ùë°;ùë°=1,‚ãØ,ùëõ} Í∞Ä Ï°¥Ïû¨ÌïòÎäî Í≤ΩÏö∞ ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò  ùëüùëò Îäî ÏÉòÌîå ÏûêÍ∏∞Í≥µÎ∂ÑÏÇ∞ Ìï®Ïàò  ùõæÃÇ ùëò Ïùò ÎπÑ(ratio)Î°ú Ï†ïÏùòÌïúÎã§.

$$\hat{\gamma}_k  = \dfrac{1}{n} \sum_{t=k+1}^n (y_t - \bar{y})(y_{t-k} - \bar{y})$$

$$\hat{r}_k = \dfrac{\hat{\gamma}_k}{\hat{\gamma}_0} = \dfrac{\sum_{t=k+1}^n (y_t - \bar{y})(y_{t-k} - \bar{y})}{\sum_{t=1}^n (y_t - \bar{y})^2}$$

ÏùºÎ∞òÏ†ÅÏúºÎ°úÎäî ÏÉòÌîå ÏûêÍ∏∞Í≥µÎ∂ÑÏÇ∞ Ìï®ÏàòÏùò Ï†ïÏùòÏóêÏÑú Î∂ÑÎ™®Î•º  ùëõ‚àíùëò Í∞Ä ÏïÑÎãå  ùëõ Î•º ÏÇ¨Ïö©ÌïúÎã§. Ïù¥Î†áÍ≤å Ìé∏Ìñ• Ïò§Ï∞®Í∞Ä ÏûàÎäî Ï∂îÏ†ïÍ∞íÏùÑ ÏÇ¨Ïö©ÌïòÎäî Í≤ÉÏùÄ ÏûêÍ∏∞Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨ÏùÑ ÏñëÌïúÏ†ï(positive definite)ÏúºÎ°ú ÎßåÎì§Í∏∞ ÏúÑÌïú Í≤ÉÏù¥Îã§.

## Python ÏóêÏÑú ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò Í≥ÑÏÇ∞

statsmodelsÏóêÏÑú ÏÉòÌîå ÏûêÍ∏∞Í≥µÎ∂ÑÏÇ∞ Ìï®Ïàò ÌòπÏùÄ ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÎ•º Íµ¨ÌïòÎ†§Î©¥ acovf ÌòπÏùÄ acf Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïúÎã§.

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("darkgrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
import statsmodels.api as sm
```

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1], [1, -0.7, 0.5])
y = p.generate_sample(100)
plt.plot(y, 'o-')
plt.show()
```

```python
acf = sm.tsa.acf(y, fft=False)  # ÏÉòÌîåÎ°ú ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÌï®Ïàò Íµ¨ÌïòÍ∏∞
acf
```

```python
plt.stem(acf, use_line_collection=True)
plt.yticks([-0.5, 0.0, 0.5, 1.0])
plt.show()
```

## ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏùò Î∂ÑÌè¨ ÌäπÏÑ±

ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏùò Í∞íÏùÄ Í≥ÑÏÇ∞Ïóê ÏÇ¨Ïö©Îêú ÏãúÍ≥ÑÏó¥ Í∞í Ï¶â, ÏÉòÌîåÏóê ÏùòÏ°¥ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú ÌôïÎ•† Î≥ÄÏàòÍ∞Ä ÎêúÎã§. ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏùò Î∂ÑÌè¨ ÌäπÏÑ±ÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

+ ÏÉòÌîå ÏãúÍ≥ÑÏó¥Ïùò Í∏∏Ïù¥  ùëõ Ïù¥ ÌÅ¨Î©¥,

    + ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÏùò Î∂ÑÌè¨Îäî Ï†ïÍ∑úÎ∂ÑÌè¨ÏôÄ ÎπÑÏä∑Ìï¥ÏßÑÎã§.

    + ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò  ùëüùëò Ïùò Í∏∞ÎåìÍ∞íÏùÄ Ïù¥Î°†Ï†ÅÏù∏ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò  ùúåùëò Ïù¥Îã§

    + ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò  ùëüùëò Ïùò Î∂ÑÏÇ∞ÏùÄ  ùëõ Ïóê Î∞òÎπÑÎ°Ä ÌïúÎã§.
    
    $$\text{Var}[r_k] \sim \dfrac{1}{n}$$

ÎßåÏïΩ ÌôïÎ•†Í≥ºÏ†ïÏù¥ Í∞ÄÏö∞ÏãúÏïà Î∞±ÏÉâ Ïû°ÏùåÏù∏ Í≤ΩÏö∞ÏóêÎäî ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÎäî Í∏∞ÎåìÍ∞í  ùúåùëò ÏôÄ Î∂ÑÏÇ∞  1/ùëõ ÏùÑ Í∞ÄÏßÄÎäî Ï†ïÍ∑úÎ∂ÑÌè¨Í∞Ä ÎêúÎã§.

$$\text{Var}[r_k] \sim \mathcal{N}\left(\rho_k, \dfrac{1}{n}\right)$$

statsmodelsÏùò acf Ìï®ÏàòÎäî alpha Ïù∏ÏàòÍ∞Ä Ï£ºÏñ¥ÏßÄÎäî Í≤ΩÏö∞, Ïù¥ Í∞íÏùÑ Ïú†Ïùò ÏàòÏ§Ä(significance level)Î°ú Í∞ÄÏßÄÎäî ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÏùò Ïã†Î¢∞Íµ¨Í∞ÑÏùÑ Ï∂úÎ†•ÌïúÎã§.

```python
acf, confint = sm.tsa.acf(y, alpha=0.05,fft=False)
confint
```

```python
plt.stem(acf, use_line_collection=True)
plt.plot(confint[:, 0], ":")
plt.plot(confint[:, 1], ":")
plt.show()
```

```python
plt.stem(acf, use_line_collection=True)
c = 0.5 * (confint[:, 1] - confint[:, 0])
plt.fill_between(np.arange(len(acf)), -c, c, alpha=0.2)
plt.show()
```

statsmodels.graphics.tsa.plot_acf Î™ÖÎ†πÏùÄ ÏÉòÌîå ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏôÄ ÌëúÏ§Ä Ïò§Ï∞®Ïùò ÌÅ¨Í∏∞Î•º ÎèôÏãúÏóê Í∑∏Î†§Ï§ÄÎã§.

```python
sm.graphics.tsa.plot_acf(y, lags=40)
plt.show()
```
