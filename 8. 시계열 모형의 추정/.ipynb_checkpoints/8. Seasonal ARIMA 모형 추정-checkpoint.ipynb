SARIMAX 클래스 이용하면 Multiplicated SARIMA(p,d,q)x(P,D,Q,s) 모형에 대한 추정 및 예측이 가능하다. 

클래스 인스턴스를 생성하기 위해서는 order 인수에 (p,d,q) 튜플을, seasonal_order 인수에 (P,D,Q,s) 튜플을 넣는다. SARIMAX 의 fit 메서드는 모수를 추정하여 그 결과를 SARIMAXResult 클래스 인스턴스로 반환한다.

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)

import statsmodels.api as sm
```

```python
def yearfraction2datetime(yearfraction, startyear=0):
    import datetime
    import dateutil
    year = int(yearfraction) + startyear
    month = int(round(12 * (yearfraction - year)))
    delta = dateutil.relativedelta.relativedelta(months=month)
    date = datetime.datetime(year, 1, 1) + delta
    return date
```

## 호흡기질환 사망자 수

```python
import pandas as pd
```

```python
data = sm.datasets.get_rdataset("deaths", "MASS")
df = data.data
df["datetime"] = df.time.map(yearfraction2datetime)
df["month"] = df.datetime.dt.month
df["y"] = np.log(df.value)
df.tail()
```

```python
df.plot(x="datetime", y="y")
plt.show()
```

```python
# 샘플의 자기상관계수함수 검토
sm.tsa.graphics.plot_acf(df.y, lags=(len(df)-1))
plt.show()
```

주기적으로 파형을 가지면서 지수적으로 감소하고 있다.

```python
m = sm.tsa.SARIMAX(df.y, order=(1, 0, 0), seasonal_order=(1, 1, 1, 12)) # AR(1) 모형 + 계절성 추세 추가 = SARIMA 모형 추정
r = m.fit()
print(r.summary())
```

잔차의 정규성과 자기상관계수 함수는 SARIMAXResult 클래스의 plot_diagnostics 메서드로 살펴볼 수 있다.

```python
r.plot_diagnostics()
plt.tight_layout()
plt.show()
```

```python
# 추정된 모수로 시뮬레이션
np.random.seed(0)

plt.subplot(7, 1, 1)
plt.plot(df.y, lw=3) # 원래 데이터

for i in range(6):
    plt.subplot(7, 1, i+2)
    plt.plot(r.simulate(len(df.y), initial_state=r.filtered_state[:, -1]))
    
plt.show()
```

```python
horizon = 60
pred = r.get_prediction(start=len(df), end=len(df) + horizon)
s = df.y.copy()
s[:-1] = np.nan # 마지막 데이터만 빼고 NaN
s = np.hstack([s, pred.predicted_mean]) # 133개 
ci = pred.conf_int(alpha=0.05)
```

```python
plt.plot(df.y)
plt.plot(s)
plt.fill_between(ci.index, ci["lower y"], ci["upper y"], color='r', alpha=0.3)
plt.show()
```

## 대기중 이산화탄소 농도 예측

```python
data = sm.datasets.get_rdataset("CO2")
df = data.data
df["datetime"] = df.time.map(yearfraction2datetime)
df["month"] = df.datetime.dt.month
df.tail()
```

```python
df.plot(x="datetime", y="value")
plt.show()
```

```python
# 모형 추정, 모수 추정
m = sm.tsa.SARIMAX(df.value, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
r = m.fit()
print(r.summary())
```

```python
r.plot_diagnostics()
plt.tight_layout()
plt.show()
```

```python
# 시뮬레이션
plt.plot(r.simulate(100, initial_state=r.filtered_state[:, -1]))
plt.show()
```

```python
np.random.seed(0)
initial_state = r.filtered_state[:, 12]
for i in range(6):
    plt.plot(r.simulate(len(df.value), initial_state=initial_state))
plt.plot(df.value, lw=5, alpha=0.5)
plt.show()
```

```python
# 예측
horizon = 120
pred = r.get_prediction(start=len(df), end=len(df) + horizon)
s = df.value.copy()
s[:-1] = np.nan
s = np.hstack([s, pred.predicted_mean])
ci = pred.conf_int(alpha=0.05)

plt.plot(df.value)
plt.plot(s)
plt.fill_between(ci.index, ci["lower value"], ci["upper value"], color='r', alpha=0.3)
plt.xlim(len(s) - 200, len(s))
plt.ylim(350, 390)
plt.show()
```

## 항공운송

```python
data = sm.datasets.get_rdataset("AirPassengers")
df = data.data
df["datetime"] = df.time.map(yearfraction2datetime)
df["y"] = np.log(df.value)
df["month"] = df.datetime.dt.month
df.tail()
```

```python
# 원래 데이터 분포 검토
df.plot(x="datetime", y="y")
plt.show()
```

```python
# 모형 추정, 모수 추정
m = sm.tsa.SARIMAX(df.y, trend="c", order=(1, 0, 0), seasonal_order=(1, 1, 1, 12), 
                   enforce_stationarity=False, enforce_invertibility=False)
r = m.fit()
print(r.summary())
```

```python
# 모형 진단 : 잔차의 정규성, 샘플자기상관계수

r.plot_diagnostics()
plt.tight_layout()
plt.show()
```

```python
# 시뮬레이션
np.random.seed(0)
initial_state = r.filtered_state[:, 12]
for i in range(6):
    plt.plot(r.simulate(len(df.y), initial_state=initial_state))
    
plt.plot(df.y, lw=3, alpha=0.5)
plt.show()
```

```python
# 예측
horizon = 360
pred = r.get_prediction(start=len(df), end=len(df) + horizon)
s = df.y.copy()
s = np.hstack([s, pred.predicted_mean])
ci = pred.conf_int(alpha=0.05)

plt.plot(df.y)
plt.plot(s)
plt.fill_between(ci.index, ci["lower y"], ci["upper y"], color='r', alpha=0.3)
plt.xlim(len(s) - 500, len(s))
plt.show()
```
