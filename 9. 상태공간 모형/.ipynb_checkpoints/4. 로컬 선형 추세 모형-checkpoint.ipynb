#### Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®Ìòï(local linear trend model)ÏùÄ ÎûúÎç§ ÏõåÌÅ¨ Î™®ÌòïÏùò Í¥ÄÏ∏°Í∞íÏù¥ Îã§Ïãú ÎàÑÏ†ÅÎêú Í≤ÉÏù¥ÎùºÍ≥† Î≥º Ïàò ÏûàÎã§.

Ïù¥ Î™®ÌòïÏùÄ ÏÉÅÌÉú Î≥ÄÏàò  ùõΩùë°,ùúáùë° Î•º Í∞ÄÏßÑÎã§.

\begin{eqnarray}
\beta_t &=& \beta_{t-1} + w_{\beta,t} &,\;\;\; & w_{\beta,t} \sim N(0, \sigma_\beta^2) \\
\mu_t &=& \mu_{t-1} + \beta_{t-1} + w_{\mu,t} &,\;\;\; & w_{\mu,t} \sim N(0, \sigma_\mu^2) \\
Y_{t} &=& \mu_{t} + v_t &,\;\;\; & v_t \sim {N}(0, \sigma_v^2) \\
\end{eqnarray}

Ïù¥ ÏàòÏãùÏùÑ ÌñâÎ†¨ Î∞©Ï†ïÏãù ÌòïÌÉúÎ°ú Ïì∞Î©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$\begin{bmatrix}
\beta_t \\ \mu_t
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 \\
1 & 1 \\
\end{bmatrix}
\begin{bmatrix}
\beta_{t-1} \\ \mu_{t-1}
\end{bmatrix}
+
\begin{bmatrix}
w_{\beta,t} \\
w_{\mu,t}
\end{bmatrix}$$

$$Y_t=
\begin{bmatrix}
0 & 1
\end{bmatrix}
\begin{bmatrix}
\beta_{t} \\ \mu_{t}
\end{bmatrix}
+
v_t$$

## ARIMA Î™®ÌòïÍ≥ºÏùò Í¥ÄÍ≥Ñ

Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏù¥ ARIMA Î™®Ìòï ÌòïÌÉúÎ°ú Î≥ÄÌôòÌï† Ïàò ÏûàÎã§.

$$\mu_t = Y_t - v_t$$

$$Y_t - v_t = Y_{t-1} - v_{t-1} + \beta_{t-1} + w_{\mu,t}$$

$$\Delta Y_t = Y_t - Y_{t-1} = \beta_{t-1} + w_{\mu,t} + v_t - v_{t-1}$$

$$\beta_{t-1} = \Delta Y_t - w_{\mu,t} - v_t + v_{t-1}$$

$$\Delta Y_{t+1} - w_{\mu,t+1} - v_{t+1} + v_{t} = \Delta Y_t - w_{\mu,t} - v_t + v_{t-1}  + w_{\beta,t}$$

$$\Delta Y_{t+1} - \Delta Y_t = w_{\mu,t+1} + v_{t+1} - v_{t} - w_{\mu,t} - v_t + v_{t-1}  + w_{\beta,t}$$

$$\Delta^2 Y_{t+1}  = w_{\mu,t+1} + v_{t+1} - v_{t} - w_{\mu,t} - v_t + v_{t-1}  + w_{\beta,t}$$

Ïù¥ ÏãùÏóêÏÑú  Œîùëåùë° Ïùò Í∏∞ÎåìÍ∞íÍ≥º ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Íµ¨ÌïòÎ©¥ Ïù¥ Í∞íÏóêÏÑú  Œî^2ùëåùë° Îäî MA(2) Î™®ÌòïÏùÑ Îî∞Î•¥Í≥†  ùëåùë° Îäî ARIMA(0,2,2) Î™®ÌòïÏùÑ Îî∞Î•¥Îäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§. 

#### Ï¶â Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùÄ ARIMA(0,2,2)Ïùò Îòê Îã§Î•∏ ÌëúÌòÑ(representation)Ïù¥ÎùºÍ≥† Î≥º Ïàò ÏûàÎã§.

## Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùò Ïòà 1

Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏóêÏÑú ÏÑ§Î™ÖÌïú Î∞∞Ïùò ÏòàÎ•º Îã§Ïãú Îì§Ïñ¥Î≥¥Ïûê. Ïñ¥Îñ§ ÏÑ¨ Í∑ºÏ≤òÏóêÏÑú Î∞∞Í∞Ä ÏóîÏßÑÏùÑ ÎÅÑÍ≥† Ï†ïÏßÄÌï¥ ÏûàÎã§Í∞Ä Ïñ¥Îäê ÏãúÏ†êÏóê ÏùºÏ†ï ÏÜçÎèÑÎ°ú ÏõÄÏßÅÏù¥Í∏∞ ÏãúÏûëÌñàÎã§.

Î∞∞Ïùò ÏÜçÎèÑ  ùõΩùë° Îäî ÏùºÏ†ïÌïòÍ≤å Ïú†ÏßÄÌïòÍ≥†Ïûê ÌïòÏßÄÎßå Í∏∞Í¥ÄÏùò ÌäπÏÑ±ÏÉÅ Ï°∞Í∏àÏî© ÏÜçÎèÑÏùò Î≥ÄÌôîÍ∞Ä ÎàÑÏ†ÅÎê† Ïàò ÏûàÎã§. Ïù¥ ÏÜçÎèÑ Î≥ÄÌôî  ùë§ùõΩ,ùë°  Îäî Í∞ÄÏö∞ÏãúÏïà Ï†ïÍ∑ú Î∂ÑÌè¨Î•º Îî∞Î•∏Îã§Í≥† Í∞ÄÏ†ïÌïòÏûê.

ÎòêÌïú Î∞∞Ïùò ÏÜçÎèÑ  ùõΩùë°  Ïóê ÏùòÌï¥ Î∞∞Ïùò ÏúÑÏπò  ùúáùë°  Îäî Î≥ÄÌôîÌïòÍ≤å ÎêòÎäîÎç∞ Ïù¥ Í≤ΩÏö∞ÏóêÎèÑ ÌååÎèÑÏùò ÌûòÏóê ÏùòÌï¥ Ï°∞Í∏àÏî© ÏïûÎí§Î°ú ÏúÑÏπòÍ∞Ä Î∞îÎÄî Ïàò ÏûàÍ≥† Ïù¥ ÎØ∏ÏÑ∏Ìïú ÏúÑÏπò Î≥ÄÌôî  ùë§ùúá,ùë°  ÎèÑ Í∞ÄÏö∞ÏãúÏïà Ï†ïÍ∑ú Î∂ÑÌè¨Î•º Îî∞Î•∏Îã§Í≥† Í∞ÄÏ†ïÌïòÏûê.

ÎßàÏßÄÎßâÏúºÎ°ú Îß§ ÏãúÍ∞ÑÎßàÎã§ Î∞∞Ïùò ÏÑ†ÏõêÏùÄ ÏÑ¨ÏúºÎ°úÎ∂ÄÌÑ∞Ïùò Í±∞Î¶¨  ùëåùë° Î•º Î∞∞Ïóê ÏûàÎäî Ï¥àÏùåÌåå Í±∞Î¶¨Í≥Ñ(ultrasonic range finder)Î°ú Ï∏°Ï†ïÌïúÎã§. Ï¥àÏùåÌåå Í±∞Î¶¨Í≥ÑÎ°ú Ïû∞ Í∞í  ùëåùë° Îäî Í∞ÄÏö∞ÏãúÏïà Ï†ïÍ∑ú Î∂ÑÌè¨Ïóê Îî∞Î•¥Îäî Ïò§Ï∞®Î•º Í∞ÄÏßÄÍ≥† ÏûàÎã§Îäî Í≤ÉÏùÑ Ï†úÏ°∞ÏÇ¨ Ï†ïÎ≥¥Î°úÎ∂ÄÌÑ∞ ÏïåÍ≥† ÏûàÎã§.

#### Ïù¥Îü¨Ìïú Í≤ΩÏö∞Ïóê Ï†ïÌôïÌïú Í±∞Î¶¨  ùúáùë° Î•º ÏïåÍ≥†Ïûê ÌïòÎäî Í≤ÉÏù¥ Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùò ÌïÑÌÑ∞ÎßÅ(filtering) Î¨∏Ï†úÍ∞Ä ÎêúÎã§.

#### Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùÄ Ïò§Ï∞®Í∞Ä ÌÅ∞ GPS Ïã†Ìò∏Î•º ÏÇ¨Ïö©Ìïú ÎÑ§ÎπÑÍ≤åÏù¥ÏÖòÏóêÏÑú Ï∞®ÎüâÏùò Ï†ïÌôïÌïú ÏúÑÏπòÎ•º Ï∂îÏ†ïÌïòÎäîÎç∞ÎèÑ ÏÇ¨Ïö©ÎêúÎã§.

+ ùúáùë° : Î∞∞Ïùò ÏúÑÏπò
+ ùõΩùë° : Î∞∞Ïùò ÏÜçÎèÑ
+ ùë§ùõΩ,ùë° : Î∞∞Ïùò ÏÜçÎèÑ Î≥ÄÌôî. Í∞ÄÏö∞ÏãúÏïà Ï†ïÍ∑úÎ∂ÑÌè¨Î•º Îî∞Î•∏Îã§Í≥† Í∞ÄÏ†ï
+  ùë§ùúá,ùë° : Î∞∞Ïùò ÏúÑÏπò Î≥ÄÌôî. Í∞ÄÏö∞ÏãúÏïà Ï†ïÍ∑úÎ∂ÑÌè¨Î•º Îî∞Î•∏Îã§Í≥† Í∞ÄÏ†ï
+ ùëåùë° : ÏÑ¨ÏúºÎ°úÎ∂ÄÌÑ∞Ïùò Î∞∞Ïùò Í±∞Î¶¨

## Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùò Ïòà 2

#### Ïã§Ï†úÎ°ú Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùÑ ÏÇ¨Ïö©ÌïòÎäî Í≤ΩÏö∞Îäî ÏÜçÎèÑ Î≥ÄÌôîÍ∞Ä Í±∞Ïùò ÏóÜÎäî Í≤ΩÏö∞ Ï¶â,  ùë§ùõΩ,ùë° Í∞Ä ÏûëÏùÄ Í≤ΩÏö∞Í∞Ä ÎßéÎã§. 

#### Ïù¥ Í≤ΩÏö∞ÏóêÎäî Ïù¥Î¶Ñ Í∑∏ÎåÄÎ°ú  ùëåùë° Í∞íÏù¥ ÏÑ†Ìòï Ï∂îÏÑ∏Ïóê Í∞ÄÍπåÏö¥ Î™®ÏäµÏùÑ Î≥¥Ïù∏Îã§. 

Îã§Îßå Ï∂îÏÑ∏Ïùò Í∏∞Ïö∏Í∏∞Í∞Ä Í≥†Ï†ïÎêú ÏùºÎ∞òÏ†ÅÏù∏ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏóê ÎπÑÌï¥ Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏Îäî Ï∂îÏÑ∏Í∞Ä Íµ≠ÏÜåÏ†Å(local)ÏúºÎ°úÎßå Ïú†ÏßÄÎêúÎã§Í≥† Í∞ÄÏ†ïÌïòÍ∏∞ ÎïåÎ¨∏Ïóê Ï∂îÏÑ∏ ÏûêÏ≤¥Í∞Ä Ï°∞Í∏àÏî© Î∞îÎÄî ÏàòÎäî ÏûàÎã§.

#### Í±∞ÏãúÍ≤ΩÏ†ú(macro economics) Î™®ÌòïÏóêÏÑú ÏÇ¨Ïö©ÌïòÎäî ÎåÄÎ∂ÄÎ∂ÑÏùò ÏßÄÌëúÍ∞íÏùÄ ÏãúÍ∞ÑÏù¥ ÏßÄÎÇòÎ©∞ ÏÑ†ÌòïÏ†Å ÌòπÏùÄ ÏßÄÏàòÌï®ÏàòÏ†ÅÏúºÎ°ú Ï¶ùÍ∞ÄÌïòÎäî Í≤ΩÏö∞Í∞Ä ÎßéÎã§. Ïù¥Îü¨Ìïú ÏãúÍ≥ÑÏó¥ Í∞íÏùÄ Î™®Îëê Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏúºÎ°ú Î™®ÌòïÌôîÍ∞Ä Í∞ÄÎä•ÌïòÎã§.

## Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùò ÏãúÎÆ¨Î†àÏù¥ÏÖò

#### statsmodelsÏùò KalmanFilter ÌÅ¥ÎûòÏä§ÏôÄ UnobservedComponents ÌÅ¥ÎûòÏä§Î•º ÏÇ¨Ïö©ÌïòÏó¨ 

Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùÑ Îî∞Î•¥Îäî ÏãúÍ≥ÑÏó¥ÏùÑ ÎßåÎì§Í≥† ÌïÑÌÑ∞ÎßÅÏùÑ Ìï¥Î≥¥Ïûê. 

 + ÌïÑÌÑ∞ÎßÅ : ÏòàÏ∏° + ÌòÑÏû¨Ïùò ÏÉÅÌÉúÎ≥ÄÏàòÍ∞í x Ï∂îÏ†ï
 + ÏπºÎßå ÌïÑÌÑ∞ : ÏÑ†Ìòï ÎèôÏ†Å Î™®ÌòïÏù¥ Í∞ÄÏö∞ÏãúÏïà Î∂ÑÌè¨Î•º Í∞ÄÏßÄÎäî ÏÑ†Ìòï ÏãúÏä§ÌÖúÏù¥ÎùºÎäî Ï†úÏïΩÏ°∞Í±¥ ÌïòÏóê ÏÉÅÌÉú Î≥ÄÏàòÏùò ÌèâÍ∑†Í≥º Î∂ÑÏÇ∞ÏùÑ Í∞±Ïã†ÌïòÎäî ÏïåÍ≥†Î¶¨Ï¶ò

#### 1. Ïö∞ÏÑ† KalmanFilter ÌÅ¥ÎûòÏä§Î°ú Ïù¥Î°†Ï†ÅÏù∏ Î°úÏª¨ ÏÑ†Ìòï Ï∂îÏÑ∏ Î™®ÌòïÏùÑ ÎßåÎì†Îã§.

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)

from statsmodels.tsa.statespace.kalman_filter import KalmanFilter
```

```python
# Î™®Ìòï ÏÉùÏÑ±

# k_state : ùõΩùë°,ùúáùë° ÏÉÅÌÉúÎ≥ÄÏàòÏùò ÏàòÍ∞Ä 2Í∞ú, ÏúÑÏùò ÏÇ¨Î°Ä
# transition : ÏÉÅÌÉúÏ†ÑÏù¥ÌñâÎ†¨.. Ïù¥Ï†ÑÏùò ÏÉÅÌÉú  ùë•ùë°‚àí1 ÏôÄ ÌòÑÏû¨ ÏÉùÏÑ±Îêú innovation ÏûÖÎ†•  ùë§ùë° Ïóê ÏùòÌï¥ ÌòÑÏû¨Ïùò ÏÉÅÌÉú  ùë•ùë° Í∞Ä ÏÉùÏÑ±ÎêòÎäî Í¥ÄÍ≥ÑÎ•º ÌëúÌòÑ
# state_intercept : ÏÉÅÌÉú Ï†ÑÏù¥ÏãùÏùò Ï†àÌé∏
# selection : Ïù¥ÎÖ∏Î≤†Ïù¥ÏÖò Í≥ºÏ†ïÏóê ÎåÄÌïú ÏÑ†ÌÉù ÌñâÎ†¨
# state_cov: Ïù¥ÎÖ∏Î≤†Ïù¥ÏÖò Í≥ºÏ†ïÏùò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨
# design : ÏÑ§Í≥Ñ ÌñâÎ†¨(design matrix)
# obs_cov : Í¥ÄÏ∏° ÌñâÎ†¨Ïùò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨
mod1 = KalmanFilter(k_states=2, k_endog=1, 
                    transition=[[1, 0], [1, 1]],
                    state_intercept=[[0], [3]],
                    selection=[[1, 0], [0, 1]],
                    state_cov = [[1, 0], [0, 1]],
                    design=[[0, 1]], obs_cov=[[500]])
```

```python
# ÏãúÎÆ¨Î†àÏù¥ÏÖò, ÌëúÎ≥∏ ÏÉùÏÑ±
np.random.seed(0)
y1, x1 = mod1.simulate(100)

plt.plot(y1, label="measurement")  # ÏÑ¨Í≥ºÏùò Í±∞Î¶¨
plt.plot(x1.T[0], label="speed state")  # Î∞∞Ïùò ÏÜçÎèÑ
plt.plot(x1.T[1], label="position state") # Î∞∞Ïùò ÏúÑÏπò
plt.legend(loc=2)
plt.title("Local Linear Trend Model")
plt.show()
```

#### 2. ÌïÑÌÑ∞ÎßÅ(ÏÉÅÌÉúÎ≥ÄÏàò Î∞è Î™®Ìòï Ï∂îÏ†ï)

Ïù¥Ï†ú ÏπºÎßå ÌïÑÌÑ∞ÎßÅÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏÉÅÌÉúÎ≥ÄÏàòÎ•º Ï∂îÏ†ïÌïòÏó¨ Î≥¥Ïûê. UnobservedComponents ÌÅ¥ÎûòÏä§Î•º ÏÇ¨Ïö©ÌïòÏó¨ Î™®Ìòï Ï∂îÏ†ïÍπåÏßÄ Ìïú Î≤àÏóê ÏàòÌñâÌï† Ïàò ÏûàÎã§.

```python
import statsmodels.api as sm
```

```python
mod1f = sm.tsa.UnobservedComponents(y1, 'local linear trend')
res1f = mod1f.fit()
print(res1f.summary())
```

```python
plt.plot(y1, label="measurement")
plt.plot(x1.T[0], label="original speed state")
plt.plot(x1.T[1], label="original position state")
plt.plot(res1f.filtered_state[0], label="filtered position state")
plt.plot(res1f.filtered_state[1], label="filtered speed state")
plt.legend(loc=2)
plt.title("Kalman Filtering of Local Linear Trend Model")
plt.show()
```

```python
# Í¥ÄÏ∏°Ïπò vs ÏòàÏ∏°Ïπò, Ïã†Î¢∞Íµ¨Í∞Ñ
res1f.plot_components(legend_loc=2, figsize=(8, 10))
plt.tight_layout(pad=3, h_pad=0.8)
plt.show()
```

```python
# ÏßÑÎã®
res1f.plot_diagnostics()
plt.tight_layout()
plt.show()
```

```python
# ÏòàÏ∏°

forecast = res1f.get_forecast(50)  # Ìñ•ÌõÑ 50Í∞ú ÏòàÏ∏°ÏùÑ ÏúÑÌïú Ïù∏Ïä§ÌÑ¥Ïä§ ÏÉùÏÑ±
plt.plot(y1, label="measurement")
forecast_ci = forecast.conf_int() # Ïã†Î¢∞Íµ¨Í∞Ñ
forecast_index = np.arange(len(y1), len(y1) + len(forecast_ci))
plt.plot(forecast_index, forecast.predicted_mean, label="predicted measurement")
plt.fill_between(forecast_index, forecast_ci[:, 0], forecast_ci[:, 1], alpha=0.1)
plt.show()
```
