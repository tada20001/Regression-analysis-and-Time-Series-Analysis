ÏãúÍ≥ÑÏó¥ ÏûêÎ£å Ï§ëÏóêÎäî ÏãúÏ∞®Í∞Ä Ï¶ùÍ∞ÄÌï¥ÎèÑ ÏûêÍ∏∞ ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä ÎπÑÍµêÏ†Å Ïò§Îû´ÎèôÏïà 0Ïù¥ ÏïÑÎãå Í∞íÏúºÎ°ú ÎÇ®ÏïÑ ÏûàÎäî Í≤ÉÎì§Ïù¥ ÎßéÎã§. 

#### Ïù¥Îü¨Ìïú ÏãúÍ≥ÑÏó¥Ïóê MA Î™®ÌòïÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Ï∞®ÏàòÍ∞Ä ÎÑàÎ¨¥ Ïª§ÏßÑÎã§. Ïù¥Î•º Í∑πÎ≥µÌïòÍ∏∞ ÏúÑÌïú Î™®Ìòï Ï§ëÏùò ÌïòÎÇòÍ∞Ä Autoregressive (AR) Î™®ÌòïÏù¥Îã§. 

AR Î™®ÌòïÏùÄ Î∞±ÏÉâ Ïû°ÏùåÏùò ÌòÑÏû¨Í∞íÍ≥º ÏûêÍ∏∞ ÏûêÏã†Ïùò Í≥ºÍ±∞Í∞íÏùò ÏÑ†Ìòï Í∞ÄÏ§ëÌï©ÏúºÎ°ú Ïù¥Î£®Ïñ¥ÏßÑ Ï†ïÏÉÅ ÌôïÎ•† Î™®ÌòïÏù¥Îã§.

$$Y_t = -\phi_1 Y_{t-1} -\phi_2 Y_{t-2} -\cdots -\phi_p Y_{t-p} + \epsilon_t$$

ÏûêÍ∏∞ ÏûêÏã†Ïùò Í≥ºÍ±∞ Í∞íÏùÑ ÏÇ¨Ïö©ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê Autoregressive ÎùºÎäî Ïù¥Î¶ÑÏù¥ Î∂ôÏó¨Ï°åÎã§. AR Î™®ÌòïÏùÄ ÏûÑÏùòÏùò Í≥ÑÏàò Í∞íÏùÑ Í∞ÄÏßà Ïàò ÏûàÎäî MA Î™®ÌòïÍ≥º Îã¨Î¶¨ Ï†ïÏÉÅ Í≥ºÏ†ïÏù¥ ÎêòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî Í≥ÑÏàòÏùò Í∞íÏóê Ï†úÌïú Ï°∞Í±¥Ïù¥ Î∂ôÎäîÎã§. 

Íµ¨Ï≤¥Ï†ÅÏù∏ Ï†úÌïú Ï°∞Í±¥ÏùÄ Ï∞®ÏàòÏóê Îî∞Îùº Îã¨ÎùºÏßÑÎã§. AR Î™®ÌòïÏùÄ ÏñºÌïè Î≥¥Í∏∞Ïóê ÏùºÎ∞ò ÏÑ†Ìòï ÌôïÎ•† Í≥ºÏ†ï(general linear process)Ïùò ÏàòÏãùÏùÑ Îî∞Î•¥ÏßÄ ÏïäÎäî Í≤ÉÏ≤òÎüº Î≥¥Ïù¥ÏßÄÎßå ÏùºÎ∞ò ÏÑ†Ìòï ÌôïÎ•† Í≥ºÏ†ïÏùò ÌòïÌÉúÎ°ú Î≥ÄÌôòÏù¥ Í∞ÄÎä•ÌïòÎã§.

## AR(1) Î™®Ìòï

1Ï∞® AR Î™®Ìòï Ï¶â, AR(1) Î™®ÌòïÏùÄ Î∞±ÏÉâ Ïû°ÏùåÏùò ÌòÑÏû¨Í∞íÍ≥º 1Ïä§ÌÖù Í≥ºÍ±∞Ïùò ÏûêÍ∏∞ ÏûêÏã†Ïùò Í∞íÎßåÏùò Í∞ÄÏ§ëÌï©ÏúºÎ°ú Ïù¥Î£®Ïñ¥ÏßÑ Î™®ÌòïÏù¥Îã§. ÏàòÏãùÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$Y_t = -\phi Y_{t-1} + \epsilon_t$$

Ïù¥ ÏàòÏãùÏóêÏÑú Í≥ÑÏàò  ùúô Îäî Îã§Ïùå Ï°∞Í±¥ÏùÑ ÎßåÏ°±Ìï¥Ïïº ÌïúÎã§.

$$-1 < \phi < 1$$

AR(1) Î™®ÌòïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏùºÎ∞ò ÏÑ†Ìòï ÌôïÎ•† Í≥ºÏ†ïÏùò ÌòïÌÉúÎ°ú Î∞îÍøÄ Ïàò ÏûàÎã§.

\begin{aligned}
Y_t 
&= -\phi Y_{t-1} + \epsilon_t \\
&= -\phi \left( -\phi Y_{t-2} + \epsilon_{t-1} \right) + \epsilon_t \\
&= \phi^2 Y_{t-2} -\phi \epsilon_{t-1} + \epsilon_t \\
&= \phi^2  \left( -\phi Y_{t-3} + \epsilon_{t-2} \right) -\phi \epsilon_{t-1} + \epsilon_t \\
&= -\phi^3 Y_{t-3} + \phi^2 \epsilon_{t-2} -\phi \epsilon_{t-1} + \epsilon_t \\
&\vdots& \\
&= \epsilon_t -\phi \epsilon_{t-1} +\phi^2 \epsilon_{t-2} -\phi^3 \epsilon_{t-3} + \cdots  \\
\end{aligned}

AR(1) Î™®ÌòïÏùò Í∏∞ÎåÄÍ∞íÏùÄ MAÎ™®ÌòïÍ≥º ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú 0Ïù¥Îã§.

$$\text{E}[Y_t] = 0$$

Ïù¥Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï¶ùÎ™ÖÌï† Ïàò ÏûàÎã§.

\begin{aligned}
\text{E}[Y_t] = \mu 
&= \text{E} \left[ -\phi  Y_{t-1} - \epsilon_{t-1} \right] \\
&= -\phi\text{E} [ Y_{t-1} ] -\text{E} [ \epsilon_{t-1} ] \\
&= -\phi \mu -   0 \\ 
(1+\phi)\mu &= 0 \\
\mu &= 0 \; \text{ if } \phi \neq -1 \\
\end{aligned}

AR(1) Î™®ÌòïÏùò Î∂ÑÏÇ∞ÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$\gamma_0 = \dfrac{\sigma_\epsilon^2}{1-\phi^2}$$

Ïù¥Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï¶ùÎ™ÖÌï† Ïàò ÏûàÎã§.

Ïö∞ÏÑ†  ùúñùë° Îäî Í≥ºÍ±∞Ïùò Í∞í  ùëåùë°‚àíùëô  (for  ùëô>1 )Ïóê ÏòÅÌñ•ÏùÑ Ï£ºÏßÄ ÏïäÍ∏∞ ÎïåÎ¨∏Ïóê ÎèÖÎ¶ΩÏù¥Îã§. Îî∞ÎùºÏÑú

$$\text{E}[\epsilon_tY_{t-l}] = \text{E}[\epsilon_t]\text{E}[Y_{t-l}] = 0$$

Ïù¥Îã§.

\begin{aligned}
\text{Var}[Y_t] = \gamma_0 = \text{E}[Y_t^2] 
&= \text{E} \left[ (-\phi  Y_{t-1} - \epsilon_{t})^2 \right] \\
&= \text{E} \left[ \phi^2  Y_{t-1}^2 + 2\phi Y_{t-1} \epsilon_{t-1} + \epsilon_{t}^2 \right] \\
&= \phi^2\text{E} [ Y_{t-1}^2 ] + 2 \phi \text{E} [ Y_{t-1} \epsilon_{t} ] + \text{E} [ \epsilon_{t}^2 ] \\
&= \phi^2 \gamma_0 + 0 + \sigma_\epsilon^2
\end{aligned}

AR(1) Î™®ÌòïÏùò ÏûêÍ∏∞Í≥µÎ∂ÑÏÇ∞ÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$\gamma_l = (-\phi)^l \dfrac{\sigma_\epsilon^2}{1-\phi^2}$$

ÏûêÍ∏∞Í≥µÎ∂ÑÏÇ∞ÏúºÎ°úÎ∂ÄÌÑ∞ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

$$\rho_l = \dfrac{\gamma_l}{\gamma_0} = (-\phi)^l$$

1Í≥º -1ÏÇ¨Ïù¥Ïùò Ïó¨Îü¨Í∞ÄÏßÄ  ùúô Ïóê ÎåÄÌï¥ AR(1) Î™®ÌòïÏùò ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò Ï¶â, ACF(Ïù¥Î°†Ï†Å ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò)Î•º Í∑∏Î¶¨Î©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

+ ùúô<0  Ïù¥Î©¥ ACFÎäî ÏßÄÏàòÌï®ÏàòÏ†ÅÏúºÎ°ú Í∞êÏÜåÌïúÎã§.
+ ùúô>0  Ïù¥Î©¥ ACFÎäî Î∂ÄÌò∏Î•º Î∞îÍøîÍ∞ÄÎ©¥ÏÑú (ÏßÑÎèôÌïòÎ©¥ÏÑú) ÏßÄÏàòÌï®ÏàòÏ†ÅÏúºÎ°ú Í∞êÏÜåÌïúÎã§.

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
lag = np.arange(12)

plt.subplot(221)
acf = 0.9 ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-0.1, 1.1)
plt.xlabel("lag")
plt.ylabel("ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò")
plt.title("$\phi=0.9$")

plt.subplot(222)
acf = 0.4 ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-0.1, 1.1)
plt.xlabel("lag")
plt.ylabel("ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò")
plt.title("$\phi=0.4$")

plt.subplot(223)
acf = (-0.8) ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-1.1, 1.1)
plt.xlabel("lag")
plt.ylabel("ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò")
plt.title("$\phi=-0.8$")

plt.subplot(224)
acf = (-0.5) ** lag
plt.stem(acf, use_line_collection=True)
plt.xlim(-0.2, 11.2)
plt.ylim(-1.1, 1.1)
plt.xlabel("lag")
plt.ylabel("ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò")
plt.title("$\phi=-0.5$")

plt.suptitle("AR(1) Î™®ÌòïÏùò ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò ÏÇ¨Î°Ä", y=1.1)
plt.tight_layout()
plt.show()
```

ÏïûÏùò Í∑∏Î¶ºÏóêÏÑú Î≥º Ïàò ÏûàÎìØÏù¥  ùúô Ïùò Ï†àÎåÄÍ∞íÏù¥ 1Ïóê Í∞ÄÍπåÏö∏ ÏàòÎ°ù ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÏùò Í∞êÏÜå ÏÜçÎèÑÍ∞Ä ÏûëÏïÑÏßÑÎã§. 

#### Ï¶â, ÏãúÏ∞®Í∞Ä Ï¶ùÍ∞ÄÌï¥ÎèÑ ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä ÌÅ¨Í≤å ÏûëÏïÑÏßÄÏßÄ ÏïäÎäîÎã§. 

Í∑∏Îüº ÎßåÏïΩ  ùúô Ïùò Ï†àÎåÄÍ∞íÏù¥ 1Ïù¥Í±∞ÎÇò 1Î≥¥Îã§ ÌÅ¨Îã§Î©¥ Ïñ¥ÎñªÍ≤å Îê†Íπå?

#### ùúô Í∞íÏùò ÌÅ¨Í∏∞Í∞Ä 1 Ïù¥ÏÉÅÏù¥ ÎêòÎ©¥ Î∂ÑÏÇ∞Ïù¥ Í≥ÑÏÜç Ï¶ùÍ∞ÄÌïòÎäî ÎπÑÏ†ïÏÉÅ Í≥ºÏ†ïÏù¥ ÎêúÎã§.  ùúô=‚àí1 Ïù∏ Í≤ΩÏö∞Î•º Î≥¥Ïûê.  (Î∂ÑÏÇ∞Ïù¥ ÏãúÍ∞ÑÏóê Îî∞Îùº Î≥ÄÌïòÍ∏∞ ÎïåÎ¨∏)

\begin{aligned}
Y_t &= Y_{t-1} + \epsilon_t \\
\text{Var}[Y_t] &= \text{Var}[Y_{t-1} + \epsilon_t] \\
 &= \text{Var}[Y_{t-1}] + \text{Var}[\epsilon_t]  \;\; (\text{independence})\\
\end{aligned}

$$\text{Var}[Y_t] > \text{Var}[Y_{t-1}]$$

AR(1) Î™®ÌòïÏùò Í≤ΩÏö∞  |ùúô|<1  Ï°∞Í±¥ÏùÄ Ï†ïÏÉÅÏÉÅÌÉú Ï°∞Í±¥(stationarity condition)Ïù¥ ÎêúÎã§.

### AR(1) Î™®ÌòïÏùò ÏãúÎÆ¨Î†àÏù¥ÏÖò

```python
import statsmodels.api as sm

np.random.seed(0)
p1 = sm.tsa.ArmaProcess([1, -0.9], [1])
y1 = p1.generate_sample(100, burnin=100)
plt.plot(y1, 'o-')
plt.title("1Ï∞® AR Î™®ÌòïÏùò ÏãúÎÆ¨Î†àÏù¥ÏÖò")
plt.show()
```

Ïù¥ ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Ïùò  ùëô -ÏãúÏ∞® ÏÉÅÍ¥ÄÍ≥ÑÏàò, Ï¶â  ùëåùë° ÏôÄ  ùëåùë°‚àíùëô Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàòÎ•º Ï°∞ÏÇ¨ÌïòÍ∏∞ ÏúÑÌï¥ Ïä§Ï∫êÌÑ∞ÌîåÎ°ØÏùÑ Í∑∏Î†§Î≥¥ÏïòÎã§. 

#### ÏßÄÏÜçÏ†ÅÏúºÎ°ú Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Î≥¥Ïù∏Îã§.

```python
import scipy as sp

plt.figure(figsize=(10, 4))

plt.subplot(131)
r, p = sp.stats.pearsonr(y1[1:], y1[:-1])
sns.scatterplot(y1[1:], y1[:-1])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-1}$")
plt.title("ÏãúÏ∞®Í∞Ä 1Ïù∏ Í≤ΩÏö∞Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàò\nr={0:.3f}, p-value={1:.3f}".format(r, p))

plt.subplot(132)
r, p = sp.stats.pearsonr(y1[2:], y1[:-2])
sns.scatterplot(y1[2:], y1[:-2])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-2}$")
plt.title("ÏãúÏ∞®Í∞Ä 2Ïù∏ Í≤ΩÏö∞Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàò\nr={0:.3f}, p-value={1:.3f}".format(r, p))

plt.subplot(133)
r, p = sp.stats.pearsonr(y1[3:], y1[:-3])
sns.scatterplot(y1[3:], y1[:-3])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-3}$")
plt.title("ÏãúÏ∞®Í∞Ä 3Ïù∏ Í≤ΩÏö∞Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàò\nr={0:.3f}, p-value={1:.3f}".format(r, p))

plt.tight_layout()
plt.show()
```

Ïù¥Î°†Ï†Å ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏôÄ ÌëúÎ≥∏ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÎ•º Í∑∏Î¶¨Î©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
plt.subplot(211)
plt.stem(p1.acf(11), use_line_collection=True)  # Ïù¥Î°†Ï†Å ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Íµ¨ÌïòÍ∏∞
plt.xlim(-1, 11)
plt.ylim(-0.4, 1.1)
plt.title("1Ï∞® AR Î™®ÌòïÏùò Ïù¥Î°†Ï†Å ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò")

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(y1, lags=10, ax=ax)
plt.xlim(-1, 11)
plt.ylim(-0.4, 1.1)
plt.title("1Ï∞® AR Î™®ÌòïÏùò ÌëúÎ≥∏ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò")

plt.tight_layout()
plt.show()
```

### AR(2) Î™®ÌòïÏùò ÏãúÎÆ¨Î†àÏù¥ÏÖò

```python
np.random.seed(0)
p2 = sm.tsa.ArmaProcess([1, -1, 0.6], [1])
y2 = p2.generate_sample(300)
plt.plot(y2, 'o-')
plt.title("2Ï∞® AR Î™®ÌòïÏùò ÏãúÎÆ¨Î†àÏù¥ÏÖò")
plt.show()
```

ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º ÏúÑÌïú Ïä§Ï∫êÌÑ∞ÌîåÎ°Ø, Ïù¥Î°†Ï†Å ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÏôÄ ÌëúÎ≥∏ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®ÏàòÎ•º Í∑∏Î¶¨Î©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
plt.figure(figsize=(10, 4))

plt.subplot(131)
r, p = sp.stats.pearsonr(y2[1:], y2[:-1])
sns.scatterplot(y2[1:], y2[:-1])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-1}$")
plt.title("ÏãúÏ∞®Í∞Ä 1Ïù∏ Í≤ΩÏö∞Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàò\n(r={0:.3f}, p-value={1:.3f})".format(r, p))

plt.subplot(132)
r, p = sp.stats.pearsonr(y2[2:], y2[:-2])
sns.scatterplot(y2[2:], y2[:-2])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-2}$")
plt.title("ÏãúÏ∞®Í∞Ä 2Ïù∏ Í≤ΩÏö∞Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàò\n(r={0:.3f}, p-value={1:.3f})".format(r, p))

plt.subplot(133)
r, p = sp.stats.pearsonr(y2[3:], y2[:-3])
sns.scatterplot(y2[3:], y2[:-3])
plt.axis("equal")
plt.xlabel("$Y_t$")
plt.ylabel("$Y_{t-3}$")
plt.title("ÏãúÏ∞®Í∞Ä 3Ïù∏ Í≤ΩÏö∞Ïùò ÏÉÅÍ¥ÄÍ≥ÑÏàò\n(r={0:.3f}, p-value={1:.3f})".format(r, p))

plt.tight_layout()
plt.show()
```

ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàòÎäî ÏãúÏ∞® 1Ïùº Í≤ΩÏö∞ 0.634Ïù¥Îã§Í∞Ä ÏãúÏ∞® 2Ïù∏ Í≤ΩÏö∞ÏóêÎäî Í±∞Ïùò Ï†úÎ°úÎ°ú ÎÇòÏò§Í≥† Îã§Ïãú ÏãúÏ∞®Í∞Ä 3Ïù∏ Í≤ΩÏö∞ ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä 0.347Ïù¥ ÎÇòÏò®Îã§. ÏßÑÎèôÌïòÎäî Í≤ÉÏ≤òÎüº Î≥¥Ïù∏Îã§.

```python
plt.subplot(211)
plt.stem(p2.acf(11), use_line_collection=True)
plt.xlim(-1, 11)
plt.ylim(-1.1, 1.1)
plt.title("2Ï∞® AR Î™®ÌòïÏùò Ïù¥Î°†Ï†Å ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò")

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(y2, lags=10, ax=ax)
plt.xlim(-1, 11)
plt.ylim(-1.1, 1.1)
plt.title("2Ï∞® AR Î™®ÌòïÏùò ÌëúÎ≥∏ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò Ìï®Ïàò")

plt.tight_layout()
plt.show()
```
