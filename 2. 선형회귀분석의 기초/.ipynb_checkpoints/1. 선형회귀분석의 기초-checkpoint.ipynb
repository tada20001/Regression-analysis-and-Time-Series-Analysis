íšŒê·€ë¶„ì„(regression analysis)ì€  ğ· ì°¨ì› ë²¡í„° ë…ë¦½ ë³€ìˆ˜  ğ‘¥ ì™€ ì´ì— ëŒ€ì‘í•˜ëŠ” ìŠ¤ì¹¼ë¼ ì¢…ì† ë³€ìˆ˜  ğ‘¦ ê°„ì˜ ê´€ê³„ë¥¼ ì •ëŸ‰ì ìœ¼ë¡œ ì°¾ì•„ë‚´ëŠ” ì‘ì—…ì´ë‹¤.

###### íšŒê·€ë¶„ì„ì—ëŠ” ê²°ì •ë¡ ì  ëª¨í˜•(deterministic Model)ê³¼ í™•ë¥ ì  ëª¨í˜•(probabilistic Model)ì´ ìˆë‹¤. 

###### ìš°ì„  ê²°ì •ë¡ ì  ëª¨í˜•ë¶€í„° ì„¤ëª…í•œë‹¤. 

ê²°ì •ë¡ ì  íšŒê·€ë¶„ì„ ëª¨í˜•ì€ ë…ë¦½ ë³€ìˆ˜  ğ‘¥ ì— ëŒ€í•´ ëŒ€ì‘í•˜ëŠ” ì¢…ì† ë³€ìˆ˜  ğ‘¦ ì™€ ê°€ì¥ ë¹„ìŠ·í•œ ê°’  ğ‘¦Ì‚  ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜  ğ‘“(ğ‘¥) ë¥¼ ì°¾ëŠ” ê³¼ì •ì´ë‹¤.

$$\hat{y} = f \left( x \right) \approx y$$

ë§Œì•½ ë…ë¦½ ë³€ìˆ˜  ğ‘¥ ì™€ ì´ì— ëŒ€ì‘í•˜ëŠ” ì¢…ì† ë³€ìˆ˜  ğ‘¦ ê°„ì˜ ê´€ê³„ê°€ ë‹¤ìŒê³¼ ê°™ì€ ì„ í˜• í•¨ìˆ˜  ğ‘“(ğ‘¥) ì´ë©´ ì„ í˜• íšŒê·€ë¶„ì„(linear regression analysis)ì´ë¼ê³  í•œë‹¤.

$$\hat{y} = w_0 + w_1 x_1 + w_2 x_2 + \cdots + w_D x_D = w_0 + w^Tx$$

ìœ„ ì‹ì—ì„œ  ğ‘¤0,â‹¯,ğ‘¤ğ· ë¥¼ í•¨ìˆ˜  ğ‘“(ğ‘¥) ì˜ ê³„ìˆ˜(coefficient)ì´ì ì´ ì„ í˜• íšŒê·€ëª¨í˜•ì˜ ëª¨ìˆ˜(parameter)ë¼ê³  í•œë‹¤.

## ìƒìˆ˜í•­ ê²°í•©

ìƒìˆ˜í•­ì´ 0ì´ ì•„ë‹Œ íšŒê·€ë¶„ì„ëª¨í˜•ì¸ ê²½ìš°ì—ëŠ” ìˆ˜ì‹ì„ ê°„ë‹¨í•˜ê²Œ ë§Œë“¤ê¸° ìœ„í•´ ë‹¤ìŒê³¼ ê°™ì´ ìƒìˆ˜í•­ì„ ë…ë¦½ë³€ìˆ˜ì— ì¶”ê°€í•œë‹¤. ì´ë¥¼ ìƒìˆ˜í•­ ê²°í•©(bias augmentation)ì‘ì—…ì´ë¼ê³  í•œë‹¤.

$$x_i =
\begin{bmatrix}
x_{i1} \\ x_{i2} \\ \vdots \\ x_{iD}
\end{bmatrix}
\rightarrow 
x_{i,a} =
\begin{bmatrix}
1 \\ x_{i1} \\ x_{i2} \\ \vdots \\ x_{iD}
\end{bmatrix}$$

ìƒìˆ˜í•­ ê²°í•©ì„ í•˜ê²Œ ë˜ë©´ ëª¨ë“  ì›ì†Œê°€ 1ì¸ ë²¡í„°ê°€ ì…ë ¥ ë°ì´í„° í–‰ë ¬ì— ì¶”ê°€ëœë‹¤.

$$X =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1D} \\
x_{21} & x_{22} & \cdots & x_{2D} \\
\vdots & \vdots & \vdots & \vdots \\
x_{N1} & x_{N2} & \cdots & x_{ND} \\
\end{bmatrix}
\rightarrow 
X_a =
\begin{bmatrix}
1 & x_{11} & x_{12} & \cdots & x_{1D} \\
1 & x_{21} & x_{22} & \cdots & x_{2D} \\
\vdots & \vdots & \vdots & \vdots & \vdots \\
1 & x_{N1} & x_{N2} & \cdots & x_{ND} \\
\end{bmatrix}$$

ì´ë ‡ê²Œ ë˜ë©´ ì „ì²´ ìˆ˜ì‹ì´ ë‹¤ìŒê³¼ ê°™ì´ ìƒìˆ˜í•­ì´ ì¶”ê°€ëœ ê°€ì¤‘ì¹˜ ë²¡í„°  ğ‘¤ ì™€ ìƒìˆ˜í•­ì´ ì¶”ê°€ëœ ì…ë ¥ ë°ì´í„° ë²¡í„°  ğ‘¥ ì˜ ë‚´ì ìœ¼ë¡œ ê°„ë‹¨íˆ í‘œì‹œëœë‹¤.

$$f(x) = w_0 + w_1 x_1 + w_2 x_2 + \cdots + w_D x_D
= 
\begin{bmatrix}
1 & x_1 & x_2 & \cdots & x_D
\end{bmatrix}
\begin{bmatrix}
w_0 \\ w_1 \\ w_2 \\ \vdots \\ w_D
\end{bmatrix}
= x_a^T w_a = w_a^T x_a$$

ìƒìˆ˜í•­ì´ 0ì´ ì•„ë‹Œ ê²½ìš°ì—ëŠ” í•­ìƒ ìƒìˆ˜í•­ ê²°í•©ì„ í•˜ê¸° ë•Œë¬¸ì— íŠ¹ë³„íˆ ë²¡í„° ê¸°í˜¸ë¥¼  ğ‘¥ğ‘  ë˜ëŠ”  ğ‘¤ğ‘ ë¼ê³  í‘œì‹œí•˜ì§€ ì•Šì•„ë„ ìƒìˆ˜í•­ ê²°í•©ì´ ë˜ì–´ ìˆëŠ” ê²½ìš°ê°€ ë§ë‹¤.

## OLS

###### OLS(Ordinary Least Squares)ëŠ” ê°€ì¥ ê¸°ë³¸ì ì¸ ê²°ì •ë¡ ì  ì„ í˜• íšŒê·€ ë°©ë²•ìœ¼ë¡œ 

##### ì”ì°¨ì œê³±í•©(RSS: Residual Sum of Squares)ì„ ìµœì†Œí™”í•˜ëŠ” ê°€ì¤‘ì¹˜ ë²¡í„°ë¥¼ í–‰ë ¬ ë¯¸ë¶„ìœ¼ë¡œ êµ¬í•˜ëŠ” ë°©ë²•ì´ë‹¤.

ìš°ë¦¬ê°€ ì‚¬ìš©í•˜ëŠ” ì˜ˆì¸¡ ëª¨í˜•ì€ ë‹¤ìŒê³¼ ê°™ì´ ìƒìˆ˜í•­ì´ ê²°í•©ëœ ì„ í˜• ëª¨í˜•ì´ë‹¤.

$$\hat{y} = Xw$$

ì´ë•Œ ì”ì°¨ ë²¡í„°(residual vector)  ğ‘’ ëŠ”

$$e = {y} - \hat{y} = y - Xw$$

ì´ê³  ì”ì°¨ ì œê³±í•©(RSS:residual sum of squares)ì€

\begin{eqnarray}
\text{RSS}
&=&  e^Te \\
&=& (y - Xw)^T(y - Xw) \\
&=& y^Ty - 2y^T X w + w^TX^TXw  
\end{eqnarray}

ì´ë‹¤. ì”ì°¨ì˜ í¬ê¸° ì¦‰, ì”ì°¨ ì œê³±í•©ì„ ê°€ì¥ ì‘ê²Œ í•˜ëŠ” ê°€ì¤‘ì¹˜ ë²¡í„°ë¥¼ êµ¬í•˜ê¸° ìœ„í•´ ì”ì°¨ ì œê³±í•©ì˜ ê·¸ë ˆë””ì–¸íŠ¸(gradient) ë²¡í„°ë¥¼ êµ¬í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

$$\dfrac{d \text{RSS}}{d w} = -2 X^T y + 2 X^TX w$$

ì”ì°¨ê°€ ìµœì†Œê°€ ë˜ëŠ” ìµœì í™” ì¡°ê±´ì€ ê·¸ë ˆë””ì–¸íŠ¸ ë²¡í„°ê°€ 0ë²¡í„°ì´ì–´ì•¼ í•˜ë¯€ë¡œ ë‹¤ìŒ ì‹ì´ ì„±ë¦½í•œë‹¤.

$$\dfrac{d \text{RSS}}{d w}  = 0$$

$$X^TX w^{\ast} = X^T y$$

ë§Œì•½  ğ‘‹ğ‘‡ğ‘‹  í–‰ë ¬ì˜ ì—­í–‰ë ¬ì´ ì¡´ì¬í•œë‹¤ë©´ ë‹¤ìŒê³¼ ê°™ì´ ìµœì  ê°€ì¤‘ì¹˜ ë²¡í„°  ğ‘¤âˆ— ë¥¼ êµ¬í•  ìˆ˜ ìˆë‹¤.

$$w^{\ast} = (X^TX)^{-1} X^T y$$

ğ‘‹ğ‘‡ğ‘‹  í–‰ë ¬ì˜ ì—­í–‰ë ¬ì´ ì¡´ì¬í•˜ê³  ìœ„ì—ì„œ êµ¬í•œ ê°’ì´ ìµœì €ê°’ì´ ë˜ë ¤ë©´  ğ‘‹ğ‘‡ğ‘‹ ê°€ ì–‘ì˜ ì •ë¶€í˜¸(positive definite) í–‰ë ¬ì´ì–´ì•¼ í•œë‹¤. ë§Œì•½  ğ‘‹ ê°€ í’€ë­í¬ê°€ ì•„ë‹ˆë©´ ì¦‰,  ğ‘‹ ì˜ ê° í–‰ë ¬ì´ ì„œë¡œ ë…ë¦½ì´ ì•„ë‹ˆë©´  ğ‘‹ğ‘‡ğ‘‹ ê°€ ì–‘ì˜ ì •ë¶€í˜¸ê°€ ì•„ë‹ˆê³  ì—­í–‰ë ¬ì´ ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë¯€ë¡œ ìœ„ì™€ ê°™ì€ í•´ë¥¼ êµ¬í•  ìˆ˜ ì—†ë‹¤.

## ì§êµ ë°©ì •ì‹

ì—¬ê¸°ì—ì„œ ê·¸ë ˆë””ì–¸íŠ¸ê°€ 0ë²¡í„°ê°€ ë˜ëŠ” ê´€ê³„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë‹¤ìŒ ì‹ì„ ì§êµ ë°©ì •ì‹(normal equation)ì´ë¼ê³  í•œë‹¤.

$$X^T y - X^TX w = 0$$

ì§êµ ë°©ì •ì‹ì„ ì¸ìˆ˜ ë¶„í•´í•˜ë©´

$$X^T (y - X w ) = 0$$

$$X^T e = 0$$

ì¦‰,  ğ‘ğ‘‘ ê°€ ëª¨ë“  ë°ì´í„°ì˜  ğ‘‘ ë²ˆì§¸ ì°¨ì›ì˜ ì›ì†Œë¡œ ì´ë£¨ì–´ì§„ ë°ì´í„° ë²¡í„°(íŠ¹ì§• í–‰ë ¬ì˜ ì—´ë²¡í„°)ë¼ê³  í•  ë•Œ ëª¨ë“  ì°¨ì›  ğ‘‘(ğ‘‘=0,â€¦,ğ·) ì— ëŒ€í•´  ğ‘ğ‘‘ ëŠ” ì”ì°¨ ë²¡í„°  ğ‘’ ì™€ ìˆ˜ì§ì„ ì´ë£¬ë‹¤.

$$c_d^T e = 0 \;\;\; (d=0, \ldots, D)$$

ë˜ëŠ”

$$c_d \perp e \;\;\; (d=0, \ldots, D)$$

ì§êµ ë°©ì •ì‹ìœ¼ë¡œë¶€í„° ë‹¤ìŒê³¼ ê°™ì€ ì„±ì§ˆì„ ì•Œ ìˆ˜ ìˆë‹¤.

(1) ëª¨í˜•ì— ìƒìˆ˜í•­ì´ ìˆëŠ” ê²½ìš°ì— ì”ì°¨ ë²¡í„°ì˜ ì›ì†Œì˜ í•©ì€ 0ì´ë‹¤. ì¦‰, ì”ì°¨ì˜ í‰ê· ì€ 0ì´ë‹¤.

$$\sum_{i=0}^N e_i = 0$$

(2)  ğ‘¥  ë°ì´í„°ì˜ í‰ê· ê°’  ğ‘¥Â¯ ì— ëŒ€í•œ ì˜ˆì¸¡ê°’ì€  ğ‘¦  ë°ì´í„°ì˜ í‰ê· ê°’  ğ‘¦Â¯ ì´ë‹¤.

$$\bar{y} = w^T \bar{x}$$

1ë²ˆ ì„±ì§ˆì€ ìƒìˆ˜í•­ ê²°í•©ì´ ë˜ì–´ ìˆìœ¼ë©´  ğ‘‹ ì˜ ì²«ë²ˆì§¸ ì—´ì´ 1-ë²¡í„°ë¼ëŠ” ê²ƒì„ ì´ìš©í•˜ì—¬ ì¦ëª…í•  ìˆ˜ ìˆë‹¤.

$$c_0^T e = \mathbf{1}^T e = \sum_{i=0}^N e_i = 0$$

2ë²ˆ ì„±ì§ˆì€ ë‹¤ìŒì²˜ëŸ¼ ì¦ëª…í•œë‹¤.

\begin{eqnarray}
\bar{y} 
&=& \dfrac{1}{N}\mathbf{1}^T y \\
&=& \dfrac{1}{N}\mathbf{1}^T (Xw + e) \\
&=& \dfrac{1}{N}\mathbf{1}^TXw + \dfrac{1}{N}\mathbf{1}^Te \\
&=& \dfrac{1}{N}\mathbf{1}^TXw \\
&=& \dfrac{1}{N}\mathbf{1}^T \begin{bmatrix}c_1 & \cdots & c_M \end{bmatrix} w \\
&=& \begin{bmatrix}\dfrac{1}{N}\mathbf{1}^Tc_1 & \cdots & \dfrac{1}{N}\mathbf{1}^Tc_D \end{bmatrix} w \\
&=& \begin{bmatrix}\bar{c}_1 & \cdots & \bar{c}_D \end{bmatrix} w \\
&=& \bar{x}^T w \\
\end{eqnarray}

## NumPyë¥¼ ì´ìš©í•œ ì„ í˜• íšŒê·€ë¶„ì„

ì´ì œ NumPyì˜ ì„ í˜•ëŒ€ìˆ˜ ê¸°ëŠ¥ì„ ì‚¬ìš©í•˜ì—¬ OLS ë°©ë²•ìœ¼ë¡œ ì„ í˜• íšŒê·€ë¶„ì„ì„ í•´ë³´ì. ìš°ì„  make_regression ëª…ë ¹ì„ ì‚¬ìš©í•˜ì—¬ ë‹¤ìŒê³¼ ê°™ì´ 1ì°¨ì› íŠ¹ì§• ë°ì´í„° xì™€ ì´ ê°’ì— ì˜ì¡´í•˜ëŠ” yë¥¼ ë§Œë“ ë‹¤.

```python
from sklearn.datasets import make_regression

bias = 100
X0, y, w = make_regression(n_samples=200, n_features=1, bias=bias, noise=10, coef=True, random_state=1)

X = sm.add_constant(X0)
y.shape
```

```python
y = y.reshape(len(y), 1)
y.shape
```

```python
y[:5]
```

ìš°ë¦¬ê°€ ì¤€ bias ê°’ì€ 100ì´ê³  make_regression ëª…ë ¹ì´ ìƒì„±í•œ ëª¨ìˆ˜ ê°’ì€ ë‹¤ìŒê³¼ ê°™ë‹¤.

```python
w
```

ë”°ë¼ì„œ xì™€ yëŠ” ë‹¤ìŒê³¼ ê°™ì€ ê´€ê³„ë¥¼ ê°€ì§„ë‹¤.

$$y = 100 + 86.44794301 x + \epsilon$$

ìœ„ì—ì„œ êµ¬í•œ ìˆ˜ì‹ì„ ì´ìš©í•˜ì—¬ ì„ í˜•íšŒê·€ ê³„ìˆ˜ë¥¼ ì¶”ì •í•˜ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

```python
# OLS í•´ë¥¼ ì§ì ‘ ì´ìš©í•˜ëŠ” ë°©ë²•
w = np.linalg.inv(X.T @ X) @ X.T @ y  # np.linalg.inv ì—­í–‰ë ¬ êµ¬í•˜ê¸°, ğ‘‹ğ‘‡ğ‘‹ğ‘¤=ğ‘‹ğ‘‡ğ‘¦
w
```

ë”°ë¼ì„œ xì™€ yëŠ” ë‹¤ìŒê³¼ ê°™ì€ ê´€ê³„ë¥¼ ê°€ì§„ë‹¤.

$$\hat{y} = 99.79150869 + 86.96171201 x$$

ì´ ê²°ê³¼ì—ì„œ ì•Œ ìˆ˜ ìˆëŠ” ê²ƒì€ ì„ í˜• íšŒê·€ë¥¼ í†µí•´ êµ¬í•œ ê²°ê³¼ëŠ” ì‹¤ì œ(ìì—° ë²•ì¹™)ì™€ ë¹„ìŠ·í•˜ì§€ë§Œ ì •í™•í•˜ì§€ëŠ” ì•Šë‹¤ëŠ” ì ì´ë‹¤.

```python
# lstsq ëª…ë ¹ì„ ì´ìš©í•˜ëŠ” ë°©ë²•
w = np.linalg.lstsq(X, y, rcond=None)[0]  
w
```

lstsq ëª…ë ¹ì€ í–‰ë ¬  ğ´ ì™€  ğ‘ ë¥¼ ëª¨ë‘ ì¸ìˆ˜ë¡œ ë°›ê³  ë’¤ì—ì„œ ì„¤ëª…í•  ìµœì†ŒììŠ¹ë¬¸ì œ(least square problem)ì˜ ë‹µ x, ì”ì°¨ì œê³±í•©(residual sum of squares) resid, ë­í¬(rank) rank, íŠ¹ì‡ê°’(singular value) së¥¼ ë°˜í™˜í•œë‹¤.  lstsq ëª…ë ¹ìœ¼ë¡œ êµ¬í•œ ë‹µì´ inv ëª…ë ¹ìœ¼ë¡œ êµ¬í•œ ë‹µê³¼ ê°™ìŒ

ì°¸ê³ : np.linalg.lstsq https://datascienceschool.net/view-notebook/927d7f7972dd434ead9d294169ae7f34/

ë‹¤ìŒ ì½”ë“œëŠ” ì—¬ëŸ¬ê°€ì§€ ìƒˆë¡œìš´  ğ‘¥ ê°’ì— ëŒ€í•´  ğ‘¦Ì‚  ì„ êµ¬í•´ë³¸ ê²°ê³¼ë¥¼ ì›ë˜ ë°ì´í„°ì™€ ë¹„êµí•œ ê²ƒì´ë‹¤.

```python
x_new = np.linspace(np.min(X0), np.max(X0), 100)
X_new = sm.add_constant(x_new)
X0[:5], X_new[:5]
```

```python
y_new = np.dot(X_new, w)

plt.scatter(X0, y, label="data")
plt.plot(x_new, y_new, 'r-', label="íšŒê·€ë¶„ì„ê²°ê³¼")
plt.xlabel("x")
plt.ylabel("y")
plt.title("ì„ í˜•íšŒê·€ë¶„ì„ì˜ ì˜ˆ")
plt.legend()
plt.show()
```
