## Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏

ÌôïÎ•†Í≥ºÏ†ïÏùò Í∏∞ÎåÄÍ∞íÏù¥ ÏãúÍ∞Ñ  ùë° Ïóê ÎåÄÌïú Ìï®ÏàòÎ°ú ÌëúÌòÑÎê† Ïàò ÏûàÏúºÎ©¥ Ïù¥Î•º Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏(trend)ÎùºÍ≥† ÌïúÎã§.

$$\mu_t = \text{E}[Y_t] = f(t)$$

Îã§Ïùå ÏãúÍ≥ÑÏó¥Îì§ÏùÄ Î™®Îëê Ï∂îÏÑ∏Î•º Í∞ÄÏßÄÎäî ÌôïÎ•† Í≥ºÏ†ïÏùÑ ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìïú Í≤ÉÏù¥Îã§.

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
import scipy as sp

np.random.seed(0)
t = np.arange(20)
y = np.zeros((30, 20)) 
for i in range(30):
    y[i, :] = t + sp.stats.norm.rvs(size=20)
    plt.plot(t, y[i], lw=1)
    
plt.title("Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏Î•º Í∞ñÎäî ÌôïÎ•†Í≥ºÏ†ï ÏãúÎÆ¨Î†àÏù¥ÏÖò")
plt.show()
```

ÏãúÍ∞ÑÏóê ÏùòÌï¥ Í≤∞Ï†ïÎêòÎäî Í≤∞Ï†ïÎ°†Ï†Å Í≥ÑÏ†àÏÑ±ÎèÑ Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏Ïóê Ìè¨Ìï®ÎêúÎã§.

```python
np.random.seed(0)
t = np.arange(20)
y = np.zeros((30, 20))

for i in range(30):
    y[i, :] = -5 * np.sin(0.25 * np.pi * t) + sp.stats.norm.rvs(size=20)
    plt.plot(t, y[i], lw=1)
plt.title("Í≤∞Ï†ïÎ°†Ï†Å Í≥ÑÏ†àÏÑ±ÏùÑ Í∞ÄÏßÄÎäî ÌôïÎ•†Í≥ºÏ†ï ÏãúÎÆ¨Î†àÏù¥ÏÖò")
plt.show()
```

## Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏ Ï∂îÏ†ï

#### Ï∂îÏÑ∏ Ï∂îÏ†ï(trend estimation)ÏùÄ ÌôïÎ•† Í≥ºÏ†ïÏùò Í≤∞Ï†ïÎ°†Ï†Å Í∏∞ÎåìÍ∞í Ìï®ÏàòÎ•º ÏïåÏïÑÎÇ¥Îäî Í≤ÉÏùÑ ÎßêÌïúÎã§. 

Î≥¥ÌÜµ Îã§ÏùåÍ≥º Í∞ôÏùÄ Í∞ÄÏ†ïÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÌôïÎ•† Í≥ºÏ†ïÏúºÎ°úÎ∂ÄÌÑ∞ Ï∂îÏÑ∏ ÏÑ±Î∂ÑÏùÑ Î∂ÑÎ¶¨ÌïúÎã§.

#### Ïö∞Î¶¨Í∞Ä Î∂ÑÏÑùÌïòÍ≥†ÏûêÌïòÎäî ÌôïÎ•† Í≥ºÏ†ï  ùëåùë° Í∞Ä ÏùºÎ∞òÏ†ÅÏù∏ ÎπÑÏ†ïÏÉÅ Í≥ºÏ†ïÏù¥ ÏïÑÎãàÎùº Ï∂îÏ†ïÏù¥ Í∞ÄÎä•Ìïú Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏ Ìï®Ïàò  ùëì(ùë°) ÏôÄ Ï†ïÏÉÅ ÌôïÎ•† Í≥ºÏ†ï  ùëãùë° Ïùò Ìï©ÏúºÎ°ú ÌëúÌòÑÎê† Ïàò ÏûàÎã§.

$$Y_t = f(t) + X_t$$

## Îã§Ìï≠Ïãù Ï∂îÏÑ∏

Îã§Ìï≠Ïãù Ï∂îÏÑ∏ Î∂ÑÏÑù Î∞©Î≤ïÏùÄ Ï∂îÏÑ∏ Ìï®Ïàò Ï¶â, ÌôïÎ•† Í≥ºÏ†ïÏùò Í∏∞ÎåìÍ∞íÏùÑ ÏãúÍ∞ÑÏóê ÎåÄÌïú Îã§Ìï≠ÏãùÏúºÎ°ú ÎÇòÌÉÄÎÇº Ïàò ÏûàÎã§Í≥† Í∞ÄÏ†ïÌïòÎäî Í≤ÉÏù¥Îã§.

$$f(t) = \sum_{i=0}^M a_i t^i = a_0 + a_1 t + a_2 t^2 + \cdots$$

Í∞ÄÏû• Îã®ÏàúÌïú Î™®ÌòïÏúºÎ°ú ÏÑ†Ìòï Ï∂îÏÑ∏Î•º Í∞ÄÏßÄÎäî Í≤ΩÏö∞Î•º ÏÇ¥Ìé¥Î≥¥Ïûê. Îã§Ïùå ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Îäî ÎåÄÍ∏∞Ï§ë CO2 ÎÜçÎèÑÎ•º Ï∏°Ï†ïÌïú Í≤ÉÏù¥Îã§.

```python
import statsmodels.api as sm

data = sm.datasets.get_rdataset("CO2", package="datasets")
df = data.data
df.tail()
```

```python
def yearfraction2datetime(yearfraction, startyear=0):
    import datetime
    import dateutil
    year = int(yearfraction) + startyear
    month = int(round(12 * (yearfraction - year)))
    delta = dateutil.relativedelta.relativedelta(months=month)
    date = datetime.datetime(year, 1, 1) + delta
    return date
```

```python
import pandas as pd

df["datetime"] = df.time.map(yearfraction2datetime)
df['month'] = df.datetime.dt.month
df.tail()
```

```python
result = sm.OLS.from_formula("value ~ time", data=df).fit()
print(result.summary())
```

ÏúÑ ÌöåÍ∑Ä Î∂ÑÏÑù Í≤∞Í≥º Î≥¥Í≥†ÏÑúÏóêÏÑú Ï∂îÏÑ∏ Ìï®ÏàòÍ∞Ä Îã§ÏùåÍ≥º Í∞ôÎã§Îäî Í≤ÉÏùÑ Ìï† Ïàò ÏûàÎã§.

$$f(t) = 1.3075 t - 2249.7742$$

```python
t = df.time
y = df.value
trend = result.params[0] + result.params[1] * t  # Ï∂îÏÑ∏Ìï®Ïàò
plt.plot(t, y, '-')
plt.plot(t, trend, '-', label="Ï∂îÏÑ∏Ìï®Ïàò")
plt.title("CO2 ÎÜçÎèÑ ÏãúÍ≥ÑÏó¥Í≥º Í∑∏Ïóê ÎåÄÌïú Ï∂îÏÑ∏Ìï®Ïàò")
plt.xlabel("ÏãúÍ∞Ñ")
plt.ylabel("CO2 ÎÜçÎèÑ")
plt.legend()
plt.show()
```

ÎßåÏïΩ Ï∂îÏÑ∏Í∞Ä 2Ï∞® Ìï®Ïàò ÌòïÌÉúÎùºÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï∂îÏÑ∏Î•º Ï∂îÏ†ïÌï† Ïàò ÏûàÎã§.

```python
result2 = sm.OLS.from_formula("value ~ time + I(time ** 2)", data=df).fit()
print(result2.summary())
```

```python
trend2 = result2.params[0] + result2.params[1] * t + result2.params[2] * t**2
plt.plot(t, y, '-', t, trend2, '-')
plt.title("CO2 ÎÜçÎèÑ ÏãúÍ≥ÑÏó¥Í≥º Í∑∏Ïóê ÎåÄÌïú Ï∂îÏÑ∏Ìï®Ïàò")
plt.show()
```

Ïù¥Î†áÍ≤å Íµ¨Ìïú Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏ Î™®ÌòïÏùÑ Ïù¥Ïö©ÌïòÎ©¥ 1998ÎÖÑ 1ÏõîÏùò CO2 ÎÜçÎèÑÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏòàÏ∏°Ìï† Ïàò ÏûàÎã§.

```python
t_test = 1998 + 1 / 12
X_test = pd.DataFrame([[t_test]], columns=["time"])
X_test
```

```python
result2.predict(X_test)
```

```python
trend2 = result2.params[0] + result2.params[1] * t + result2.params[2] * t ** 2
plt.plot(t[-30:], y[-30:], 'o-', t[-30:], trend2[-30:], '-')
plt.plot(t_test, result2.predict(X_test).values[0], 'o', ms=10, lw=5)
plt.title("Co2 ÎÜçÎèÑ ÏãúÍ≥ÑÏó¥Í≥º 1998ÎÖÑ 1Ïõî ÏòàÏ∏°Ïπò")
plt.show()
```

## Í≥ÑÏ†àÏÑ± Ï∂îÏÑ∏

Í≥ÑÏ†àÏÑ± Ï∂îÏÑ∏Îäî ÌäπÏ†ïÌïú Îã¨(month)Ïù¥ÎÇò ÏöîÏùº(day of week)Ïóê Îî∞Îùº Í∏∞ÎåìÍ∞íÏù¥ Îã¨ÎùºÏßÄÎäî Í≤ÉÏùÑ ÎßêÌïúÎã§. Ïù¥Îäî Îã¨ Ïù¥Î¶ÑÏù¥ÎÇò ÏöîÏùº Ïù¥Î¶ÑÏùÑ Ïπ¥ÌÖåÍ≥†Î¶¨(category) Í∞íÏúºÎ°ú ÏÇ¨Ïö©ÌïòÏó¨ ÌöåÍ∑ÄÎ∂ÑÏÑùÌïòÏó¨ Ï∂îÏ†ïÌï† Ïàò ÏûàÎã§.

Îã§ÏùåÏùÄ Ìò∏Ìù°Í∏∞ÏßàÌôò ÏÇ¨ÎßùÏûêÏàòÎ•º ÎÇòÌÉÄÎÇ∏ ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Ïù¥Îã§.

```python
df2 = sm.datasets.get_rdataset("deaths", "MASS").data
df2["datetime"] = df2.time.map(yearfraction2datetime)
df2.tail()
```

```python
df2["month"] = df.datetime.dt.month
df2.tail()
```

```python
result = sm.OLS.from_formula("value ~ C(month) - 1", data=df2).fit()  # -1 : ÏÉÅÏàòÌï≠ Ï†úÏô∏
print(result.summary())
```

Î™®ÌòïÏãù Î¨∏ÏûêÏó¥ÏóêÏÑú "-1"ÏùÄ y Ï†àÌé∏ÏùÑ Î™®ÌòïÏóê ÎÑ£ÏßÄ ÏïäÎäî Í≤ÉÏùÑ ÏùòÎØ∏ÌïúÎã§. Ï¶â, Î™®Îì† monthÎ≥ÄÏàòÏùò Í∞ÄÎä•Ìïú Í∞íÏùÑ ÏÇ¨Ïö©ÌïúÎã§. 

#### month Î≥ÄÏàòÍ∞Ä 0Í≥º 1Í∞íÏùÑ Í∞ÄÏßÄÎäî dummy Î≥ÄÏàòÏù¥Í∏∞ ÎïåÎ¨∏Ïóê Ïù¥ Î∂ÑÏÑùÏóêÏÑú Íµ¨Ìïú Í≥ÑÏàòÎäî ÌäπÏ†ïÌïú Îã¨Ïùò Í∏∞Ïò® ÌèâÍ∑†Í∞íÏù¥ ÎêúÎã§.

Ï∂îÏ†ïÌïú Í≥ÑÏ†àÏÑ± ÏÑ±Î∂ÑÍ≥º ÎÇòÎ®∏ÏßÄ ÏÑ±Î∂ÑÏùÑ Î∂ÑÎ¶¨ÌïòÏó¨ Í∑∏Î¶¨Î©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
plt.plot(df2.value, label="ÏÇ¨ÎßùÏûê Ïàò ÏãúÍ≥ÑÏó¥")
plt.plot(result.fittedvalues, lw=3, alpha=0.5, label="Ï∂îÏ†ïÌïú ÏãúÍ≥ÑÏó¥")
plt.plot(result.resid, label="ÏûîÏ∞®")
plt.title('ÏãúÍ≥ÑÏó¥Ïùò Í≥ÑÏ†àÏÑ± Ï∂îÏ†ï')
plt.legend(loc=1)
plt.show()
```

ÏúÑ Î∞©Î≤ïÏúºÎ°ú Ï∞æÏïÑÎÇ∏ ÏûîÏ∞® ÏãúÍ≥ÑÏó¥ÏùÑ Î≥¥Î©¥ ÏãúÍ∞ÑÏù¥ ÏßÄÎÇòÍ∞à ÏàòÎ°ù Ï†êÏ†ê Í∞êÏÜåÌïòÍ≥† ÏûàÎäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§. Ïù¥Îü¨Ìïú ÏÑ†Ìòï Ï∂îÏÑ∏ÍπåÏßÄ ÌïúÍ∫ºÎ≤àÏóê Ïû°ÏïÑÎÇ¥Î†§Î©¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÌöåÍ∑ÄÎ∂ÑÏÑùÏùÑ ÌïòÎ©¥ ÎêúÎã§.

```python
result2 = sm.OLS.from_formula("value ~ time + C(month) - 1", data=df2).fit()  # time Ï∂îÍ∞Ä
print(result2.summary())
```

```python
plt.plot(df2.value, label="ÏÇ¨ÎßùÏûêÏàò ÏãúÍ≥ÑÏó¥")
plt.plot(result2.fittedvalues, lw=3, alpha=0.5, label="Ï∂îÏ†ïÌïú ÏãúÍ≥ÑÏó¥")
plt.plot(result2.resid, label="ÏûîÏ∞®")
plt.title("ÏãúÍ≥ÑÏó¥Ïùò Í≥ÑÏ†àÏÑ±Í≥º Ï∂îÏÑ∏ Ï∂îÏ†ï")
plt.legend()
plt.show()
```

Ïù¥Î†áÍ≤å Íµ¨Ìïú Í≤∞Ï†ïÎ°†Ï†Å Ï∂îÏÑ∏ Î™®ÌòïÏùÑ Ïù¥Ïö©ÌïòÎ©¥ 1980ÎÖÑ 1ÏõîÏùò ÏÇ¨ÎßùÏûê ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏòàÏ∏°Ìï† Ïàò ÏûàÎã§.

```python
t_test = 1980 + 1 / 12
month_test = 1
X_test = pd.DataFrame([[t_test, month_test]], columns=["time", "month"])
X_test
```

```python
result2.predict(X_test)
```

```python
plt.plot(df2.value, label="ÏÇ¨ÎßùÏûêÏàò ÏãúÍ≥ÑÏó¥")
plt.plot(result2.fittedvalues, lw=3, alpha=0.5, label="Ï∂îÏ†ïÌïú ÏãúÍ≥ÑÏó¥")
plt.plot(len(df2) + 1, result2.predict(X_test).values[0], 'o', ms=10, lw=5)
plt.title("ÏÇ¨ÎßùÏûê Ïàò ÏãúÍ≥ÑÏó¥Í≥º 1980ÎÖÑ 1Ïõî ÏòàÏ∏°Ïπò")
plt.show()
```

Ïù¥Î≤àÏóêÎäî ÏúÑÏóêÏÑú Î≥¥ÏïòÎçò CO2 ÎÜçÎèÑÎ•º Í≥ÑÏ†àÏÑ±ÍπåÏßÄ Ìè¨Ìï®ÌïòÏó¨ Ï∂îÏÑ∏ Ï∂îÏ†ïÏùÑ Ìï¥Î≥¥Ïûê.

```python
result3 = sm.OLS.from_formula('value ~ 0 + C(month) + time + I(time ** 2)', data=df).fit()  # 0 : ÏÉÅÏàòÌï≠ Ìè¨Ìï® ÏïàÌï®
print(result3.summary())
```

```python
plt.subplot(211)
plt.title("CO2 ÎÜçÎèÑ ÏãúÍ≥ÑÏó¥ Ï∂îÏ†ï")
plt.plot(df.value, label="CO2 ÎÜçÎèÑ ÏãúÍ≥ÑÏó¥")
plt.plot(result3.fittedvalues, lw=2, alpha=0.5, label="Ï∂îÏ†ïÌïú ÏãúÍ≥ÑÏó¥")

plt.subplot(212)
plt.title("ÏûîÏ∞® ÏãúÍ≥ÑÏó¥")
plt.plot(df.value - result3.fittedvalues)

plt.tight_layout()
plt.show()
```

```python
data = sm.datasets.get_rdataset("PeaceBridge2012", package="Stat2Data")
df = data.data
df.tail()
```

```python
df.plot(x="t", y="Traffic", style=["bo-"])
plt.show()
```

```python
result = sm.OLS.from_formula("Traffic ~ C(Month) + t - 1", data=df).fit()  # ÏÉÅÏàòÌï≠ Ìè¨Ìï®ÌïòÏßÄ ÏïäÏùå
print(result.summary())
```

```python
plt.plot(df.Traffic, label="ÏõîÍ∞Ñ ÍµêÌÜµÎüâ ÏãúÍ≥ÑÏó¥")
plt.plot(result.fittedvalues, label="Ï∂îÏ†ïÌïú ÏãúÍ≥ÑÏó¥")
plt.plot(result.resid, label="ÏûîÏ∞®")
plt.title("2012-2015ÎÖÑ ÏÇ¨Ïù¥Ïùò ÏõîÍ∞Ñ ÍµêÌÜµÎüâ ÏãúÍ≥ÑÏó¥Í≥º Í∑∏Ïóê ÎåÄÌïú Í≥ÑÏ†àÏÑ± Ï∂îÏ†ï")
plt.legend()
plt.show()
```

```python
t_test = len(df) + 1
month_test = 1

X_test = pd.DataFrame([[month_test, t_test]], columns=["Month", "t"])
X_test
```

```python
result.predict(X_test)
```

```python
plt.plot(df.Traffic, label="ÏõîÍ∞Ñ ÍµêÌÜµÎüâ ÏãúÍ≥ÑÏó¥")
plt.plot(result.fittedvalues, label="Ï∂îÏ†ïÌïú ÏãúÍ≥ÑÏó¥")
plt.plot(len(df)+1, result.predict(X_test).values[0], 'o', ms=10, lw=5)
plt.title("ÏõîÍ∞Ñ ÍµêÌÜµÎüâ ÏãúÍ≥ÑÏó¥Í≥º 2016ÎÖÑ 1Ïõî ÏòàÏ∏°Ïπò")
plt.show()
```
