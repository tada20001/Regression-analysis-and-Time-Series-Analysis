Îã§ÏùåÍ≥º Í∞ôÏùÄ ARMA(p,q) Î™®ÌòïÍ≥º ÌëúÎ≥∏ ÏãúÍ≥ÑÏó¥ ÏûêÎ£åÍ∞Ä ÏûàÎäî Í≤ΩÏö∞ 
#### ÏµúÎåÄ Í∞ÄÎä•ÎèÑ Ï∂îÏ†ïÎ≤ïÏúºÎ°ú Î™®Ïàò  ùúôùëñ ,  ùúÉùëñ Î•º Íµ¨Ìï† Ïàò ÏûàÎã§.

$$Y_t + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} 
= \epsilon_t + \theta_1 \epsilon_{t-1}  + \theta_2 \epsilon_{t-2} - \cdots  + \theta_q \epsilon_{t-q}$$

MLEÎ•º ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌïú Í∞ÄÎä•ÎèÑ Ìï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

\begin{eqnarray}
\mathcal{L}(\theta; \{y_j\}) 
&=& \mathcal{L}(\phi_1, \cdots, \phi_p, \theta_1, \cdots, \theta_q; y_1, y_2, \cdots, y_N) \\
&=& f_{Y_1,\cdots , Y_N}(y_1, y_2, \cdots, y_N;\phi_1, \cdots, \phi_p, \theta_1, \cdots, \theta_q) \\
\end{eqnarray}

## ARMA(p,q)Ïùò Í∞ÄÎä•ÎèÑ Ìï®Ïàò

ÌôïÎ•†Î≥ÄÏàòÏôÄ Îã¨Î¶¨ ÌôïÎ•†Í≥ºÏ†ï Î™®ÌòïÏóêÏÑúÎäî  ùúñ  Í∞íÏùÑ ÏïåÏïÑÏïº ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê Îã§ÏùåÍ≥º Í∞ôÏùÄ Î∞òÎ≥µÏ†Å Î∞©Î≤ïÏùÑ ÏÇ¨Ïö©ÌïúÎã§.

#### (1) ùúñ  Í∞íÏùò Ï¥àÍ∏∞Í∞í Í∞ÄÏ†ïÍ≥º ÏÑ†ÌòïÌöåÍ∑ÄÎ∞©Î≤ïÏùÑ ÌÜµÌïú Î™®Ïàò Ï∂îÏ†ï

Ïö∞ÏÑ† ÏûÑÏùòÏùò  ùúñ  Í∞íÏùÑ Ï¥àÍ∏∞Í∞íÏúºÎ°ú Í∞ÄÏ†ïÌïúÎã§.

ùëåùë° Îäî  ùëåùë°‚àí1,‚ãØ,ùëåùë°‚àíùëù,ùúñùë°‚àí1,‚ãØ,ùúñùë°‚àíùëû  Ïóê ÏùòÏ°¥ÌïòÎäî Ï†ïÍ∑ú Î∂ÑÌè¨ ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÎêúÎã§.

\begin{eqnarray}
Y_{t} 
&\sim& 
-\phi_1 Y_{t-1} -\phi_2 Y_{t-2} -\cdots -\phi_p Y_{t-p} 
+\theta_1 \epsilon_{t-1} +\theta_2 \epsilon_{t-2} +\cdots +\theta_q \epsilon_{t-q}
\end{eqnarray}

ÏÑ†ÌòïÌöåÍ∑ÄÎ∞©Î≤ïÏúºÎ°ú Î™®Ïàò  ùúô1,‚ãØ,ùúôùëù,ùúÉ1,‚ãØ,ùúÉùëû Î•º Ï∂îÏ†ïÌïúÎã§.

#### (2) Îã§Ïãú ùúñ  Í∞íÏùÑ Í≥ÑÏÇ∞ÌïòÍ≥† ÌöåÍ∑ÄÎ∂ÑÏÑùÏùÑ Ïã§ÏãúÌïòÏó¨ Î™®Ïàò Ï∂îÏ†ï 


ùúñ  Í∞íÏùÑ Í≥ÑÏÇ∞ÌïúÎã§. Ïö∞ÏÑ† Îã§ÏùåÍ≥º Í∞ôÏùÄ Ï¥àÍ∏∞Í∞í Í∞ÄÏ†ïÏùÑ ÌïòÏûê.

$$\epsilon_p = \epsilon_{p-1} = \cdots = \epsilon_{p-q} = 0$$

Ïù¥  ùúñùëù+1,‚ãØ,ùúñùëÅ ÏùÄ Îã§Ïùå Í≥µÏãùÏóêÏÑú Íµ¨Ìï† Ïàò ÏûàÎã§.

$$\epsilon_{p+1}=Y_{p+1}+\phi_1Y_{p}+\phi_2 Y_{p-1} + \cdots + \phi_p Y_{1}$$

$$\epsilon_{p+2}=Y_{p+2}+\phi_1Y_{p+1}+\phi_2Y_{p}+\cdots+\phi_pY_{2}-\theta_1\epsilon_{p+1}$$

$$\vdots$$

Ïù¥  ùúñ  Í∞íÏùÑ Ïù¥Ïö©ÌïòÏó¨ Îã§Ïãú (1)Ïùò ÌöåÍ∑ÄÎ∂ÑÏÑùÏùÑ Ïã§ÏãúÌïòÏó¨ Î™®Ïàò  ùúô1,‚ãØ,ùúôùëù,ùúÉ1,‚ãØ,ùúÉùëû Î•º Ï∂îÏ†ïÌïúÎã§.

## statsmodels Ìå®ÌÇ§ÏßÄÎ•º ÏÇ¨Ïö©Ìïú ARMA Î™®Ïàò Ï∂îÏ†ï

statsmodels Ìå®ÌÇ§ÏßÄÎäî ARMA Î™®Ïàò Ï∂îÏ†ïÏùÑ ÏúÑÌïú ARMAÎùºÎäî ÌÅ¥ÎûòÏä§ÏôÄ Ïù¥ ÌÅ¥ÎûòÏä§Ïùò fit Î©îÏÑúÎìúÎ•º Ï†úÍ≥µÌïúÎã§. ARMA ÌÅ¥ÎûòÏä§Ïùò Ïù∏ÏàòÎ°ú ÏãúÍ≥ÑÏó¥ ÌëúÎ≥∏Í≥º Ï∞®ÏàòÎ•º ÎÑ£Ïñ¥ Ïù∏Ïä§ÌÑ¥Ïä§Î•º ÏÉùÏÑ±Ìïú Îí§ fit Î©îÏÑúÎìúÎ•º Ìò∏Ï∂úÌïòÎ©¥ Ï∂îÏ†ï Í≤∞Í≥ºÍ∞íÏùÑ Í∞ÄÏßÑ ARMAResults ÌÅ¥ÎûòÏä§ ÏûêÎ£åÎ•º Î∞òÌôòÌïúÎã§. 

#### Îã§Îßå Ï£ºÏùòÌï† Ï†êÏùÄ ARMAResultsÏóêÏÑúÎäî ARÍ≥ÑÏàò  ùúô Ïóê ÎåÄÌï¥ ArmaProcess ÌÅ¥ÎûòÏä§ÏóêÏÑú Ï†ïÌïú Î∂ÄÌò∏ÏôÄ Î∞òÎåÄÏùò Î∂ÄÌò∏Î•º Ï∂úÎ†•ÌïúÎã§.

ARMA ÌÅ¥ÎûòÏä§ÏôÄ Î™áÍ∞ÄÏßÄ Ïù¥Î°†Ï†Å Î™®ÌòïÏóêÏÑú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïòÏó¨ ÎÇòÏò® ÏãúÍ≥ÑÏó¥ ÏûêÎ£åÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Î™®ÏàòÎ•º Ï∏°Ï†ïÌïòÏó¨ Î≥¥Ïûê.

## MA(1) Î™®Ïàò Ï∂îÏ†ïÏùò Ïòà

$$Y_t = \epsilon_t + 0.9 \epsilon_{t-1}$$

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)

import statsmodels.api as sm
```

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1, -0.9], [1]) # MA Î™®Ìòï ÏÉùÏÑ±
y = p.generate_sample(1000)  # ÏÉòÌîå ÏÉùÏÑ±
m = sm.tsa.ARMA(y, (1, 0))   # Ï∂îÏ†ïÍ∞í ÏÉùÏÑ±ÏùÑ ÏúÑÌï¥ ÏãúÍ≥ÑÏó¥ ÌëúÎ≥∏Í≥º Ï∞®ÏàòÎ•º ÎÑ£ÏùÄ ÌõÑ Ïù∏Ïä§ÌÑ¥Ïä§ ÏÉùÏÑ±
r = m.fit()  # Ï∂îÏ†ïÍ≤∞Í≥ºÍ∞í
print(r.summary())
```

ar.L1.y Í∞íÏù¥ 0.9144Î°ú Î™®ÌòïÏùò 0.9ÏôÄ Í±∞Ïùò ÏùºÏπòÌïú Í∞íÏù¥ Ï∂îÏ†ïÎêòÍ≥† ÏûàÎã§.

## AR(2) Î™®Ïàò Ï∂îÏ†ïÏùò Ïòà

$$Y_t = 1.5 Y_{t-1} - 0.75 Y_{t-2} + \epsilon_t$$

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1, -1.5, 0.75], [1])
y = p.generate_sample(1000)
m = sm.tsa.ARMA(y, (2, 0))
r = m.fit()
print(r.summary())
```

ar.L1.y 1.49, ar.L2.y -0.74Î°ú ÏõêÎûò Î™®ÌòïÏùò Î™®ÏàòÍ∞íÍ≥º Í±∞Ïùò ÏùºÏπòÌïúÎã§. ÏÉÅÏàòÌï≠ÏùÄ 14%Î°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§ Ï±ÑÌÉùÏúºÎ°ú 0ÎùºÍ≥† Ìï† Ïàò ÏûàÎã§.

## ARMA(1,1) Î™®Ïàò Ï∂îÏ†ïÏùò Ïòà

$$Y_t = 0.6 Y_{t-1} + \epsilon_t + 0.3 \epsilon_{t-1}$$

```python
np.random.seed(0)
p = sm.tsa.ArmaProcess([1, -0.6], [1, 0.3])
y = p.generate_sample(1000)
m = sm.tsa.ARMA(y, (1, 1))
r = m.fit()
print(r.summary())
```

ar.L1.y 0.62, ma.L1.y 0.26ÏúºÎ°ú ÏõêÎûò Î™®ÌòïÏùò Î™®Ïàò 0.6, 0.3Í≥º Ïú†ÏÇ¨ÌïòÍ≤å Ï∂îÏ†ïÎêòÍ≥† ÏûàÎã§.

## Ïã§Ï†ú ÏãúÍ≥ÑÏó¥ ÏûêÎ£åÏóê ÎåÄÌïú ARMA Î™®Ìòï Í≥ÑÏàò Ï∂îÏ†ï

Ïù¥Î≤àÏóêÎäî Ïã§Ï†ú ÏãúÍ≥ÑÏó¥ ÏûêÎ£åÏóê ÎåÄÌï¥ ARMA Î™®Ìòï Í≥ÑÏàòÎ•º Ï∂îÏ†ïÌï¥ Î≥¥Ïûê.

### ÏÇ¨Î°Ä1. Ìô©Ï≤¥ÌòïÏÑ± Ìò∏Î•¥Î™¨ Î™®Ìòï Í≥ÑÏàò Ï∂îÏ†ï

Ìô©Ï≤¥ÌòïÏÑ± Ìò∏Î•¥Î™¨Ïùò Î™®Ìòï Ï∞®ÏàòÎäî AR(1)Î°ú Ï∂îÏ†ïÌñàÏóàÎã§. Ïù¥ Îïå Í≥ÑÏàòÎäî Îã§ÏùåÏ≤òÎüº Íµ¨ÌïúÎã§.

```python
data = sm.datasets.get_rdataset("lh")
df = data.data
df.plot(x="time", y="value", style=["o-"])
plt.show()
```

Ïù¥ ÏãúÍ≥ÑÏó¥ÏùÑ AR(1) Î™®ÌòïÏúºÎ°ú Î≥¥Í≥† Î™®ÏàòÎ•º Ï∂îÏ†ïÌïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
m = sm.tsa.ARMA(df.value, (1, 0))
r = m.fit()
print(r.summary())
```

Ïù¥ Î™®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Îã§Ïãú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§. ÍµµÏùÄ Ïã§ÏÑ†Ïù¥ ÏõêÎûò ÏûêÎ£åÏù¥Îã§.

```python
plt.figure(figsize=(10, 10))
#plt.plot(r.fittedvalues, label="Ï∂îÏ†ïÌïú Îç∞Ïù¥ÌÑ∞")
np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], [1])
plt.subplot(6, 1, 1)
plt.plot(df.value, lw=5)  # ÏõêÎûò Îç∞Ïù¥ÌÑ∞

for i in range(5):
    plt.subplot(6, 1, i+2)
    y = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0] # Í∞Å ÏÉòÌîå Í∞íÏóê Ìé∏Ï∞®Î•º Í≥±Ìïú ÌõÑ ÏÉÅÏàòÌï≠ ÎçîÌï®
    plt.plot(y)
    
plt.tight_layout()
plt.show()
```

MA(1) Î™®ÌòïÏúºÎ°ú Ï∂îÏ†ïÌïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§. AR(1) Î™®ÌòïÎ≥¥Îã§ ÏÑ±Îä•Ïù¥ Ï¢ãÏßÄ ÏïäÏùåÏùÑ Ïïå Ïàò ÏûàÎã§.  (Loglikelihood : Î°úÍ∑∏ Í∞ÄÎä•ÎèÑ, AIC, BIC ÎπÑÍµê)

```python
m = sm.tsa.ARMA(df.value, (0, 1))
r = m.fit()
print(r.summary())
```

ARMA(1,1)Î™®ÌòïÏúºÎ°ú Ï∂îÏ†ïÌïòÎ©¥ Í∞ÄÎä•ÎèÑÎäî ÎÜíÏßÄÎßå AIC, BICÍ∞Ä ÎÇòÏÅòÎã§.

```python
m = sm.tsa.ARMA(df.value, (1, 1))
r = m.fit()
print(r.summary())
```

### ÏÇ¨Î°Ä2. Ìò∏Ìù°Í∏∞ÏßàÌôò ÏÇ¨ÎßùÏûêÏàò

```python
data = sm.datasets.get_rdataset("deaths", "MASS")
df = data.data
df.tail()
```

```python
df.plot(x="time", y="value")
plt.show()
```

```python
# Ï∂îÏÑ∏ Ï†úÍ±∞Î•º ÏúÑÌï¥ Î°úÍ∑∏Î≥ÄÌôò
df["logvalue"] = np.log(df.value)
df.plot(x="time", y="logvalue")
plt.show()
```

Ïù¥ ÏãúÍ≥ÑÏó¥ÏùÑ AR(2) Î™®ÌòïÏúºÎ°ú Î≥¥Í≥† Î™®ÏàòÎ•º Ï∂îÏ†ïÌïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
m = sm.tsa.ARMA(df.logvalue, (2, 0))
r = m.fit()
print(r.summary())
```

Ïù¥ Î™®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Îã§Ïãú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§. ÍµµÏùÄ Ïã§ÏÑ†Ïù¥ ÏõêÎûò ÏûêÎ£åÏù¥Îã§.

```python
plt.figure(figsize=(10, 10))
np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], [1])
plt.subplot(6, 1, 1)
plt.plot(df.logvalue, lw=5)

for i in range(5):
    plt.subplot(6, 1, i+2)
    y = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0]
    plt.plot(y)
    
plt.tight_layout()
plt.show()
```

ARMA(1,0), ARMA(1,1), ARMA(2,0), ARMA(2,1), ARMA(1,2), ARMA(2,2) Î™®ÌòïÏùÑ ÏÇ¨Ïö©Ìïú Í≤∞Í≥ºÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
from itertools import product

result = []
for p, q in product(range(3), range(3)):
    if (p == 0): 
        continue 
    m = sm.tsa.ARMA(df.logvalue, (p, q))
    try:
        r = m.fit()
        result.append({"p": p, "q": q, "LLF": r.llf, "AIC": r.aic, "BIC": r.bic})
    except:
        pass
```

```python
import pandas as pd

pd.DataFrame(result)
```

ÏúÑÏùò Í≤∞Í≥ºÎ•º ÌÜµÌï¥ Î≥¥Î©¥, log likelihood Í∏∞Ï§ÄÏúºÎ°ú Î≥ºÎïå ARMA(2, 1) Î™®ÌòïÏù¥ Í∞ÄÏû• ÏÑ±Îä•Ïù¥ Ï¢ãÎã§Í≥† Ìï† Ïàò ÏûàÎã§. Îî∞ÎùºÏÑú Îã§Ïãú Ïù¥ Î™®ÌòïÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Î™®ÏàòÎ•º Ï∂îÏ†ïÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
m = sm.tsa.ARMA(df.logvalue.astype(float), (2, 1))
r = m.fit()
print(r.summary())
```

ÏúÑÏùò Î™®ÏàòÏ∂îÏ†ï Í≤∞Í≥ºÎ°ú Îã§Ïãú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
plt.figure(figsize=(10, 10))

np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], [1, r.maparams])
plt.subplot(6, 1, 1)
plt.plot(df.logvalue, lw=5)
for i in range(5):
    plt.subplot(6, 1, i+2)
    y = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0]
    plt.plot(y)
    
plt.tight_layout()
plt.show()
```

### ÏÇ¨Î°Ä3. Ìò∏Ìù°Í∏∞ ÏßàÌôò ÏÇ¨ÎßùÏûê ÏàòÏùò Í≤∞Ï†ïÎ°†Ï†Å Seasonality Î™®Ìòï

```python
def yearfraction2datetime(yearfraction, startyear=0):
    import datetime
    import dateutil
    year = int(yearfraction) + startyear
    month = int(round(12 * (yearfraction - year)))
    delta = dateutil.relativedelta.relativedelta(months=month)
    date = datetime.datetime(year, 1, 1) + delta
    return date

df["datetime"] = df.time.map(yearfraction2datetime)
df["month"] = df.datetime.dt.month

df.tail()
```

ÏÑ†ÌòïÌöåÍ∑ÄÎ∂ÑÏÑùÏùÑ ÌÜµÌï¥ Í≥ÑÏ†àÏÑ±ÏùÑ Ï∂îÏ†ïÌïúÎã§.

```python
result = sm.OLS.from_formula('logvalue ~ C(month) + time + 0', data=df).fit()
y_seasonal = result.fittedvalues # Í≥ÑÏ†àÏÑ± Ï∂îÏ†ïÌïú Í∞í
y_nonseasonal = df.logvalue - y_seasonal # Í≥ÑÏ†àÏÑ± Ï†úÍ±∞
```

```python
from itertools import product
```

```python
result = []
for p, q in product(range(3), range(3)):
    if (p==0):
        continue
    m = sm.tsa.ARMA(y_nonseasonal, (p, q))
    try:
        r = m.fit()
        result.append({"p": p, "q": q, "LLF": r.llf, "AIC": r.aic, "BIC": r.bic})
    except:
        pass
    
pd.DataFrame(result)
```

#### Í≥ÑÏ†àÏÑ± Ï∂îÏÑ∏Î•º Ï†úÍ±∞Ìïú ÌõÑÏóê Îã§Ïãú Î™®Ïàò Ï∂îÏ†ï Í≤∞Í≥º ARMA(2, 2)Î™®ÌòïÏù¥ Í∞ÄÏû• ÏÑ±Îä•Ïù¥ Ï¢ãÎã§. ARMA(2, 2) Î™®ÌòïÏúºÎ°ú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

```python
m = sm.tsa.ARMA(y_nonseasonal, (2, 2))
r = m.fit()
```

```python
plt.figure(figsize=(10, 10))
np.random.seed(0)
p = sm.tsa.ArmaProcess(np.r_[1, -r.arparams], np.r_[1, r.maparams],)
plt.subplot(6, 1, 1)
plt.plot(df.logvalue, lw=5)
for i in range(5):
    plt.subplot(6, 1, i + 2)
    y2 = p.generate_sample(len(df), burnin=100) * np.sqrt(r.sigma2) + r.params[0] 
    y = y_seasonal + y2
    plt.plot(y)
plt.tight_layout()
plt.show()
```
