Î°úÏª¨ Î†àÎ≤® Î™®Ìòï(local level model)ÏùÄ ÎûúÎç§ÏõåÌÅ¨ Î™®ÌòïÏóê Í¥ÄÏ∏° Ïû°ÏùåÏù¥ Ï∂îÍ∞ÄÎêú Í≤ÉÏù¥Îã§. Îã§ÏùåÍ≥º Í∞ôÏù¥ ÎûúÎç§ ÏõåÌÅ¨ Í≥ºÏ†ï Î™®ÌòïÏùÑ Îî∞Î•¥Îäî Îã®Î≥ÄÏàò ÏÉÅÌÉú Î≥ÄÏàò  ùúáùë° Î•º Í∞ÄÏßÑÎã§.

\begin{eqnarray}
\mu_t &=& \mu_{t-1} + w_t &,\;\;\; & w_t \sim N(0, \sigma_w^2) \\
Y_{t} &=& \mu_{t} + v_t &,\;\;\; & v_t \sim {N}(0, \sigma_v^2) \\
\end{eqnarray}

[Ï∞∏Í≥†] Ïù¥ÏÇ∞ ÏãúÍ∞Ñ ÎûúÎç§ ÏõåÌÅ¨(discrete-time random walk)Îäî Î∞±ÏÉâ Ïû°Ïùå(white noise)ÏùÑ ÎàÑÏ†ÅÌïú ÌôïÎ•† Í≥ºÏ†ïÏùÑ ÎßêÌïúÎã§. 
+ https://datascienceschool.net/view-notebook/6b963e771dc54f8c8cb23437274a86d6/

## ARIMA Î™®ÌòïÍ≥ºÏùò Í¥ÄÍ≥Ñ

Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏù¥ ARIMA Î™®Ìòï ÌòïÌÉúÎ°ú Î≥ÄÌôòÌï† Ïàò ÏûàÎã§.

$$\mu_t = Y_t - v_t$$

$$Y_t - v_t = Y_{t-1} - v_{t-1} + w_t$$

$$\Delta Y_t = Y_t - Y_{t-1} = w_t + v_t - v_{t-1}$$

Ïù¥ ÏãùÏóêÏÑú  Œîùëåùë° Ïùò Í∏∞ÎåìÍ∞íÍ≥º ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Íµ¨ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$\text{E} \left[ \Delta Y_t \right] = \text{E} \left[ w_t + v_t - v_{t-1} \right] =0$$

\begin{eqnarray}
\text{E} [ \Delta Y_t^2 ] 
&=& \text{E} \left[ (w_t + v_t - v_{t-1})(w_{t} + v_{t} - v_{t-1}) \right] \\
&=& \text{E} \left[ w_t^2 + v_t^2 + v_{t-1}^2 \right] \\
&=& \sigma_w^2 + 2\sigma_v^2 \\
\end{eqnarray}

\begin{eqnarray}
\text{E} [ \Delta Y_t \Delta Y_{t-1} ] 
&=& \text{E} \left[ (w_t + v_t - v_{t-1})(w_{t-1} + v_{t-1} - v_{t-2}) \right] \\
&=& - \text{E} \left[ v_{t-1}^2 \right] \\
&=& -\sigma_v^2 \\
\end{eqnarray}

Ïù¥ Í∞íÏóêÏÑú  Œîùëåùë° Îäî MA(1) Î™®ÌòïÏùÑ Îî∞Î•¥Í≥†  ùëåùë° Îäî ARIMA(0,1,1) Î™®ÌòïÏùÑ Îî∞Î•¥Îäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§. Ï¶â Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùÄ ARIMA(0,1,1)Ïùò Îòê Îã§Î•∏ ÌëúÌòÑ(representation)Ïù¥ÎùºÍ≥† Î≥º Ïàò ÏûàÎã§.

## ARIMA Î™®ÌòïÍ≥ºÏùò Ï∞®Ïù¥Ï†ê

ARIMA(0,1,1)Ïù¥ ÏïÑÎãå Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùÑ ÏÇ¨Ïö©ÌïòÎäî Ïù¥Ïú†Îäî Îã§ÏùåÍ≥º Í∞ôÏùÄ Ïù¥Ïú†ÏôÄ Í∞ÄÏ†ï ÎïåÎ¨∏Ïù¥Îã§.

+ Ïö∞Î¶¨Í∞Ä Í¥ÄÏã¨ÏùÑ Í∞ÄÏßÄÎäî Í∞íÏùÄ Ïñ¥Îñ§ Ïù¥Ïú†Ïóê ÏùòÌï¥ Î∞òÎìúÏãú ÎûúÎç§ ÏõåÌÅ¨ Î™®ÌòïÏùÑ Îî∞ÎùºÏïº ÌïúÎã§.
+ Í∑∏Îü¨ÎÇò ÏãúÍ≥ÑÏó¥ ÏûêÎ£å  ùëåùë° Îäî ÎûúÎç§ ÏõåÌÅ¨ Î™®ÌòïÏùÑ Îî∞Î•¥ÏßÄ ÏïäÎäîÎã§. (ARIMA(0,1,1) Î™®ÌòïÏùÑ Îî∞Î¶Ñ)
+ Í∑∏Îü¨ÎØÄÎ°ú ÏãúÍ≥ÑÏó¥ ÏûêÎ£å  ùëåùë° Îäî Ïö∞Î¶¨Í∞Ä ÏõêÌïòÎäî ÎûúÎç§ ÏõåÌÅ¨  ùúáùë° Ïóê Í¥ÄÏ∏° Ïû°ÏùåÏù¥ ÎçîÌï¥ÏßÑ Í±∏Î°ú Í∞ÄÏ†ïÌïòÏûê.
+ Ïù¥Ï†ú ÌíÄÏñ¥Ïïº ÌïòÎäî Î¨∏Ï†úÎäî ÏãúÍ≥ÑÏó¥ ÏûêÎ£å  ùëåùë° Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÎûúÎç§ ÏõåÌÅ¨ Î™®ÌòïÏùÑ Îî∞Î•¥Îäî, Ïû°ÏùåÏùÑ Ï†úÍ±∞Ìïú ÏõêÎûòÏùò Í∞í  ùëöùë¢ùë° Ïù¥Îã§.

## Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùò Ïòà 1

ÏòàÎ•º Îì§Ïñ¥ Ïñ¥Îñ§ ÏÑ¨ Í∑ºÏ≤òÏóêÏÑú Î∞∞Í∞Ä ÏóîÏßÑÏùÑ ÎÅÑÍ≥† Ï†ïÏßÄÌï¥ ÏûàÎã§. Î¨ºÎ¶¨ÌïôÏ†ÅÏù∏ ÏßÄÏãùÍ≥º Í∞ÄÏ†ïÏóê Îî∞Îùº Ïù¥ Î∞∞Îäî ÌååÎèÑÏóê ÏùòÌïú Î∏åÎùºÏö¥ Ïö¥Îèô(Brown montion)ÏùÑ ÌïòÍ≤å ÎêúÎã§. Ï¶â, ÌååÎèÑÏùò ÎûúÎç§Ìïú ÌûòÏóê Î∞ÄÎ†§ Î∞∞Ïùò ÏúÑÏπò  ùúáùë° Í∞Ä Ï∞®Ï∏∞ Î≥ÄÌïòÎ©∞ Í∑∏ Î≥ÄÌôîÎäî Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏàòÏãùÏùÑ Îî∞Î•∏Îã§Í≥† Í∞ÄÏ†ïÌïúÎã§.

$$\mu_t = \mu_{t-1} + w_t$$

Ïù¥ ÏãùÏóêÏÑú  ùë§ùë° Îäî ÏãúÍ∞Ñ  ùë° ÏóêÏÑú ÌååÎèÑÏùò ÏòÅÌñ•Ïóê ÏùòÌïú ÏõÄÏßÅÏûÑÏù¥Î©∞ Í∞ÄÏö∞ÏãúÏïà Ï†ïÍ∑ú Î∂ÑÌè¨Î•º Îî∞Î•∏Îã§Í≥† Í∞ÄÏ†ïÌïúÎã§.

Îß§ ÏãúÍ∞ÑÎßàÎã§ Î∞∞Ïùò ÏÑ†ÏõêÏùÄ ÏÑ¨ÏúºÎ°úÎ∂ÄÌÑ∞Ïùò Í±∞Î¶¨  ùëåùë° Î•º Î∞∞Ïóê ÏûàÎäî Ï¥àÏùåÌåå Í±∞Î¶¨Í≥Ñ(ultrasonic range finder)Î°ú Ï∏°Ï†ïÌïúÎã§. Ï¥àÏùåÌåå Í±∞Î¶¨Í≥ÑÎ°ú Ïû∞ Í∞í  ùëåùë° Îäî Í∞ÄÏö∞ÏãúÏïà Ï†ïÍ∑ú Î∂ÑÌè¨Ïóê Îî∞Î•¥Îäî Ïò§Ï∞®Î•º Í∞ÄÏßÄÍ≥† ÏûàÎã§Îäî Í≤ÉÏùÑ Ï†úÏ°∞ÏÇ¨ Ï†ïÎ≥¥Î°úÎ∂ÄÌÑ∞ ÏïåÍ≥† ÏûàÎã§.

#### Ïù¥Îü¨Ìïú Í≤ΩÏö∞Ïóê Ï†ïÌôïÌïú Í±∞Î¶¨  ùúáùë° Î•º ÏïåÍ≥†Ïûê ÌïòÎäî Í≤ÉÏù¥ Î∞îÎ°ú Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùò ÌïÑÌÑ∞ÎßÅ(filtering) Î¨∏Ï†úÍ∞Ä ÎêúÎã§.

 + ùëåùë° : ÏÑ¨ÏúºÎ°úÎ∂ÄÌÑ∞Ïùò Í±∞Î¶¨
 + ùúáùë° : ÌååÎèÑÏùò ÎûúÎç§Ìïú ÌûòÏóê Î∞ÄÎ†§ Î∞∞Ïùò ÏúÑÏπò Î≥ÄÌôî
 + ùë§ùë° : ÏãúÍ∞Ñ tÏóêÏÑú ÌååÎèÑÏùò ÏòÅÌñ•Ïóê ÏùòÌïú ÏõÄÏßÅÏûÑ

## Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùò Ïòà 2

Ìö®Ïú®Ï†ÅÏù∏ ÏãúÏû•ÏóêÏÑú Í±∞ÎûòÎêòÎäî Ï£ºÏãùÏùò Í∞ÄÍ≤©, Ï¶â Ï£ºÍ∞ÄÎäî ÏùºÎ∞òÏ†ÅÏúºÎ°ú ÎûúÎç§ÏõåÌÅ¨Î•º Îî∞Î•¥Îäî Í≤ÉÏúºÎ°ú ÏÉùÍ∞ÅÌï† Ïàò ÏûàÎã§.

#### Ïã§Ï†ú Ï£ºÍ∞ÄÏùò ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Ï∏°Ï†ïÌïòÎ©¥ ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÍ∞Ä ÏãúÏ∞®(lag)Í∞Ä 0Ïù¥ ÏïÑÎãå Í≤ΩÏö∞ÏóêÎèÑ Ï°¥Ïû¨Ìï† Ïàò ÏûàÎã§.
#### Ïù¥Î•º ÏãúÏû• ÎØ∏ÏãúÍµ¨Ï°∞ Ïû°Ïùå(market microstructure noise)ÎùºÍ≥† ÌïòÎ©∞ Ï£ºÍ∞ÄÏùò ÏãúÍ∞Ñ ÏÉòÌîåÎßÅ Í∞ÑÍ≤©Ïù¥ ÏûëÏùÑÏàòÎ°ù, ÌäπÌûà ÏùºÏ§ë(intraday) Í±∞ÎûòÏùò Í≤ΩÏö∞Ïóê ÎëêÎìúÎü¨ÏßÑÎã§. 

Ïù¥Îäî Ï£ºÏãùÏùò Ïú†ÎèôÏÑ± Î∞è Í±∞Îûò Íµ¨Ï°∞ÏôÄ Í¥ÄÎ†®ÌïòÏó¨ ÎÇòÌÉÄÎÇòÎäî ÌäπÏÑ±ÏúºÎ°ú Î≥º Ïàò ÏûàÎã§.

Ïù¥Îü¨Ìïú ÏãúÏû• ÎØ∏ÏãúÍµ¨Ï°∞ Ïû°ÏùåÏù¥ ÏûàÎäî Ï£ºÍ∞ÄÎ°úÎ∂ÄÌÑ∞ Í∑ºÎ≥∏Ï†ÅÏù∏ Ï£ºÏãùÏùò ÎÇ¥Ïû¨ Í∞ÄÏπò Ï¶â, ÎûúÎç§ ÏõåÌÅ¨Î•º Îî∞Î•¥Îäî Í∞ÄÏπòÎ•º Ï∂îÏ†ïÌïòÍ≥†Ïûê ÌïòÎäî Í≤ΩÏö∞ÎèÑ Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùò ÌïÑÌÑ∞ÎßÅ Î¨∏Ï†úÎ°ú Î≥º Ïàò ÏûàÎã§.

## Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùò ÏãúÎÆ¨Î†àÏù¥ÏÖò

statsmodelsÏùò KalmanFilter ÌÅ¥ÎûòÏä§ÏôÄ UnobservedComponents ÌÅ¥ÎûòÏä§Î•º ÏÇ¨Ïö©ÌïòÏó¨ Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùÑ Îî∞Î•¥Îäî ÏãúÍ≥ÑÏó¥ÏùÑ ÎßåÎì§Í≥† ÌïÑÌÑ∞ÎßÅÏùÑ Ìï¥Î≥¥Ïûê.

Ïö∞ÏÑ† KalmanFilter ÌÅ¥ÎûòÏä§Î°ú Ïù¥Î°†Ï†ÅÏù∏ Î°úÏª¨ Î†àÎ≤® Î™®ÌòïÏùÑ ÎßåÎì†Îã§.

```python
from statsmodels.tsa.statespace.kalman_filter import KalmanFilter

model1 = KalmanFilter(k_states=1, k_endog=1, transition=[[1]], selection=[[1]], state_cov=[[1]],
                     design=[[1]], obs_cov=[[1]])
```

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
np.random.seed(0)

y1, x1 = model1.simulate(100)

plt.plot(y1, "r:", label="Í¥ÄÏ∏°Í∞í")
plt.plot(x1, "g-", label="ÏÉÅÌÉúÍ∞í")
plt.legend()
plt.title("Î°úÏª¨Î†àÎ≤® Î™®Ìòï ($\sigma_w^2 = 1$, $\sigma_v^2 = 1$)")
plt.show()
```

Ïù¥ Î™®ÌòïÏùÄ Ïù¥ÎÖ∏Î≤†Ïù¥ÏÖò Í≥ºÏ†ïÏùò Í≥µÎ∂ÑÏÇ∞  ùúé^2ùë§ ÏôÄ Í¥ÄÏ∏° Ïû°ÏùåÏùò Í≥µÎ∂ÑÏÇ∞  ùúé^2ùë£ Í∞Ä Í∞ÅÍ∞Å  ùúé^2ùë§=1 ,  ùúé^2ùë£=1 Ïù∏ Í≤ΩÏö∞Ïù¥Îã§. Ïù¥ Í∞íÏù¥ Îã¨ÎùºÏßÄÎ©¥ Ïñ¥ÎñªÍ≤å ÎêòÎäîÏßÄ ÏÇ¥Ìé¥Î≥¥Ïûê.

```python
# state_cov : Ïù¥ÎÖ∏Î≤†Ïù¥ÏÖò Í≥ºÏ†ïÏùò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨ÏùÑ [[10]] Ìïú Í≤ΩÏö∞

model2 = KalmanFilter(k_states=1, k_endog=1, transition=[[1]], selection=[[1]], state_cov=[[10]],
                     design=[[1]], obs_cov=[[1]])
```

```python
np.random.seed(0)

y2, x2 = model2.simulate(100)
plt.plot(y2, "r:", label='Í¥ÄÏ∏°Í∞í')
plt.plot(x2, "g-", label="ÏÉÅÌÉúÍ∞í")
plt.legend()
plt.title("Î°úÏª¨Î†àÎ≤®Î™®Ìòï ($\sigma_w^2 = 10$, $\sigma_v^2 = 1$)")
plt.show()
```

Ïù¥ÎÖ∏Î≤†Ïù¥ÏÖò Í≥ºÏ†ïÏùò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨Ïù¥ Ïª§Ïßà ÏàòÎ°ù ÏÉÅÌÉúÍ∞í ÏòàÏ∏°Ïù¥ Ïûò ÎêòÍ≥† ÏûàÏùåÏùÑ Ïïå Ïàò ÏûàÎã§.

```python
# obs_cov : Í¥ÄÏ∏° ÌñâÎ†¨Ïùò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨ [[10]] Í≤ΩÏö∞

model3 = KalmanFilter(k_states=1, k_endog=1, transition=[[1]], selection=[[1]], state_cov=[[1]],
                     design=[[1]], obs_cov=[[10]])

np.random.seed(0)
y3, x3 = model3.simulate(100)
plt.plot(y3, "r:", label="Í¥ÄÏ∏°Í∞í")
plt.plot(x3, "g-", label="ÏÉÅÌÉúÍ∞í")
plt.legend()
plt.title("Î°úÏª¨Î†àÎ≤®Î™®Ìòï ($\sigma_w^2 = 1$, $\sigma_v^2 = 10$)")
plt.show()
```

Í¥ÄÏ∏° ÌñâÎ†¨Ïùò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨Ïù¥ Ïª§ÏßàÏàòÎ°ù Ïò§Ï∞®Í∞Ä Ïª§ÏßÑÎã§Îäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§.

### ÏπºÎßå ÌïÑÌÑ∞ÎßÅÏùÑ Ïù¥Ïö©Ìïú ÏÉÅÌÉúÎ≥ÄÏàò Ï∂îÏ†ï

Ïù¥Ï†ú ÏπºÎßå ÌïÑÌÑ∞ÎßÅÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏÉÅÌÉúÎ≥ÄÏàòÎ•º Ï∂îÏ†ïÌïòÏó¨ Î≥¥Ïûê. UnobservedComponents ÌÅ¥ÎûòÏä§Î•º ÏÇ¨Ïö©ÌïòÏó¨ Î™®Ìòï Ï∂îÏ†ïÍπåÏßÄ Ìïú Î≤àÏóê ÏàòÌñâÌï† Ïàò ÏûàÎã§.

```python
import statsmodels.api as sm

model3f = sm.tsa.UnobservedComponents(y3, 'local level')  # ÏÉÅÌÉúÏ∂îÏ†ï + Î™®ÌòïÏ∂îÏ†ï
res3f = model3f.fit()
print(res3f.summary())
```

```python
plt.plot(y3, "r:", label="Í¥ÄÏ∏°Í∞í")
plt.plot(x3, "g-", label="ÏõêÎûò ÏÉÅÌÉúÍ∞í")
plt.plot(res3f.filtered_state[0], "b-", label="ÏÉÅÌÉúÍ∞íÏùò Ï∂îÏ†ïÏπò")
plt.legend(loc=1)
plt.title("Kalman Filtering of Local Level Model")
plt.show()
```

```python
res3f.plot_components(legend_loc="lower right", figsize=(8, 10))
plt.tight_layout(pad=3, h_pad=0.8)
plt.show()
```

```python
# Î™®ÌòïÏùò ÏßÑÎã® : Ï†ïÍ∑úÏÑ±, ÏûêÍ∏∞ÏÉÅÍ¥ÄÍ≥ÑÏàò
res3f.plot_diagnostics()
plt.tight_layout()
plt.show()
```

```python
# ÏòàÏ∏°
forecast = res3f.get_forecast(50)

plt.plot(y3, label="Í¥ÄÏ∏°Í∞í")
forecast_ci = forecast.conf_int()
forecast_index = np.arange(len(y1), len(y1) + len(forecast_ci)) # 100 ~ 150
plt.plot(forecast_index, forecast.predicted_mean, label="ÏòàÏ∏°Í∞í")
plt.fill_between(forecast_index, forecast_ci[:, 0], forecast_ci[:, 1], alpha=0.1)
plt.show()
```
